```{div} full-width

In this lecture, we will discuss the deformable Earth, denote with $\bb{u}$, $\rho^E$ and $\phi^E$ its displacement field, density distribution and gravitational potential and adopt the translating and rotating reference system discussed in {numref}`Section %s <sec-reference>` - {ref}`sec-reference`. In this respect, we will adopt a spherical a spherical reference frame centred at the translating origin $\bb{X}_O$ and defined with respect to the rotating unit vectors $\bb{e}_i$

$$
\bb{x} = x_i\,\bn{e}_i(t) = x\,\bn{r}(\vartheta,\varphi,t)
$$ (E:1)

with

$$
\bn{r}(\vartheta,\varphi,t) = \sin\vartheta\,\big(\cos\varphi\,\bn{e}_1(t)+\sin\varphi\,\bn{e}_2(t)\big) + \sin\vartheta\,\bn{e}_3(t)
$$ (E:2)

The spherical unit vectors along the meridians, $\sn{\vartheta}$, and parallels, $\sn{\varphi}$, will be also defined with respect to the rotating unit vectors, as well as the spherical harmonic vectors $\bb{P}_n^m$, $\bb{B}_n^m$ and $\bb{C}_n^m$.

Due to the instantaneous relatation between the density distribution and the gravitational potential, all the results discussed in {numref}`Section %s <chap-potentials>` - {ref}`chap-potentials` can be written by (i) adding the time depencence to these two fields, (ii) replacing the Cartesian unit vectors with the rotating ones and (iii) taking into account the translating origin of the reference system. 

```

```{admonition} Mass conservation
:class: full-width

The mass contained with a time-dependent region $\mathcal{V}_t$ of a deformable continuum body evolving according to the displacement field  $\bb{u}$ can be obtained as

$$
M = \int_{\mathcal{V}_t} \bar{\rho}(\bb{r},t)\,\md \bar{V}
$$ (E:3)

where $\md \bar{V}=\md r_1\,\md r_2\,\md r_3$ being the infinitesimal volume element. By noting that

$$
\md \bar{V} = J(\bb{x},t)\,\md V
$$ (E:4)

with $\md V=\md x_1\,\md x_2\,\md x_3$ and $J$ being the Jacobian of the transformation $\bb{r}\rightarrow \bb{x}$

$$
J(\bb{x},t) = \mathrm{det}\big(\bb{J}(\bb{x},t)\big) \approx 1+\grad\bb{u}(\bb{x},t) \qquad \mathrm{with}\qquad \bb{J}(\bb{x},t) = \grad\bb{r} = \one + \grad\bb{u}(\bb{x},t)
$$ (E:5)

where the approximation holds at the first-order approximation. In light of this, after the change of variable, eq. {eq}`E:3` becomes 

$$
M = \int_{\mathcal{V}_0} \rho(\bb{x},t)\,J(\bb{x},t)\,\md V
$$ (E:6)

In this respect and requiring the mass conservation of any deformed region system, we thus understand that the product between $\rho$ and $J$ corresponds to the initial density distribution $\rho_0$ that, by definition, does not depend on time

$$
\rho_0(\bb{x}) =  \rho(\bb{x},t)\,J(\bb{x},t)
$$ (E:8)

This interpretation is correct also in view of the fact that $J$ represent the relative volume variations of the infinitesimal volume element of the deformable continuum body.
```

## Centre of mass and orbital motion

Let us now consider the mass and centre of mass of the Earth system

$$
M = \int \bar{\rho}(\bb{r},t)\,\md \bar{V} = \int \rho_0(\bb{x})\,\md V
$$
$$
M\,\bb{Z} = \int \bar{\rho}(\bb{r},t)\,\bb{R}(\bb{r},t)\,\md \bar{V} = M\,\bb{X}_O + \bb{C}
$$

with $\bb{C}$ being the first moment of the density distribution with respect to the translating origin

$$
\bb{C} = \int \bar{\rho}(\bb{r},t)\,\bb{r}\,\md \bar{V} = \int \rho_0(\bb{x})\,\bb{r}\,\md V
$$

```{admonition} Density first moment time derivatives
:class: full-width

We note that

$$
\dercla{\bb{C}} = \int \rho_0(\bb{x})\,\big(\sb{\omega}\times\bb{r}+\bb{v}(\bb{r},t)\big)\,\md V = \sb{\omega}\times\bb{C} + \derpar{\bb{C}}
$$

with

$$
\derpar{\bb{C}} = \int \bar{\rho}(\bb{r},t)\,\bb{v}(\bb{r},t)\,\md \bar{V} = \int \rho_0(\bb{x})\,\bb{v}(\bb{r},t)\,\md V
$$

and that 

$$
\ddercla{\bb{C}} = \sb{\omega}\times\sb{\omega}\times\bb{C} + \derpar{\sb{\omega}}\times\bb{C} + 2\,\sb{\omega}\times\derpar{\bb{C}} + \dderpar{\bb{C}}
$$

with

$$
\dderpar{\bb{C}} = \int \bar{\rho}(\bb{r},t)\,\bb{a}(\bb{r},t)\,\md \bar{V} = \int \rho_0(\bb{x})\,\bb{a}(\bb{r},t)\,\md V
$$

```

### Linear momentum

The total linear momentum and its time derivative read

$$
\bb{P} = \int \bar{\rho}(\bb{r},t)\,\bb{V}(\bb{r},t)\,\md \bar{V} = M\,\dercla{\bb{X}_O} +\dercla{\bb{C}}
$$
$$
\dercla{\bb{P}} = \int \bar{\rho}(\bb{r},t)\,\bb{A}(\bb{r},t)\,\md \bar{V} = M\,\ddercla{\bb{X}_O} +\ddercla{\bb{C}}
$$

As first choice, we fix the origin so that

$$
\bb{Z} = \bb{X}_O \qquad \bb{C} = \zero
$$

and write the total linear momentum and its derivative simply as

$$
\bb{P} =  M\,\dercla{\bb{Z}}\qquad
\dercla{\bb{P}} = M\,\ddercla{\bb{Z}} 
$$

Then by considering that the total forces from the celestial body acting on nearly spherical density distribution of total mass $M$ reads

$$
\bb{F}_T = M\,\bb{g}_1
$$

from the second Newton law

$$
\ddercla{\bb{P}} = \bb{F}_T
$$

we obatin the acceleration of the centre of mass in the inertial frame

$$
\ddercla{\bb{Z}}  = \bb{g}_1^T
$$

### Solid earth and athmosphere

Let us now distinguish the density distribution $\rho$ into the solid Earth, $\rho^E$, and athmosphere, $\rho^A$, densities

$$
\rho(\bb{x},t) = \rho^E(\bb{x},t) + \rho^A(\bb{x},t)
$$

and write the centres of mass and the linear momentum of the solid Earth or the athmosphere

$$
M_X\,\bb{Z}_X = M_X\,\bb{Z}+\bb{C}_X
$$
$$
\bb{P}_X = M_X\,\dercla{Z} + \dercla{\bb{C}_X}
$$

with $X=E$ or $A$, and

$$
\bb{C}_X = \int \bar{\rho}^X(\bb{r},t)\,\bb{r}\,\md \bar{V}
$$

By definition, the following relation between the masses, centres of mass and linear momenta of the Earth system, of the solid Earth and of the athmosphere hold

$$
M_E + M_A = M
$$
$$
M_E\,\bb{Z}_E + M_A\,\bb{Z}_A = M\,\bb{Z}
$$
$$
\bb{P}_E + \bb{P}_A = \bb{P} 
$$

We note that, from eqs. , we can obtain

$$
\bb{C}_E = - \bb{C}_A
$$

Let us now consider the total force from the athmosphere on the Earth

$$
\bb{F}_{AE} =  \int_{\partial V_t}\bb{T}^A(\bb{r},t)\,\md S - \int  \bar{\rho}^E(\bb{r},t)\,\bar{\grad}\bar{\phi}^A(\bb{r},t)\,\md \bar{V}
$$

where $\phi^A$ being the athmosphere gravitational potential and $\bb{T}^A$ being the traction of the athmosphere acting on the solid Earth. Also, let us consider the total force from the celestial body on the solid Earth, $\bb{F}_{TE}$, and on the athmosphere, $\bb{F}_{TA}$, 

$$
\bb{F}_{TX} = M_X\,\bb{g}_1^T
$$

In view of the second Newton law applied to the solid Earth or athmosphere, we obatin

$$
\dercla{\bb{P}_E} = \bb{F}_{TE} + \bb{F}_{AE}  \qquad \dercla{\bb{P}_A} = \bb{F}_{TA} - \bb{F}_{AE} 
$$

and, so, 

$$
\ddercla{\bb{C}_E} = -\ddercla{\bb{C}_A} = \bb{F}_{AE} 
$$

Note that the total force from the solid Earth to the athmosphere used in eq. {eq}`EQ` is just $-\bb{F}_{AE}$.


## The choice of the angular velocity

Let us now consider the total linear momentum of the solid Earth or the athmosphere under the assumption that $\bb{Z}=\bb{X}_O$

$$
\begin{align}
\bb{H}_X(t) &= \int\bar{\rho}^X(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\bb{V}}(\bb{r},t)\,\md \bar{V} = \bb{X}_O(t)\times\bb{P} + \int\bar{\rho}^X(\bb{r},t)\,\bb{r}\times\bar{\bb{V}}(\bb{r},t)\,\md \bar{V}
\\
&= \bb{Z}(t)\times\bb{P}_X(t) + \bb{C}_X(t)\times\dercla{\bb{Z}(t)} + \int\bar{\rho}^X(\bb{r},t)\,\bb{r}\times\big(\sb{\omega}(t)\times\bb{r}+\bar{\bb{v}}(\bb{r},t)\big)\,\md \bar{V}
\\
&= \bb{Z}(t)\times\bb{P}_X(t) + \bb{C}_X(t)\times\dercla{\bb{Z}(t)} + \bb{I}_X(t)\cdot\sb{\omega}(t) + \bb{h}_X(t)
\end{align}
$$ (E:24)

with $\bb{I}_X$ and $\bb{h}_X$ being the inertia tensor of the Earth as defined in eq. {eq}`G:31`  and the relative angular momentum

$$
\bb{h}_X(t) = \int\bar{\rho}^X(\bb{r},t)\,\bb{r}\times\bar{\bb{v}}(\bb{r},t)\,\md \bar{V} =  \int \rho_0^X(\bb{x})\,\bb{r}\times\bb{v}(\bb{x},t)\,\md V 
$$ (E:25)

In order to define the angular velocity, we adopt the Tisserand reference frame in which the angular velocity $\sb{\omega}$ is choosen in such a way that relative angular momentum of the solid Earthn is zero

$$
\bb{h}_E(t) = \zero
$$


```{note}
:class: full-width, dropdown

This choice is always possible. Let us indeed consider the following angular velocity

$$
\sb{\omega}^*(t) = \sb{\omega}(t)+\mat\sb{\omega}(t)
$$

and denote with $\bb{e}^*_i$ the respective rotating unit vectors. Then, let us expresse the velocity in the two distinct rotating reference frame

$$
\bb{V}(\bb{x},t) = \dercla{\bb{Z}(t)} + \sb{\omega}^*(t)\times\bb{r} + \bb{v}^*(\bb{x},t) = \dercla{\bb{Z}(t)} + \sb{\omega}(t)\times\bb{r} + \bb{v}(\bb{x},t)
$$

We thus understand that

$$
\bb{v}^*(\bb{x},t) = \bb{v}(\bb{x},t) - \mat\sb{\omega}(t)\times\bb{r}
$$

and the relative angular momentum can be arrange as follows

$$
\bb{h}_E^*(t) = \int\bar{\rho}^E(\bb{r},t)\,\bb{r}\times\bar{\bb{v}}^*(\bb{r},t)\,\md \bar{V} =  \bb{h}_E(t) - \int \bar{\rho}^E(\bb{r},t)\,\bb{r}\times \mat\sb{\omega}(t)\times\bb{r}\,\md \bar{V} = \bb{h}_E(t) - \bb{I}_E(t)\cdot\mat\sb{\omega}(t)
$$ (E:25)

In this respect, we thus understand that it can be set to zero by properly choosing $\mat\sb{\omega}$

$$
\mat\sb{\omega}(t) = \bb{I}_E(t)^{-1}(t)\cdot\bb{h}_E(t) = \zero  \qquad\Rightarrow \qquad \bb{h}_E^*(t)
$$

and, so, there exist an angular velocity vector $\sb{\omega}^*$ such that the relative angular momentum is zero.

```

This choice of the angulra velocity $\sb{\omega}$ based on the solid Earth system, along with the choice of the translating origin $\bb{X}_O$ taken as the centre of mass of the Earth system (including the athmosphere), fully defines the translating and rotating reference system discussed in  {numref}`Chapter %s <chap-kinematics>` - {ref}`chap-kinematics`. 

### Conservation of the angular momentum

Within this framework, the angular momentum of the solid Earth becomes

$$
\bb{H}_E = \bb{Z}(t)\times\bb{P}_E(t) + \bb{C}_E(t)\times\dercla{\bb{Z}(t)} + \bb{I}_E(t)\cdot\sb{\omega}(t) 
$$

and its time derivative yields

$$
\begin{align}
\dercla{\bb{H}_E} &= \dercla{\bb{Z}(t)}\times\bb{P}_E(t) + \bb{Z}\times\dercla{\bb{P}_E} + \dercla{\bb{C}_E(t)}\times\dercla{\bb{Z}(t)} + \bb{C}_E\times\ddercla{\bb{Z}} + \dercla{\big(\bb{I}_E(t)\cdot\sb{\omega}(t) \big)} 
\\
&= \dercla{\bb{Z}(t)}\times\left(M_E\,\dercla{\bb{Z}} + \dercla{\bb{C}_E}\right) + \bb{Z}\times\left(M_E\,\bb{g}_1^T + \bb{F}_{AE}\right) + \dercla{\bb{C}_E(t)}\times\dercla{\bb{Z}(t)} + \bb{C}_E\times\ddercla{\bb{Z}} + \dercla{\big(\bb{I}_E(t)\cdot\sb{\omega}(t) \big)} 
\\
&= \big(M_E\,\bb{Z}+\bb{C}_E\big)\times\bb{g}_1^T + \bb{Z}\times\ddercla{\bb{C}_E} + \dercla{\big(\bb{I}_E(t)\cdot\sb{\omega}(t) \big)} 
\end{align}
$$

Let us now consider the torque from the athmosphere and from the celestial bodies acting on the solid Earth

$$
\begin{align}
\bb{K}_{AE} &= \int_{\partial V_t} \bar{\bb{R}}(\bb{r},t)\times\bb{T}^A(\bb{r},t)\,\md S - \int  \bar{\rho}^E(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\grad}\bar{\phi}^A(\bb{r},t)\,\md \bar{V} = \bb{Z}\times\bb{F}_{AE} + \bb{k}_{AE}
\\
&= \bb{Z}\times\ddercla{\bb{C}_E} + \bb{k}_A
\\
\bb{K}_{TX} &= -\int \bar{\rho}^X(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\grad}\bar{\phi}^T(\bb{r},t)\,\md \bar{V} = \big(M_X\,\bb{Z} +\bb{C}_X\big)\times\bb{g}_1 -\int \bar{\rho}^X(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\grad}\bar{\psi}^T(\bb{r},t)\,\md \bar{V}
\\
&= \big(M_X\,\bb{Z} +\bb{C}_X\big)\times\bb{g}_1^T + \bb{Z}\times\int \bar{\rho}^X(\bb{r},t)\,\bar{\grad}\bar{\psi}^T(\bb{r},t)\,\md\bar{V} + \int \bar{\rho}^X(\bb{r},t)\,\bb{r}\times\bar{\grad}\bar{\psi}^T(\bb{r},t)\,\md \bar{V}
\\
&= \big(M_X\,\bb{Z} +\bb{C}_X\big)\times\bb{g}_1^T 
\end{align}
$$

where, in the last identities, we neglect the remaining volume integrals in view of eq. and we define the relative torque from the athmosphere with respect to the centre of mass of the Earth system

$$
\bb{k}_{AE} = \int_{\partial V_t} \bb{r}\times\bb{T}^A(\bb{r},t)\,\md S - \int  \bar{\rho}^E(\bb{r},t)\,\bb{r}\times\bar{\grad}\bar{\phi}^A(\bb{r},t)\,\md \bar{V} 
$$

In light of this, the conservation of the angular momentum of the solid Earth system

$$
\dercla{\bb{H}_E(t)} = \bb{K}_{TE}(t) + \bb{K}_{AE}(t)
$$

yields the so called Euler rotational equation

$$
\dercla{\big(\bb{I}_E(t)\cdot\sb{\omega}(t) \big)}  = \derpar{\big(\bb{I}_E(t)\cdot\sb{\omega}(t) \big)} + \sb{\omega}\times\big(\bb{I}_E(t)\cdot\sb{\omega}(t) \big) = \bb{k}_{AE}(t)
$$

We note that the relative torques of the athmosphere can thus change the angular momentum of the solid Earth. Different, the relative torque from celestial bodies does not enter the above equation because, within the assumption of infinitesimal departure from the spherical symmetry and as already pointed out by eq. , it is zero at the first order.

### Initial and incremental Poisson equations

```{div} full-width

Let us also consider the decomposition of the Earth gravitational potential into its initial and local increment

$$
\bar{\phi}^E(\bb{r},t) = \phi_0^E(\bb{r}) + \loc\bar{\phi}^E(\bb{r},t)
$$ (E:12)

By definition, the integral and differential equations governing the initial gravitational potential read

$$
\begin{align}
& \phi_0^E(\bb{r}) = -G\,\int \frac{\varrho_0(\dif{\bb{r}})}{\|\bb{r}-\dif{\bb{r}}\|}\,\md \dif{\bar{V}} \qquad  \bar{\nabla}^2 \phi_0^E(\bb{r}) = C\,\varrho_0(\bb{r}) 
\\
& \phi_0^E(\bb{x})  = -G\,\int \frac{\varrho_0(\dif{\bb{x}})}{\|\bb{x}-\dif{\bb{x}}\|}\,\md \dif{V}  \qquad  \lap \phi_0^E(\bb{x}) = C\,\varrho_0(\bb{x})
\end{align}
$$ (E:13)

For the local incremental gravitational potential, instead, we have the following integral identities

$$
\begin{align}
& \loc\bar{\phi}^E(\bb{r},t) = \loc\phi^E(\bb{x},t) = -G\,\int \frac{\loc\bar{\rho}^E(\dif{\bb{r}},t)}{\|\bb{r}-\dif{\bb{r}}\|}\,\md \dif{\bar{V}} \approx -G\,\int \frac{\loc\rho^E(\dif{\bb{x}},t)}{\|\bb{x}-\dif{\bb{x}}\|}\,\md \dif{V} 
\end{align}
$$ (E:14a)

and the following incremental Poisson equations

$$
\begin{align}
& \bar{\nabla}^2 \loc\bar{\phi}^E(\bb{r},t) = C\,\loc\bar{\rho}^E(\bb{r},t) \qquad \nabla^2 \loc\phi^E(\bb{x},t) \approx C\,\loc\rho^E(\bb{x},t)  
\end{align}
$$ (E:14)

Here we have made use of the fact that the gradient or the Laplacian of first-order terms in the Eulerian and Lagrangian frameworks differ only by second-order terms, eqs. {eq}`O:31` and {eq}`O:32`.
```

### Mass and centre of mass

```{div} full-width

Let us now consider the mass $M_E$ of the Earth system

$$
M_E = \int\bar{\rho}^E(\bb{r},t)\,\md \bar{V} = \int \varrho_0(x)\,\md V = 4\,\pi\,\int \varrho_0(x)\,x^2\,\md x
$$ (E:17)

Different, by making use of the local decomposition, it reads

$$
\begin{align}
M_E &= \int\bar{\rho}^E(\bb{r},t)\,\md \bar{V}  = \int \varrho_0(r)\,\md \bar{V} + \int \loc\bar{\rho}^E(\bb{r},t)\,\md\bar{V} \approx M + \int_0^{a^+} \loc\rho^E(\bb{x},t)\,\md V 
\\
&= M +  4\,\pi\,\int \loc\rho^E_{0,0}(x)\,x^2\,\md x
\end{align}
$$ (E:18)

and, so, from the comparison between eqs. {eq}`E:17` and {eq}`E:18`, we obtain the following constraint on the degree-0 local incremental density 

$$
4\,\pi\,\int \loc\rho^E_{0,0}(x)\,x^2\,\md x \approx 0
$$ (E:19)

As it concerns the centre of mass $\bb{Z}_E$, making use of eqs. {eq}`O:12`, it reads

$$
M_E\,\bb{Z}_E(t) = \int \bar{\rho}^E(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\,\md \bar{V} = \int \varrho_0(\bb{x})\,\bb{R}(\bb{x},t)\,\md \bar{V}  = M_E\,\bb{X}_O(t) + \bb{C}_E(t) 
$$

with

$$
\bb{C}_E(t) = \int \bar{\rho}^E(\bb{r},t)\,\bb{r}\,\md \bar{V}  =  \int \varrho_0(\bb{r})\,\bb{r}\,\md V
$$ (E:20)

Furthermore, in view of eq. {eq}`G:28`, we note that

$$
\begin{align}
\bb{C}_E(t) &= \int \bar{\rho}^E(\bb{r},t)\,\bb{r}\,\md \bar{V} = \int \bar{\rho}_0^E(r)\,\bb{r}\,\md \bar{V} + \int \loc\bar{\rho}^E(\bb{r},t)\,\bb{r}\,\md \bar{V} \approx \int \loc\rho^E(\bb{x},t)\,\bb{x}\,\md V 
\\
&= -\frac{\sqrt{3}\,a^2}{G}\sum_m \bb{p}_m\,\loc\phi^E_{1,m}(a,t)
\end{align}
$$ (E:21)

and, so, the degree-1 local incremental gravitational potential describes a shift of the centre of mass with respect to the translating origin $\bb{X}_O$.

```

### Linear momentum

```{div} full-width

Let us now consider the total linear momentum of the Earth 

$$
\begin{align}
 \bb{P}_E(t) &= \int \bar{\rho}^E(\bb{r},t)\,\bar{\bb{V}}(\bb{r},t)\,\md \bar{V} \\
 &= \int \varrho_0(\bb{x})\,\bb{V}(\bb{x},t)\,\md V = M_E\,\dercla{\bb{X}_O(t)} + \sb{\omega}(t)\times\bb{C}_E(t) + \derpar{\bb{C}_E(t)}
\end{align}
$$ (E:22a)

and its time derivative

$$
\begin{align}
\dercla{\bb{P}_E(t)} &= \int \bar{\rho}^E(\bb{r},t)\,\bar{\bb{A}}(\bb{r},t)\,\md \bar{V} = \int \varrho_0(\bb{x})\,\bb{A}(\bb{x},t)\,\md \bar{V} \\
&= M_E\,\ddercla{\bb{X}_O(t)} + \sb{\omega}(t)\times\sb{\omega}(t)\times\bb{C}_E(t) + 2\,\sb{\omega}(t)\times\derpar{\bb{C}_E(t)} + \derpar{\sb{\omega}(t)}\times\bb{C}_E(t) + \dderpar{\bb{C}_E(t)}
\end{align}
$$ (E:22)

with

$$
\begin{align}
& \derpar{\bb{C}_E(t)} = \int \bar{\rho}^E(\bb{r},t)\,\bar{\bb{v}}(\bb{r},t)\,\md \bar{V} = \int \varrho_0(\bb{x})\,\bb{v}(\bb{x},t)\,\md V
\end{align}
$$ (E:23a)
$$
\begin{align}
& \dderpar{\bb{C}_E(t)} = \int \bar{\rho}^E(\bb{r},t)\,\bar{\bb{a}}(\bb{r},t)\,\md \bar{V} = \int \varrho_0(\bb{x})\,\bb{a}(\bb{x},t)\,\md V
\end{align}
$$ (E:23)

```

### Angular momentum

```{div} full-width

Let us now consider the total linear momentum of the Earth

$$
\begin{align}
\bb{H}_E(t) &= \int\bar{\rho}^E(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\bb{V}}(\bb{r},t)\,\md \bar{V} = \bb{X}_O(t)\times\bb{P}_E + \int\bar{\rho}^E(\bb{r},t)\,\bb{r}\times\bar{\bb{V}}(\bb{r},t)\,\md \bar{V}
\\
&= \bb{X}_O(t)\times\bb{P}_E(t) + \bb{C}_E(t)\times\dercla{\bb{X}_O(t)} + \int\bar{\rho}^E(\bb{r},t)\,\bb{r}\times\big(\sb{\omega}(t)\times\bb{r}+\bar{\bb{v}}(\bb{r},t)\big)\,\md \bar{V}
\\
&= \bb{X}_O(t)\times\bb{P}_E(t) + \bb{C}_E(t)\times\dercla{\bb{X}_O(t)} + \bb{I}_E(t)\cdot\sb{\omega}(t) + \bb{h}_E(t)
\end{align}
$$ (E:24)

with $\bb{I}_E$ and $\bb{h}_E$ being the inertia tensor of the Earth as defined in eq. {eq}`G:31`  and the relative angular momentum

$$
\bb{h}_E(t) = \int\bar{\rho}^E(\bb{r},t)\,\bb{r}\times\bar{\bb{v}}(\bb{r},t)\,\md \bar{V} =  \int \varrho_0(\bb{x})\,\bb{r}\times\bb{v}(\bb{x},t)\,\md V  \approx  \int \varrho_0(\bb{x})\,\bb{x}\times\bb{v}(\bb{x},t)\,\md V
$$ (E:25)

We note that, in view of the local decomposition of the Earth density distribution, we can write

$$
\bb{I}_E(t) = \big(I_E + \loc I_E(t)\big)\,\one + \bb{D}_E(t)
$$ (E:26)

where $I_E$ is the initial (scalar) inertia of the Earth, $\loc  I_E$ is its perturbation and $\bb{D}_E$ being the deviatoric inertia tensor

$$
I_E = \frac{8\,\pi}{3}\,\int \rho_{0}^E(x)\,x^4\,\md x \qquad \loc I_E(t) = \frac{8\,\pi}{3}\,\int \loc\rho_{0,0}^E(x,t)\,x^4\,\md x \qquad \bb{D}_E = - \frac{5\,a^3}{G}\,\sum_m \loc\phi^E_{2,m}(a)\,\bb{Q}_m
$$ (E:27)

with $\bb{Q}_m$ given by eq. {eq}`G:34` with the Cartesian unit vector $\bn{x}_i$ replaced by the rotating unit vectors $\bb{e}_i$.

```

### Definition of the Earth reference system

```{div} full-width

So far the translating origin $\bb{X}_O$ and the angular velocity $\sb{\omega}$ have not been specified. According to the results of the previous section, we can choose $\bb{X}_O$ to coincide with the centre of mass of the Earth and $\sb{\omega}$ in such a way that the relative angular momentum is zero

$$
\bb{X}_O(t) = \bb{Z}_E(t) \qquad 
\bb{h}_E(t) = \zero
$$ (E:29)

Within this framework, from eq. {eq}`E:21`, we thus have 

$$
\bb{C}_E(t) = \zero
\qquad
\loc\phi^E_{1,m}(a,t) = 0
$$ (E:30)

while eqs. {eq}`E:22` and {eq}`E:24` becomes

$$
\begin{align}
& \bb{P}(t) =  M_E\,\dercla{\bb{X}_O(t)} 
\qquad
\dercla{\bb{P}_E(t)} = M_E\,\ddercla{\bb{X}_O(t)} 
\\
& \bb{H}_E(t) = M_E\,\bb{Z}_E(t)\times\dercla{\bb{Z}_E(t)} + \bb{I}_E(t)\cdot\sb{\omega}(t) 
\end{align}
$$ (E:31)

```

```{admonition} The athmosphere
:class: full-width

This choice of the reference system is based only on the Earth system and, in particular, on its solid part leaving away the athmosphere. By considering that the total mass of the athmosphere is about six order of magnitude smaller than the total Earth mass, we can consider the athmosphere as first-order term and, in principle, its centre of mass can be different from that of the solid part. In this respet, the athmosphere's centre of mass, $\bb{Z}_A$, does not necessarily coincide with that of the solid Earth

$$
\begin{align}
& \bb{Z}_A(t) =  \bb{Z}_E(t) + \frac{\bb{C}_A(t)}{M_A}
\end{align}
$$

Also, the athmosphere's linear momentum and its time derivative reads

$$
\begin{align}
& \bb{P}_A(t) =  M_A\,\dercla{\bb{Z}_E(t)} + \sb{\omega}(t)\times\bb{C}_A(t) + \derpar{\bb{C}_A(t)}
\\
& \dercla{\bb{P}_A(t)} = M_A\,\ddercla{\bb{Z}_E(t)} + \sb{\omega}(t)\times\sb{\omega}(t)\times\bb{C}_A(t) + 2\,\sb{\omega}(t)\times\derpar{\bb{C}_A(t)} + \derpar{\sb{\omega}(t)}\times\bb{C}_A(t) + \dderpar{\bb{C}_A(t)}
\end{align}
$$

while its angular momentum read

$$
\begin{align}
& \bb{H}_A(t) = \bb{Z}_E(t)\times\bb{P}_A(t) + \bb{C}_A(t)\times\dercla{\bb{Z}_E(t)} + \bb{I}_A(t)\cdot\sb{\omega}(t) + \bb{h}_A(t)
\end{align}
$$ (E:32)

where $M_A$, $\bb{I}_A$ and $\bb{h}_A$ being its total mass, inertia tensor and relative angular momentum.
```

### Earth's orbital motion

```{div} full-width

Let us denote with $\bb{Z}_T$ the centre of mass of the celestial body and, according to {numref}`Section %s <sec-celestial>` - {ref}`sec-celestial`, define its relative position with respect to the Eart centre of mass $\bb{Z}_E$ as follows

$$
\bb{x}_T(t) = \bb{Z}_T(t) - \bb{Z}_E(t) = x_T(t)\,\bn{r}(\vartheta_T(t),\varphi_T(t))
$$ (E:33)

By considering any departure from the spherical symmetry of the Earth in the adopted reference frame as a perturbation which contribute only with second-order terms to the total force from the the celestial body, eqs. {eq}`G:68`  and {eq}`G:69`, we can write

$$
\dercla{\bb{P}_E(t)} + \dercla{\bb{P}_A(t)} = M\,\bb{g}_1^T(t) \qquad \mathrm{with}\qquad M = M_E+M_A
$$ (E:34)

By considering, instead, the total force acting on the celestial body due to the Earth and the athmosphere, we obatin

$$
\dercla{\bb{P}_T(t)} = M_T\,\ddercla{\bb{Z}_T(t)} = -M\,\bb{g}_1^T(t)
$$ (E:35)

By making use of eqs. {eq}`E:22` and {eq}`E:23`, eq. {eq}`E:34` becomes

$$
M\,\ddercla{\bb{Z}_E(t)} + \sb{\omega}(t)\times\sb{\omega}(t)\times\bb{C}_A(t) + 2\,\sb{\omega}(t)\times\derpar{\bb{C}_A(t)} + \derpar{\sb{\omega}(t)}\times\bb{C}_A(t) + \dderpar{\bb{C}_A(t)} = M\,\bb{g}_1^T(t)
$$ (E:37)

In this respect and further considering the terms related to centrifugal, Coriolis and Euler accelerations as first-order terms, as well as the departure of centre of mass of the athmosphere $\bb{C}_A$, it becomes, correct to the first-order, the following equation

$$
M\,\ddercla{\bb{Z}_E(t)} = M\,\bb{g}_1^T(t)
$$ (E:38)

that, toghether with eq. {eq}`E:35`, defines the classical two-body problem for determining the orbital motions of the Earth and of the celestial body resulting into the classical elliptical orbits as predicted by the Keplero's theory for two massive body with total mass $M_T$ and $M=M_E+M_A$. For our purpose, however, it is sufficient to note that eq. {eq}`E:38` implies

$$
\ddercla{\bb{Z}_E(t)} = \bb{g}_1^T(t)
$$ (E:39)

```

### Total torque and conservation of the angular momentum

Because the torque depend on the origin with respect it is calculated,  the torque due to the celestial body on the Earth in the inertial frame here considered reads

$$
\bb{F}_{TE} =  \int  \bar{\rho}^E(\bb{r},t)\,\bar{\grad}\bar{\phi}^T(\bb{r},t)\,\md \bar{V} = M_E\,\bb{g}_1^T
$$
$$
\bb{F}_{TA} =  \int  \bar{\rho}^A(\bb{r},t)\,\bar{\grad}\bar{\phi}^T(\bb{r},t)\,\md \bar{V} = M_A\,\bb{g}_1^T
$$
$$
\bb{K}_{TE} = \int  \bar{\rho}^E(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\grad}\bar{\phi}^T(\bb{r},t)\,\md \bar{V} \approx \bb{Z}_E(t)\times\bb{F}_{TE}(t) \approx M_E\,\bb{Z}_E(t)\times\bb{g}_1^T
$$
$$
\bb{K}_{TA} = \int  \bar{\rho}^A(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\grad}\bar{\phi}^T(\bb{r},t)\,\md \bar{V} \approx \bb{Z}_A(t)\times\bb{F}_{TA}(t) \approx \big(M_A\,\bb{Z}_E(t)+\bb{C}_A(t)\big)\times\bb{g}_1^T
$$

where we use the fact that, as discussed in {numref}`Section %s <sec-total-torque>` - {ref}`sec-total-torque` and within the first order perturbation theory, the torque with respect to the Earth's centre of mass is negligible, eq. {eq}`G:73`. 

Also, let us consider the total force and torque of the athmosphere on the Earth

$$
\begin{align}
\bb{F}_{AE} &= \int_{\partial V_t} \bar{\rho}^E(\bb{r},t)\,\bb{T}^A(\bb{r},t)\,\md S - \int  \bar{\rho}^E(\bb{r},t)\,\bar{\grad}\bar{\phi}^A(\bb{r},t)\,\md \bar{V}
\\
\bb{K}_{AE} &= \int_{\partial V_t} \bar{\rho}^E(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bb{T}^A(\bb{r},t)\,\md S - \int  \bar{\rho}^E(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\grad}\bar{\phi}^A(\bb{r},t)\,\md \bar{V} 
\\
&\approx \bb{Z}_E(t)\times\bb{F}_{AE}(t) + \bb{k}_A(t)
\end{align}
$$

with $\bb{k}_A$ being the torque from the athmosphere with respect to the Earth's centre of mass

$$
\bb{k}_A(t) = \int_{\partial V_t} \bar{\rho}^E(\bb{r},t)\,\bb{r}\times\bb{T}^A(\bb{r},t)\,\md S - \int  \bar{\rho}^E(\bb{r},t)\,\bb{r}\times\bar{\grad}\bar{\phi}^A(\bb{r},t)\,\md \bar{V} 
$$



Summary:

\qquad
\bb{F}_{TA} = M_A\,\bb{g}_1^T
\qquad
\bb{F}_{AE} = OK
