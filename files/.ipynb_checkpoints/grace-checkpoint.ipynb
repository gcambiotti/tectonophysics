# International Earth Rotation and Reference Systems Service (IERS)

```python
dire = "RL06/isdcftp.gfz-potsdam.de/grace/Level-2/GFZ/RL06/"
import glob
gaa_files = glob.glob(dire+"GAA*")
gab_files = glob.glob(dire+"GAB*")
gac_files = glob.glob(dire+"GAC*")
gsm_files = glob.glob(dire+"GSM*")
gsm_files = [file for file in gsm_files if "BB01" in file]
gaa_files.sort()
gab_files.sort()
gac_files.sort()
gsm_files.sort()
```

```python
import gzip
import numpy as np

N = 180
P = (N+1)*(N+2)//2
gaa_shs = np.empty((len(gaa_files),N+1,N+1,2))
gab_shs = np.empty((len(gab_files),N+1,N+1,2))
gac_shs = np.empty((len(gac_files),N+1,N+1,2))
```

```python
skiprows = 125
for j,file in enumerate(gac_files):
    with gzip.open(file, 'rb') as f:
        data = np.loadtxt(f,skiprows=skiprows,usecols=[3,4])
        k0 = 0
        for m in range(N+1):
            dk = N+1-m
            k1 = k0 + dk 
            gac_shs[j,m:m+dk,m] = data[k0:k1]
            k0 = k1
for j,file in enumerate(gab_files):
    with gzip.open(file, 'rb') as f:
        data = np.loadtxt(f,skiprows=skiprows,usecols=[3,4])
        k0 = 0
        for m in range(N+1):
            dk = N+1-m
            k1 = k0 + dk 
            gab_shs[j,m:m+dk,m] = data[k0:k1]
            k0 = k1
for j,file in enumerate(gaa_files):
    with gzip.open(file, 'rb') as f:
        data = np.loadtxt(f,skiprows=skiprows,usecols=[3,4])
        k0 = 0
        for m in range(N+1):
            dk = N+1-m
            k1 = k0 + dk 
            gaa_shs[j,m:m+dk,m] = data[k0:k1]
            k0 = k1
```
