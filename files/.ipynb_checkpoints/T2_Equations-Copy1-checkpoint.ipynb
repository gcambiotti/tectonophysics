## Momentum equation

In the perspective of obtaining the momentum equation for the deformable Earth, it is convinient to introduce the total gravitational potential $\phi$ including the contribution from the Earth, $\phi$,  athmosphere, $\phi^A$, and the celestial bodies, $\phi^T$, 

$$
\Phi(\bb{x},t) = \phi(\bb{x},t) +  \phi^A(\bb{x},t) + \phi_0^T(\bb{x},t) + \phi_1^T(\bb{x},t) + \psi^T(\bb{x},t)
$$ (M:1)

where we use the decomposition given by eq. {eq}`G:56`, $\phi^T = \phi_0^T +\phi_1^T + \psi^T$ with $\psi^T$ being the tidal potential. In addition to it, we shall define  also the total (generalized) potential

$$
\Psi(\bb{x},t) = \phi(\bb{x},t)  + \phi^A(\bb{x},t)  + \psi^T(\bb{x},t) + \psi^C(\bb{x},t)
$$ (M:2)

where we have added the centrifugal potential and removed the degree-0 and -1 contributions of the gravitational potential of the celestial bodies. Its local decomposition reads

$$
\bar{\Psi}(\bb{r},t) = \phi_0(\bb{r}) +\loc\bar{\Psi}(\bb{r},t)
\qquad\mathrm{with}\qquad
\loc\bar{\Psi}(\bb{r},t) = \loc\bar{\phi}(\bb{r},t) + \bar{\phi}^A(\bb{r},t)  +\bar{\psi}^T(\bb{r},t) +\bar{\psi}^C(\bb{r},t) 
$$ (M:3)

where the gravitational potentials due to the athmosphere and the celestial bodies, as well as the centrifugal potential, have been considered as first-order perturbations.

Let us now obtain the momentum equation describing the balance between surface and volume forces acting on an infinitesimal element of the Earth system. Let us consider a time-dependent region $\mathcal{V}_t$ evolving according to the displacement field $\bb{u}$ and the time derivative of the linear momentum to the total surface and volume forces

$$
\begin{align}
\int_{\mathcal{V}_t} \bar{\rho}(\bb{r},t)\,\bar{\bb{A}}(\bb{r},t)\,\md V = \int_{\partial \mathcal{V}_t} \bar{\bb{n}}(\bb{r},t)\cdot\bar{\sb{\sigma}}(\bb{r},t)\,\md \bar{S} - \int_{\mathcal{V}_t}\bar{\rho}(\bb{r},t)\,\bar{\grad}\bar{\Phi}(\bb{r},t)\,\md \bar{V}  = \int_{ \mathcal{V}_t} \big(\bar{\grad}\bar{\sb{\sigma}}(\bb{r},t)-\bar{\rho}(\bb{r},t)\,\bar{\grad}\bar{\Phi}(\bb{r},t)\big)\,\md \bar{V} 
\end{align}
$$ (M:4)

where $\bar{\sb{\sigma}}$ is the Eulerian Cauchy stress tensor, $\partial \mathcal{V}_t$ is the surface of the deforming region $\mathcal{V}_t$ and $\bar{\bb{n}}$ is its outward normal unit vector. By the arbitraryness of the region $\mathcal{V}_t$, the above equation holds at the local level and it yields the so called momentum equation

$$
\begin{align}
\bar{\grad}\cdot\bar{\sb{\sigma}}(\bb{r},t) -\bar{\rho}(\bb{r},t)\,\bar{\grad}\bar{\Phi}(\bb{r},t) = \bar{\rho}(\bb{r},t)\,\bar{\bb{A}}(\bb{r},t)
\end{align}
$$ (M:5)

By making use of eqs. {eq}`O:54` and {eq}`E:29`, it can be recasted as follows

$$
\begin{align}
& \bar{\grad}\cdot\bar{\sb{\sigma}}(\bb{r},t) - \bar{\rho}(\bb{r},t)\,\bar{\grad} \bar{\Phi}(\bb{r},t)
= \bar{\rho}(\bb{r},t)\,\ddercla{ \bb{Z}_E(t)}  + \bar{\rho}(\bb{r},t)\,\bar{\grad}\bar{\psi}^C + \bar{\rho}(\bb{r},t)\,\left(\derpar{\sb{\omega}(t)}\times\bb{r} + 2\,\sb{\omega}(t)\times\bar{\bb{v}}(\bb{r},t)  +\bar{\bb{a}}(\bb{r},t)\right)
\end{align}
$$ (M:6)

that, in view of eq. {eq}`M:2`, can be rewritten in terms of the total (generalized) potential $\Psi$

$$
\begin{align}
& \bar{\grad}\cdot\bar{\sb{\sigma}}(\bb{r},t) - \bar{\rho}(\bb{r},t)\,\bar{\grad}\bar{\Psi}
= \bar{\rho}(\bb{r},t)\,\bar{\bb{a}}(\bb{r},t) + \bar{\rho}(\bb{r},t)\,\left(\derpar{\sb{\omega}(t)}\times\bb{r} + 2\,\sb{\omega}(t)\times\bar{\bb{v}}(\bb{r},t)\right) + \bar{\rho}(\bb{r},t)\,\left(\ddercla{\bb{Z}_E(t)}-\bb{g}_1^T(t)\right)
\end{align}
$$ (M:7)

In view of eq. {eq}`E:29`, the third term in the right-hand side yields zero. Furthermore,  the Euler and Coriolis accelerations can be neglected for studying solid Earth deformation and, so, eq. {eq}`M:7` simplifies into

$$
\begin{align}
& \bar{\grad}\cdot\bar{\sb{\sigma}}(\bb{r},t) - \bar{\rho}(\bb{r},t)\,\bar{\grad}\bar{\Psi}
= \bar{\rho}(\bb{r},t)\,\bar{\bb{a}}(\bb{r},t) 
\end{align}
$$ (M:8)

that is the momentum equation that we shall solve.


### Initial and incremental momentum equations

In addition to the decomposition of the Earth's density and gravitational potentials, we also consider a similar decomposition for the initial Cauchy stress tensor

$$
\bar{\sb{\sigma}}(\bb{r},t) = \sb{\sigma}_0(\bb{r}) + \loc\bar{\sb{\sigma}}(\bb{r},t) 
$$ (M:9)

with

$$
\loc\bar{\sb{\sigma}}(\bb{r},t) = \loc\sb{\sigma}(\bb{x},t) \approx  -\bb{u}(\bb{x},t)\cdot\grad\sb{\sigma}_0(\bb{x}) + \mat\sb{\sigma}(\bb{x},t)
$$ (M:10)

We also define it in such a way that, at the reference configuration, the following initial momentum equation holds

$$
\begin{align}
& \bar{\grad}\cdot\sb{\sigma}_0(\bb{r}) - \rho_0(\bb{r})\,\bar{\grad}\phi_0(\bb{r})
=\zero
\end{align}
$$ (M:11)

or, equivalently,

$$
\begin{align}
& \grad\cdot\sb{\sigma}_0(\bb{x}) - \rho_0(\bb{x})\,\grad\phi_0(\bb{x})
=\zero
\end{align}
$$ (M:12)

By subtracting eq. {eq}`M:11`  from eq. {eq}`M:8` and neglecting second-order terms, we thus obtain the incremental momentum equation

$$
\begin{align}
& \bar{\grad}\cdot\loc\bar{\sb{\sigma}}(\bb{r},t) -\rho_0(\bb{r},t)\,\bar{\grad}\loc\bar{\Psi}(\bb{r},t) - \loc\bar{\rho}(\bb{r},t)\,\bar{\grad}\phi_0(\bb{r},t)
= \rho_0(\bb{r})\,\bar{\bb{a}}(\bb{r},t) 
\end{align}
$$ (M:13)

Furthermore, by considering the Lagrangian representation of the local incremental fields and that the gradient or Laplacian of fields with respect to the current coordinates can be approximated, at the first order, with their gradient or Laplacian with respect to the initial coordinates, eqs. {eq}`O:31` and {eq}`O:32`, we can also write

$$
\begin{align}
& \grad\cdot\loc\sb{\sigma}(\bb{x},t) -\rho_0(\bb{x})\,\grad\loc\Psi(\bb{x},t) - \loc\rho(\bb{x},t)\,\grad\phi_0(\bb{x}) = \rho_0(\bb{x})\,\bb{a}(\bb{x},t) 
\end{align}
$$ (M:14)


## Differential system for self-gravitating Earth models

Let us now summarize the differential equations describing the initial and incremental state of the Earth and consider both the momentum and Poisson equation. For the initial state, from eqs. {eq}`M:12` and {eq}`E:13`, we have

$$
\begin{align}
& \grad\cdot\sb{\sigma}_0(\bb{x}) - \rho_0(\bb{x})\,\grad\phi_0(\bb{x})
=\zero
\\
& \nabla^2 \phi_0(\bb{x}) = C\,\rho_0(\bb{x})
\end{align}
$$ (M:15)

As it concern the incremental state, instead, from eqs. {eq}`M:14`

$$
\begin{align}
& \grad\cdot\loc\sb{\sigma}(\bb{x},t) -\rho_0(\bb{x})\,\grad\loc\Psi(\bb{x},t) - \loc\rho(\bb{x},t)\,\grad\phi_0(\bb{x}) = \rho_0(\bb{x})\,\bb{a}(\bb{x},t) 
\\
& \nabla^2\loc\Phi(\bb{x},t) = C\,\big(\loc\rho(\bb{x},t) + \rho^A(\bb{x},t)\big)
\end{align}
$$ (M:16)

Note that we have included the density distribution of the athmosphere in the incremental Poisson equation due to the fact that the (generalized) total potential accounts also for the gravitational potential of the athmosphere. In order to fix the ideas, the incremental momentum equation will be solved within the solid Earth, i.e. for all $\bb{x}$ such that $x=\big\|\bb{x}\big\|<a$, while the incremental Poisson equation will be solved up to just above the Earth surface, i.e., for all $\bb{x}$ such that $x=\big\|\bb{x}\big\|\leq a^+$.




### Spherical symmetry

Let us now adopt the spherical symmetry of the initial state and assume that the initial Earth is in hydrostatic equilibrium

$$
\rho_0(\bb{x}) = \rho_0(x) \qquad \phi_0(\bb{x}) = \phi_0(x)  \qquad \sb{\sigma}_0(\bb{x}) = - p_0(x)\,\one
$$ (M:18)

with $p_0$ being the initial hydrostatic pressure. Within these assumptions, eq. {eq}`M:15` simplifies as follows
$$
-\frac{\partial p_0(x)}{\partial x} - \rho_0(x)\,g_0(x) = 0 \qquad \frac{\partial g_0(x)}{\partial x}+\frac{2}{x}\,g_0(x)  = C\,\rho_0(x)
$$

with $g_0$ being the initial gravity acceleration

$$
g_0(x) = \frac{\partial \phi_0(x)}{\partial x} 
$$


Furthermore, the local incremental Cauchy stress given by eq. {eq}`M:10` becomes

$$
\loc\bar{\sb{\sigma}}(\bb{r},t) = \loc\sb{\sigma}(\bb{x},t) \approx  -\big(\bb{u}(\bb{x},t)\cdot\grad p_0(\bb{x})\big)\,\one + \mat\sb{\sigma}(\bb{x},t) = - \rho_0(x)\,g_0(x)\,u(\bb{x},t)\,\one + \mat\sb{\sigma}(\bb{x},t)
$$

with $u$ being the radial component of the displacement field

$$
u(\bb{x},t) = \bb{u}(\bb{x},t)\cdot\bn{r}(\vartheta,\varphi)
$$



By substituting the above expression into the incremental momentum equation, eq. {eq}`M:16`, we thus obtain

$$
\grad\cdot\mat\sb{\sigma}(\bb{x},t) - \grad\big(\rho_0(x)\,g_0(x)\,u(\bb{x},t)\big) -\rho_0(x)\,\grad\loc\Psi(\bb{x},t) - \loc\rho(\bb{x},t)\,\grad\phi_0(x)
= \rho_0(x)\,\bb{a}(\bb{x},t) 
$$

### Earthquakes, athmosphere and celestial bodies

Before considering internal surface loadings, we will consider the case in which the only forcings are caused by earthquakes, the athmosphere and celestial bodies. In this case, the total external force acting on the deformable Earth reads

$$
\bb{F}(t) = \int_{\partial\mathcal{V}}\bar{\bb{T}}^A(\bb{r},t)\,\md S - \int\bar{\rho}(\bb{r},t)\,\bar{\grad}\big(\bar{\phi}^A(\bb{r},t)+\bar{\phi}^T(\bb{r},t)\big)\,\md \bar{V} = M\,\bb{g}_1^T(t) + \bb{F}^A(t) + \bb{F}^L(t)
$$

where $\bb{T}^A$ is the traction on the Earth surface due to the athmosphere, we made use of eq. to decompose the celestial gravitational potential into the reference degree-1 contribution and the tidal potential, $M$ is the total mass

$$
M = \int\bar{\rho}(\bb{r},t)\,\md \bar{V} = \int\rho_0(\bb{x})\,\md V
$$

while $\bb{F}^A$ and $\bb{F}^T$ are the total forces due to the athmosphere and tidal forcings

$$
\begin{align}
& \bb{F}^A(t) = \int_{\partial\mathcal{V}}\bar{\bb{T}}^A(\bb{r},t)\,\md S - \int\bar{\rho}(\bb{r},t)\,\bar{\grad}\bar{\phi}^A(\bb{r},t)\,\md \bar{V}
\\
& \bb{F}^T(t) = - \int\bar{\rho}(\bb{r},t)\,\bar{\grad}\bar{\psi}^T(\bb{r},t)\,\md \bar{V}
\end{align}
$$


The total external torque acting on the deformable Earth reads

$$
\bb{K}(t) = \int_{\partial\mathcal{V}}\bar{\bb{R}}(\bb{r},t)\times\bb{T}^A(\bb{r},t)\,\md S - \int\bar{\rho}(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\grad}\big(\bar{\phi}^A(\bb{r},t)+\bar{\phi}^T(\bb{r},t)\big)\,\md \bar{V} = \bb{X}_O(t)\times\bb{F}  + \bb{C}(t)\times\bb{g}_1^T(t) + \bb{K}^A(t) + \bb{K}^L(t)
$$

with $\bb{K}^A$ and $\bb{K}^T$ being the relative torques with respect to the time-dependent origin due to the athmosphere and tidal forcings

$$
\begin{align}
&\bb{K}^A = \int_{\partial\mathcal{V}}\bb{r}\times\bar{\bb{T}}^A(\bb{r},t)\,\md \bar{S} - \int\bar{\rho}(\bb{r},t)\,\bb{r}\times\bar{\grad}\bar{\phi}^A(\bb{r},t)\,\md \bar{V} 
\\
& \bb{K}^T =  - \int\bar{\rho}(\bb{r},t)\,\bb{r}\times\bar{\grad}\bar{\psi}^T(\bb{r},t)\,\md \bar{V} = - \int\bar{\rho}(\bb{r},t)\,\bb{r}\times\bar{\grad}\bar{\psi}^T(\bb{r},t)\,\md \bar{V}
\end{align}
$$

and

$$
\bb{C}(t) = \int \bar{\rho}(\bb{r},t)\,\bb{r}\,\md\bar{V} = \int \rho_0(\bb{x})\,\bb{r}\,\md V
$$

Within the assumption of the first-order perturbation theory and that the initial region occupied by the Earth is a sphere of radius $a$, we note that 

$$
\begin{align}
& \bb{F}^A(t) \approx \int_{\partial\mathcal{V}}\bb{T}^A(\bb{x},t)\,\md S - \int \rho_0(\bb{x})\,\grad\phi^A(\bb{x},t)\,\md V
\\
& \bb{F}^T(t) \approx - \int \rho(\bb{x},t)\,\grad \psi^T(\bb{x},t)\,\md V
\end{align}
\\
&\bb{K}^A \approx \int_{\partial\mathcal{V}}\bb{x}\times\bb{T}^A(\bb{x},t)\,\md S - \int \rho(\bb{x},t)\,\bb{x}\times\grad\phi^A(\bb{r},t)\,\md V
\\
& \bb{K}^T \approx  - \int\rho_0(\bb{x},t)\,\bb{x}\times\grad\bar{\psi}^T(\bb{r},t)\,\md \bar{V} = - \int\bar{\rho}(\bb{r},t)\,\bb{r}\times\bar{\grad}\bar{\psi}^T(\bb{r},t)\,\md \bar{V}
\end{align}
$$
#### Spherical harmonic expansion

Let us consider the following spherical harmonic expansions

$$
\bar{\rho}(\bb{r},t) = \rho(\bb{x},t) = \sum_n\sum_m \bar{\rho}_{n,m}(r,t)\,Y_n^m(\vartheta,\varphi)
\qquad
\bar{\Phi}(\bb{r},t) = \sum_n\sum_m \bar{\Phi}_{n,m}(r,t)\,Y_n^m(\vartheta,\varphi)
$$

and

$$
\bar{\phi}(\bb{r},t) = \sum_n\sum_m \bar{\phi}_{n,m}(r,t)\,Y_n^m(\vartheta,\varphi)
\qquad
\bar{\phi}^A(\bb{r},t) = \sum_n\sum_m \bar{\phi}_{n,m}^A(r,t)\,Y_n^m(\vartheta,\varphi)
\qquad
\bar{\psi}^T(\bb{r},t) = \sum_{n\geq 2}\sum_m \bar{\psi}_{n,m}^T(r,t)\,Y_n^m(\vartheta,\varphi)
$$

with

$$
\bar{\Phi}_{n,m}(r,t) =\bar{\phi}_{n,m}(r,t) + \bar{\phi}_{n,m}^A(r,t) + \bar{\psi}_{n,m}^T(r,t)
$$

Furthermore, we also consider the spherical harmonic expansion of the athmospheric incremental traction

$$
\bar{\bb{T}}^A(\bb{r},t) = \sum_n\sum_m \big(\bar{R}_{n,m}^A(r,t)\,\bb{P}_n^m(\vartheta,\varphi)+\bar{S}_{n,m}^A(r,t)\,\bb{B}_n^m(\vartheta,\varphi)+\bar{T}_{n,m}^A(r,t)\,\bb{C}_n^m(\vartheta,\varphi)\big)
$$

By sustituting these expansions in eq. we obtain

$$
\begin{align}
& \bb{F}^A(t) = \int_{\partial\mathcal{V}}\bb{T}^A(\bb{r},t)\,\md S - \int\bar{\rho}(\bb{r},t)\,\bar{\grad}\bar{\phi}^A(\bb{r},t)\,\md \bar{V}
\\
& \bb{F}^T(t) = - \int\bar{\rho}(\bb{r},t)\,\bar{\grad}\bar{\psi}^T(\bb{r},t)\,\md \bar{V}
\end{align}
$$

### Centre of mass and linear and angular momenta

$$
\bb{X}_C(t) = \frac{1}{M}\,\int \bar{\rho}(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\,\md\bar{V} = \frac{1}{M}\,\int \rho_0(\bb{x})\,\bb{R}(\bb{x},t)\,\md V = \bb{X}_O(t) + \frac{\bb{C}(t)}{M}
$$

$$
\bb{P}(t) = \int \bar{\rho}(\bb{r},t)\,\bar{\bb{V}}(\bb{r},t)\,\md\bar{V}  = M\,\dercla{\bb{X}_C(t)} = M\,\dercla{\bb{X}_O(t)} + \dercla{\bb{C}(t)}

\bb{H} = \int \bar{\rho}(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\bb{V}}(\bb{r},t)\,\md\bar{V}  = \bb{X}_O(t)\times\bb{P}(t) + \bb{C}(t)\times\dercla{\bb{X}_O(t)} + \bb{I}(t)\cdot\sb{\omega}(t) + \bb{h}(t)

$$
\bb{h} = \int \bar{\rho}(\bb{r},t)\,\bb{r}\times\bar{\bb{v}}(\bb{r},t)\,\md\bar{V}

We can choose $\bb{X}_O$ and $\sb{\omega}$ such that 


$$
\bb{C}(t) = \int \bar{\rho}(\bb{r},t)\,\bb{r}\,\md\bar{V} = \zero

\dercla{\bb{P}} = M\,\ddercla{\bb{X}_C(t)}

and that

$$
\bb{F} = M\,\bb{g}_1^T(t) + \bb{F}^A  + \bb{F}^T

We note that we can

\qquad
\loc\bb{I}(t) = \int \loc\bar{\rho}_0(\bb{r},t)\,\big(r^2\,\one - \bb{r}\,\bb{r}\big)\,\md \bar{V}
$$


### Conservation of the linear and angular momenta

$$
\dercla{\bb{P}} = \int \bar{\rho}(\bb{r},t)\,\bar{\bb{A}}(\bb{r},t)\,\md\bar{V}  = M\,\ddercla{\bb{X}_C(t)}

$$

\end{align}$$

\bar{\bb{c}}(\bb{r},t) = \bar{\rho}(\bb{r},t)\,\frac{\bb{F}^A(t)+\bb{F}^T(t)}{M} \approx \rho_0(\bb{r})\,\frac{\bb{F}^A(t)+\bb{F}^T(t)}{M}
$$

Within this framework, we thus obtain

\end{align}$$

$$

\dif{\bb{F}}(t) = (M_E+M_A)\,\bb{g}_1(t) - \int \bar{\rho}(\bb{r},t)\,\frac{\psi^T_{n,m}(a)}{a}\,\left(\frac{r}{a}\right)^{n-1}\,\big(n\,\bb{P}_{n,m}(\vartheta,\varphi)+\gamma_n\,\bb{B}_{n,m}(\vartheta,\varphi)\big)

## Earth and load system

### Load system

In the present theory, we will not enter into the details of the deformation of the load system and we will assume that its Eulerian density distribution $\bar{\rho}^L$ is simply prescribed according to the physical problem that we intend to discuss. In this respect, we do not model the displacement field of the deformable load but we assume that its particle can move through time and, so, we associate a load relative velocity field $\bb{v}^L$

$$
\bar{\bb{V}}^L(\bb{r},t) = \dercla{\bb{X}_O(t)} + \sb{\omega}(t)\times\bb{r} + \bar{\bb{v}}^L(\bb{r},t)
$$

Also, their particle can accelerate and, within the Eulerian framework and similar to eq. , we can write

$$
\bar{\bb{A}}^L(\bb{r},t)
= \ddercla{ \bb{X}_O(t)} + \derpar{\sb{\omega}(t)}\times\bb{r} + \sb{\omega}(t)\times\sb{\omega}(t)\times\bb{r} +2\,\sb{\omega}(t)\times\bar{\bb{v}}^L(\bb{r},t)  +\bar{\bb{a}}^L(\bb{r},t)
$$

with $\bar{\bb{a}}^L$ being the load relative acceleration with respect to the rotating unit vectors 

$$
\bar{\bb{a}}^L(\bb{r},t) = \derpar{\bar{\bb{v}}^L(\bb{r},t)} +\bar{\bb{v}}^L(\bb{r},t)\cdot\bar{\grad}\bar{\bb{v}}^L(\bb{r},t)
$$

Different from the deformable Earth, for the case of loads, we adopt an Eulerian approach for describing its density, velocity and acceleration fields. On the other hand, we can express these fields with respect to the initial coordinates $\bb{x}$ of the particle composing the deformable Earth which occupy the current position $\bb{r}$ at present. In this sense, we can also introduce the Lagrangian representations of the above fields

$$
\rho^L(\bb{x},t) = \bar{\rho}^L(\bb{r},t)
\qquad
\bb{V}^L(\bb{x},t)
 = \bar{\bb{V}}^L(\bb{r},t)
\qquad
\bb{v}^L(\bb{x},t)
 = \bar{\bb{v}}^L(\bb{r},t)
\qquad
\bb{A}^L(\bb{x},t)
=\bar{\bb{A}}^L(\bb{r},t)
\qquad
\bb{a}^L(\bb{x},t)
 = \bar{\bb{a}}^L(\bb{r},t)
$$

In addition to it, we also define the following density distribution

$$
\rho_0^L(\bb{x},t) =
\rho^L(\bb{x},t)\,J(\bb{x},t) 
$$ 

which can be seen as the Lagrangian load density once corrected by the volume change of the deformable Earth. Different from $\rho_0$ given by eq., $\rho_0^L$ can depend on time because we $J$ does not describe the volume change of the load system, but that of the Earth one.

### First-order perturbation theory

Assuming that the displacement field, $\bb{u}$, and its gradient, $\grad\bb{u}$, are (infinitesimal) first-order terms, we note that the Jacobian $J$ can be approximated as follows

$$
\begin{align} &
J(\bb{x},t) \approx 1+ \grad\cdot\bb{u}(\bb{x},t)
\end{align}
$$

and eq. becomes

$$
\begin{align} &
\rho(\bb{x},t) = \rho_0(\bb{x}) + \mat\rho(\bb{x},t) \qquad \mat\rho(\bb{x},t) \approx -\rho_0(\bb{x})\,\grad\cdot\bb{x}(\bb{x},t)
\end{align}
$$

Here $\mat\rho$ represents the so called material incremental Earth density which describes the perturbation of the current density of a particle with respect to its initial density. Within the Eulerian framework, it is natural to introduce the following decomposition

$$
\begin{align} &
\bar{\rho}(\bb{r},t) = \rho_0(\bb{r})+\loc\bar{\rho}(\bb{r},t)
\end{align}
$$

with $\loc\bar{\rho}$ being the local incremental Earth density. From the comparison between eqs. we obtain

$$
\begin{align} &
\loc\bar{\rho}(\bb{r},t) = \loc\rho(\bb{x},t) =  \rho_0(\bb{x}) - \rho_0(\bb{x}+\bb{u}(\bb{x},t)) + \mat\rho(\bb{x},t) 
\end{align}
$$

that, within the first-order approximation, can be approximated as follows

$$
\begin{align} &
\loc\bar{\rho}(\bb{r},t) = \loc\rho(\bb{x},t) \approx  -\bb{u}(\bb{x},t)\cdot\grad\rho_0(\bb{x}) + \mat\rho(\bb{x},t)  \approx  -\grad\cdot\big(\rho_0(\bb{x})\,\bb{u}(\bb{x},t)\big)
\end{align}
$$

In the following we will assume that the load density is an (infinitesimal) first-order term and, so, we will neglet its product with other infinitesimal terms. In this respect, eq. becomes

$$
\rho_0^L(\bb{x},t) \approx
\rho^L(\bb{x},t)
$$ 

Moreover, we will also assume that the relative velocities of the deformable Earth and of the loads are first-order terms.

### Total mass and centre of mass

Let us first calculate the total mass of the Earth and load system

$$
M(t) = \int\big(\bar{\rho}(\bb{r},t)+\bar{\rho}^L(\bb{r},t)\big)\,\md \bar{V} = M + M^L
$$

with $M$ and $M^L$ being the Earth and load masses

$$
M = \int\bar{\rho}(\bb{r},t)\,\md \bar{V} = \int \rho_0(\bb{x})\,\md V \qquad M^L = \int\bar{\rho}^L(\bb{r},t)\,\md \bar{V}  = \int \rho_0^L(\bb{x},t)\,\md V
$$

We note that, in view of mass conservation, we have required that both masses does not depend on time. This requirement is consistent within the present description of the deformable Earth. For the load mass, instead, it implies that we are neglectinbg the possibility of mass exchanges between the load and the athmosphere systems.

Let us now introduce the following vector

$$
\bb{C}(t) = \int\big(\bar{\rho}(\bb{r},t)+\bar{\rho}^L(\bb{r},t)\big)\,\bar{\bb{R}}(\bb{r},t)\,\md \bar{V} = M\,\bb{X}_O(t) + \bb{c}(t)
$$

with

$$
\begin{align}
& \bb{c}(t) = \int\big(\bar{\rho}(\bb{r},t)+\bar{\rho}^L(\bb{r},t)\big)\,\bb{r}\,\md \bar{V} = \int\big(\rho_0(\bb{x})+\rho_0^L(\bb{x},t)\big)\,\bb{r}\,\md V = \bb{c}(t) + \bb{c}^L(t)
\\
& \bb{c}(t) = \int\bar{\rho}(\bb{r},t)\,\bb{r}\,\md \bar{V} = \int \rho_0(\bb{x})\,\bb{r}\,\md V
\\
& \bb{c}^L(t) = \int\bar{\rho}^L(\bb{r},t)\,\bb{r}\,\md \bar{V}  = \int \rho_0^L(\bb{x},t)\,\bb{r}\,\md V
\end{align}
$$


In light of eqs, the centre of mass $\bb{X}_C$  of the Earth and load system thus reads

$$
\bb{X}_C(t) = \frac{1}{M}\,\int\big(\bar{\rho}(\bb{r},t)+\bar{\rho}^L(\bb{r},t)\big)\,\bar{\bb{R}}(\bb{r},t)\,\md \bar{V}  = \frac{\bb{C}(t)}{M} = \bb{X}_O(t) + \frac{\bb{c}(t)}{M}
$$

and it thus composes of the time-dependent origin $\bb{X}_O$ and its relative position with respect to it, $\bb{c}/M$. This result suggests us to choose the time-dependent origin $\bb{X}_O$  as the centre of mass $\bb{X}_O$, which means

$$
\bb{X}_C(t) = \bb{X}_O(t) \qquad \bb{c}(t) = \zero
$$

At the initial state, before the perturbation, we thus have to require that

$$
\bb{c}(0) = \int \rho_0(\bb{r})\,\bb{r}\,\md \bar{V} = \int \rho_0(\bb{x})\,\bb{x}\,\md V = \zero
$$

while, at later times, we have that

$$
\begin{align}
& \bb{c}(t) = \int\loc\bar{\rho}(\bb{r},t)\,\bb{r}\,\md \bar{V} = \int \rho_0(\bb{x})\,\bb{u}(\bb{x},t)\,\md V 
\end{align}
$$

In particular, within the first order approximation theory, we have

$$
\bb{c}(t) \approx \int\big(\loc\rho(\bb{x},t)+\rho^L(\bb{x},t)\big)\,\bb{x}\,\md V  = \zero
$$

Summarizing the above results, thanks to the choice of the centre of mass as the time-dependent origin, we have

$$
\begin{align}
& \bb{X}_C(t) = \bb{X}_O(t) \qquad 
 \bb{c}(t) = \bb{c}(t) + \bb{c}^L(t) = \zero
\\
& \bb{c}(t) = \int\loc\bar{\rho}(\bb{r},t)\,\bb{r}\,\md \bar{V} \approx \int \loc\rho(\bb{x},t)\,\bb{x}\,\md V 
\\
& \bb{c}^L(t) = \int\bar{\rho}^L(\bb{r},t)\,\bb{r}\,\md \bar{V}  \approx \int \rho^L(\bb{x},t)\,\bb{x}\,\md V
\end{align}
$$

The total angular momentum of the Earth and load system reads

$$
\begin{align}

\end{align}
$$

### The choice of the reference system

The centre of mass and the angular momentum describe global feature of the Earth and load system can be used as starting point for defining the translating and rotating reference system. First, we require that the centre of mass $\bb{X}_C$ coincides with the time-dependent origin $\bb{X}_O$

$$
\bb{X}_C(t) = \bb{X}_O(t) 
$$

This means that

$$
\bb{X}_C(t) = \bb{X}_O(t) \qquad \bb{c}(t)=\zero \qquad \bb{c}(t) = - \bb{c}^L(t)
$$

and the linear and angular momenta become

$$
\bb{P}(t) = M(t)\,\dercla{\bb{X}_O(t)} + \bb{p}(t) 
\qquad
\bb{H}(t) = \bb{X}_O(t)\times\bb{P}(t)  + \bb{I}(t)\cdot\sb{\omega}(t) + \bb{h}(t)
$$

Let us now prove that the angular velocity can be choosen so that the angular momentum with respect to the rotating vectors is zero, $\bb{h}=\zero$. Let us first decompose the angular velocity of the rotating unit vectors such that 

$$
\sb{\omega}(t) = \sb{\omega}^*(t)+\mat\sb{\omega}(t) \qquad \mathrm{with} \qquad \mat\sb{\omega}(t) = -\bb{I}^{-1}(t)\cdot\bb{h}(t)
$$

where $\sb{\omega}^*$ is the angular velocity vector of the so called co-rotating reference frame. By considering that

$$
\sb{\omega}(t)\times\bb{r} + \bb{v} = \sb{\omega}^*\times\bb{r} + \bb{v}^*
$$

we obtain the following relation between the velocity in the rotating and co-rotating frames

$$
\bb{v} = \bb{v}^* -\mat\sb{\omega}\times\bb{r}
\qquad
\bb{v}^L = \bb{v}^{L,*} -\mat\sb{\omega}\times\bb{r}
$$

Then, by making use of this decomposition for computing the linear momentum in the rotating frame

$$
\bb{P}(t) = M(t)\,\dercla{\bb{X}_C(t)} + \bb{p}(t) = M(t)\,\dercla{\bb{X}_C(t)} + \bb{p}^*(t) 
$$

where we have use the following identity

$$
\bb{p}(t) = \int \big(\bar{\rho}(\bb{r},t)\,\bar{\bb{v}}(\bb{r},t)+\bar{\rho}^L(\bb{r},t)\,\bar{\bb{v}}^L(\bb{r},t)\big)\,\md \bar{V} = \bb{p}^*(t) - \sb{\omega}(t)\times\bb{c}(t) = \bb{p}^*(t) 
$$

with $\bb{p}$^* being the linear momentum in the co-rotating frame

$$
\bb{p}^*(t) = \int \big(\bar{\rho}(\bb{r},t)\,\bar{\bb{v}}^*(\bb{r},t)+\bar{\rho}^L(\bb{r},t)\,\bar{\bb{v}}^{L*}(\bb{r},t)\big)\,\md \bar{V}
$$

Thanks to the above results, the total angular momentum becomes

$$
\bb{H}(t) = \bb{X}_O(t)\times\bb{P}(t)  + \bb{I}(t)\cdot\sb{\omega}^*(t) 
$$

and, so, from the comparison with eq. , this means that, properly choosing the angular velocity vector $\sb{\omega}=\sb{\omega}^*$, the angular momentum with respect to the co-rotating frame is zero.


### Total mass

with

### Centre of mass

\begin{align}

Then, by making use of the initial and local incremental decomposition of the Earth density distribution, we also obtain

### Lagrangian and Eulerian representation

Let us now consider the Eulerian, $\bar{f}$ and $\bar{\bb{f}}$, and Lagrangian, $f$ and $\bb{f}$, representations of scalar and vector fields

& \bar{\bb{f}}(\bb{r},t) = \bb{f}(x_1,x_2,x_3,t) \qquad\mathrm{with}\qquad \bar{f}_i(\bb{r},t) = f_i(x_1,x_2,x_3,t)
\end{align}

with $r_i$ given by eq. {eq}`O:7` and the components of the vector field are with respect to the rotating unit vectors

\bar{\bb{f}}(\bb{r},t) = \bar{f}_i(\bb{r},t)\,\bb{e}_i(t)

The Lagrangian representation provides the value of the field of a particle of which the initial position was $\bb{X}_0$ (i.e., identified by the initial coordinates $x_i$). ### Gradients

Let us consider the gradient of a field within the Eulerian framework, $\bar{\grad}\bar{f}$, and obtain its Lagrangian representation

$$
\bar{\grad}\bar{f}(\bb{r},t) = \frac{\partial \bar{f}(r_1,r_2,r_3,t)}{\partial r_j}\,\bb{e}_j(t) = \frac{\partial x_k}{\partial r_j}\,\frac{\partial f(x_1,x_2,x_3,t)}{\partial x_k}\,\bb{e}_j(t) = \big(\one+\grad\bb{u}(x_1,x_2,x_3,t)\big)^{-1} \cdot\grad f(x_1,x_2,x_3,t) \approx \big(\one - \grad\bb{u}(x_1,x_2,x_3,t)\big)\cdot\grad f(x_1,x_2,x_3,t)
$$

In this respect, $\bar{\grad}\bar{f}$ and $\grad f$ differ from each other due to the first order term  $-\grad\bb{u}\cdot\grad f$. If the field $f$ is already a first order terms, in this case the gradients would differ by a second-order term which can be neglected.

## First order perturbation theory

Let us assume that the displacement field $\bb{u}$ is infinitesimal, and consider the Eulerian, $\bar{f}$, and Lagrangian, $f$, representation of a (scalar, vector or tensor) field

$$
\bar{f}(\bb{r},t) = f(\bb{x},t) 
$$

and consider its decomposition into the initial field $f_0$ and the local, $\loc\bar{f}$, and material, $\mat f$, incremental fields

$$
\bar{f}(\bar{\bb{r}},t) = f_0(\bar{\bb{r}}) + \loc\bar{f}(\bar{\bb{r}},t) 
\qquad
f(\bb{x},t) = f_0(\bb{x}) + \mat f(\bb{x},t) 
$$

We note that, also for the incremental field, we can adopt both the Eulerian and Lagrangian formulations

$$
\loc\bar{f}(\bb{r},t) = \loc f(\bb{x},t) \qquad \mat f(\bb{x},t) = \mat\bar{f}(\bb{r},t)
$$

Within a first-order perturbation theory, we can write

$$
f_0(\bb{r})= f_0(\bb{x}+\bb{u}(\bb{x},t))  \approx f_0(\bb{x}) + \bb{u}(\bb{x},t)\cdot\grad f_0(\bb{x})
$$

and, so, we obtain the following relation between local and material incremental fields in the Lagrangian formulation

$$
\loc \bar{f}(\bb{r},t) = \loc f(\bb{x},t) \approx \mat f(\bb{x},t) - \bb{u}(\bb{x},t)\cdot\grad f_0(\bb{x})
$$


## Deformable body and density distribution

Let us consider the scalar field describing the density distribution of a deformable body and, for the sake of the example, let us focus on the Earth density distrubtion. According to the mass conservation, we can write

$$
\bar{\rho}(\bar{\bb{r}},t) = \rho(\bb{x},t) = \frac{\rho_0(\bb{x})}{J(\bb{x},t)}
$$

where $J$ being the Jacobian of the transformation $\bb{r}\rightarrow \bb{x}$ given by

$$
J(\bb{x},t) = \mathrm{det}\big(\one+\grad\bb{u}(\bb{x},t)\big) \approx 1 + \grad\cdot\bb{u}(\bb{x},t)
$$

with the second identity holding within the first-order approximation. In this respect, eq. becomes

$$
 \rho(\bb{x},t) \approx \rho_0(\bb{x}) + \mat\rho(\bb{x},t)
$$

with material incremental density given by

$$
\mat\rho(\bb{x},t)  \approx -\rho_0(\bb{x})\,\grad\cdot\bb{u}(\bb{x},t)
$$

while the local incremental density reads

$$
\loc \bar{\rho}(\bb{r},t) = \loc \rho(\bb{x},t) \approx \mat \rho(\bb{x},t) - \bb{u}(\bb{x},t)\cdot\grad \rho_0(\bb{x}) \approx -\grad\cdot\big(\rho_0(\bb{x})\,\bb{u}(\bb{x},t)\big)
$$

### Mass and centre of mass

The total mass of the deformable body reads

$$
M = \int \bar{\rho}(\bar{\bb{r}},t)\,\md \bar{V} = \int \rho(\bb{x},t)\,J(\bb{x},t)\,\md V =  \int \rho_0(\bb{x})\,\md V 
$$

Making use of the decomposition into the intial and local incremental Earth densities

$$
M = \int \rho_0(\bb{r})\,\md \bar{V} + \int \loc\bar{\rho}(\bar{\bb{r}},t)\,\md \bar{V} = M_0 + \mat M(t)  = M_0
$$

with

$$
\begin{align}
& M_0 = \int \rho_0(\bb{r})\,\md \bar{V}  = \int \rho_0(\bb{x})\,\md V
\\
& \mat M(t) =  \int \loc\bar{\rho}(\bar{\bb{r}},t)\,\md \bar{V} \approx \int \loc\rho(\bb{x},t)\,\md V = 0
\end{align}
$$

Note that we set to zero $\mat M$ because the total mass of the Earth must coincide with the initial one, $M=M_0$.

The centre of mass, instead, gives

$$
\bb{X}_C = \frac{\bb{M}(t)}{M}
$$

with 

$$
\begin{align}
& \bb{M}(t) = \int \bar{\rho}(\bar{\bb{r}},t)\,\bar{\bb{R}}(\bar{\bb{r}},t)\,\md \bar{V} =  \int \rho_0(\bb{x})\,\bb{R}(\bb{x},t)\,\md V = M\,\bb{X}_O(t) + \loc\bb{M}(t)
\\
& \loc\bb{M}(t) = \int\bar{\rho}(\bar{\bb{r}},t)\,\bar{\bb{r}}\,\md \bar{V} = \int \rho_0(\bb{x})\,\bb{r}\,\md V   
\end{align}
$$

In this respect, eq. becomes

$$
\bb{X}_C(t) = \bb{X}_O(t) + \frac{\loc\bb{M}(t)}{M}
$$

Making use of the decomposition into the intial and local incremental Earth densities

$$
\loc\bb{M}(t) = \int\rho_0(\bar{\bb{r}})\,\bar{\bb{r}}\,\md \bar{V} + \int\loc\bar{\rho}(\bar{\bb{r}},t)\,\bar{\bb{r}}\,\md \bar{V} = \loc\bb{M}_0 + \mat\bb{M}(t)
$$

with

$$
\loc\bb{M}_0 = \int\rho_0(\bar{\bb{r}})\,\bar{\bb{r}}\,\md \bar{V}  = \int\rho_0(\bb{x})\,\bb{x}\,\md V 
$$
$$
\mat\bb{M}(t) = \int\loc\bar{\rho}(\bar{\bb{r}},t)\,\bar{\bb{r}}\,\md \bar{V} \approx \int \loc\rho(\bb{x},t)\,\bb{x}\,\md V 
$$

### Linear momentum

The linear momentum of the deformable body reads

$$
\bb{P}(t) = \int \bar{\rho}(\bar{\bb{r}},t)\,\bar{\bb{V}}(\bar{\bb{r}},t)\,\md \bar{V} =  \int \rho_0(\bb{x})\,\bb{V}(\bb{x},t)\,\md V = M\,\frac{\md\bb{X}_O(t)}{\md t} + \sb{\omega}(t)\times\loc\bb{M}(t)+\loc\bb{P}(t)
$$

with $\loc\bb{P}$ being the relative angular momentum

$$
\loc\bb{P}(t) = \int\bar{\rho}(\bar{\bb{r}},t)\,\bar{\bb{v}}(\bar{\bb{r}},t)\,\md \bar{V} = \int \rho_0(\bb{x})\,\bb{v}(\bb{x},t)\,\md V
$$ 

We note that, in view of $\bb{v}=\dot{\bb{u}}$, we can write

$$
\loc\bb{P}(t) = \loc\dot{\bb{M}}(t) \qquad \frac{\md \loc\bb{M}(t)}{\md t} = \loc\bb{P}(t) + \sb{\omega}(t)\times\loc\bb{M}(t)
$$



### Density distribution of the Earth and load system

By definition the density distribution of the Earth and load system reads

$$
\bar{\rho}(\bar{\bb{r}},t) = \bar{\rho}(\bar{\bb{r}},t) + \bar{\rho}^L(\bar{\bb{r}},t) + \bar{\rho}^S(\bar{\bb{r}},t) = 
 \rho(\bb{x},t) + \rho^L(\bb{x},t) + \rho^S(\bb{x},t) = \rho(\bb{x},t) 
$$

where the internal and surface load density distribution are considered first-order terms and they yields zero at the initial time. In this respect, the decompisition into the initial and incremental densities read

$$
\rho(\bb{x},t)  = \rho_0(\bb{x}) + \loc\rho(\bb{x},t)
\qquad
\loc\rho(\bb{x},t) = \mat\rho(\bb{x},t) + \rho^L(\bb{x},t) + \rho^S(\bb{x},t) 
$$

and

$$
\bar{\rho}(\bar{\bb{r}},t) = \rho_0(\bar{\bb{r}}) + \loc\bar{\rho}(\bar{\bb{r}},t)
\qquad
\loc\bar{\rho}(\bar{\bb{r}},t) = \loc\bar{\rho}(\bar{\bb{r}},t) + \bar{\rho}^L(\bar{\bb{r}},t) + \bar{\rho}^S(\bar{\bb{r}},t) 
$$

and, within the first-order perturbation theory, we can write

$$
\loc\bar{\rho}(\bb{r},t) = \loc\rho(\bb{x},t) = \loc\rho(\bb{x},t) + \rho^L(\bb{x},t) + \rho^S(\bb{x},t)
$$

We note that, we do not need to account for volume variations in defining the time evolution of the internal and surface loads just because they are infinitesimal and can be neglected for terms which are already of the first-order. In this respect, $\rho^L(\bb{x},t)$ and $\rho^S(\bb{x},t)$ can be seen as prescribed fields.

### Gravitational potentials

The natural framework for dealing with gravitational potentials is the Eulerian one and the Poisson equation becomes

$$
\bar{\grad}^2\bar{\phi}(\bar{\bb{r}},t) = c\,\bar{\rho}(\bar{\bb{r}},t)
$$

At the initial state, it gives

$$
\bar{\grad}^2\phi_0(\bar{\bb{r}}) = c\,\rho_0(\bar{\bb{r}})
\qquad\mathrm{or, equivalently,}\qquad
\grad^2\phi_0(\bb{x}) = c\,\rho_0(\bb{x})
$$

Then, by subtracting eq. to eq. we obtain the incremental Poisson equation

$$
\bar{\grad}^2\loc\bar{\phi}(\bar{\bb{r}},t) = c\,\loc\bar{\rho}(\bar{\bb{r}},t)
$$

and considering the result of SECTION, we can approximate correct to the first-order as follows

$$
\grad^2\loc\phi(\bb{x},t) = c\,\loc\rho(\bb{x},t)
\qquad
\grad^2\loc\psi(\bb{x},t) = c\,\loc\rho(\bb{x},t)
\qquad\grad^2\psi^L(\bb{x},t) = c\,\rho^L(\bb{x},t)  
\qquad\grad^2\psi^S(\bb{x},t) = c\,\rho^S(\bb{x},t)  
$$

### Total mass 

$$
M(t) = \int \bar{\rho}(\bar{\bb{r}},t)\,\md \bar{V} =  M_0 + \loc M(t)
$$

with 

$$

\qquad
\loc M(t) = \int \loc\bar{\rho}(\bar{\bb{r}},t)\,\md \bar{V} \approx \int \loc\rho(\bb{x},t)\,\md V  = 4\,\pi\,\int \loc\rho_{0,0}(x,t)\,x^2\,\md x

\bb{X}_C(t) = \frac{1}{M(t)}\int_{\mathbb{R}^3}\bar{\rho}\,\bar{\bb{R}}\,\md \bar{V} = \bb{X}_O(t) + \frac{\loc\bb{M}(t)}{M(t)}

\end{align}

### Angular momentum

Let us now consider the angular momentum of the Earth and load system

$$
\bb{H}(t) = \int_{\mathcal{V}}\rho\,\bb{R}\times\bb{V}\,\md V = \bb{X}_O(t)\times \int_{\mathcal{V}}\rho\,\bb{V}\,\md V + \int_{\mathcal{V}}\rho\,\bb{r}\times\bb{V}\,\md V = M(t)\,\bb{X}_O(t)\times\frac{\md \bb{X}_O(t)}{\md t} + \loc\bb{P}\times\frac{\md \bb{X}_O(t)}{\md t} + \bb{I}(t)\cdot\sb{\omega}(t)+ \loc\bb{H}(t)
$$

with $\bb{I}$ being the inertia tensor

$$
\bb{I}(t)\cdot\sb{\omega}(t) = \int_\mathcal{V}\rho\,\bb{r}\times(\sb{\omega}(t)\times\bb{r})\,\md V 
\qquad \bb{I}(t) = \int_\mathcal{V}\rho\,\big(r^2\,\one-\bb{r}\,\bb{r}\big)\,\md V 
$$

and $\loc\bb{H}$ being the relative angular momentum

$$
\loc\bb{H}(t) = \int_{\mathcal{V}}\rho\,\bb{r}\times\bb{v}\,\md V  
$$

In view of the choice of the time-dependent origin as the centre of mass, eq. simplifies into

$$
\bb{H}(t) = M(t)\,\bb{X}_O(t)\times\frac{\md \bb{X}_O(t)}{\md t}+ \bb{I}(t)\cdot\sb{\omega}(t)  + \loc\bb{H}(t)
$$

Furthermore, in the perspective of choising the angular velocity vector $\sb{\omega}$ so that the relative angular momentum is zero

$$
\loc\bb{H}=\zero
$$

eq. further simplifies into

$$
\bb{H}(t) = M(t)\,\bb{X}_O(t)\times\frac{\md \bb{X}_O(t)}{\md t}+ \bb{I}(t)\cdot\sb{\omega}(t) 
$$

## Spherical reference frame

In the following we will introduce the spherical reference frame with respect to the rotating unit vectors and write the relative position with respect to the time dependent origin $\bb{X}_O$ as follows

$$
\bb{x}(x,\vartheta,\varphi,t) = x\,\big[\sin\vartheta\,\big(\cos\varphi\,\bn{e}_1(t)+\sin\varphi\,\bn{e}_2(t)\big) + \cos\vartheta\,\bn{e}_3(t)\big]
$$ (O:38)

where $x$ is the radial distance relative to the time-dependent origin, while $\vartheta$ and $\varphi$ are the colatitude and longitude with respect to the rotating unit vectors. Within this framework, the spharical unit vectors depend also on time

$$
\begin{align}
& \bn{r}(\vartheta,\varphi,t) = \sin\vartheta\,\big(\cos\varphi\,\bn{e}_1(t)+\sin\varphi\,\bn{e}_2(t)\big) + \cos\vartheta\,\bn{e}_3(t)
\\
& \sn{\vartheta}(\vartheta,\varphi,t) = \cos\vartheta\,\big(\cos\varphi\,\bn{e}_1(t)+\sin\varphi\,\bn{e}_2(t)\big) - \sin\vartheta\,\bn{e}_3(t)
\\
& \sn{\varphi}(\vartheta,\varphi,t) = -\sin\varphi\,\bn{e}_1(t)+\cos\varphi\,\bn{e}_2(t)
\end{align}
$$ (O:39)

and eq. {eq}`O:38` becomes

$$
\bb{x}(x,\vartheta,\varphi,t) = x\,\bn{r}(\vartheta,\varphi,t) 
$$ (O:40)

Similar, for the relative current position of the particle, we will write

$$
\bar{\bb{r}}(r_1,r_2,r_3,t) = \bar{r}\,\big[\sin\bar{\vartheta}\,\big(\cos\bar{\varphi}\,\bn{e}_1(t)+\sin\bar{\varphi}\,\bn{e}_2(t)\big) + \cos\bar{\vartheta}\,\bn{e}_3(t)\big] = \bar{r}\,\bn{r}(\bar{\vartheta},\bar{\varphi},t)
$$ (O:41)




### Spherical harmonic expansions

Within the framwork of the (rotating) spherical reference frame, the spherical harmonic expansion of scalar fields reads

$$
f(\bb{x},t) = \sum_n\sum_m f_{n,m}(x,t)\,Y_n^m(\vartheta,\varphi) $$


As it concerns vector fields, instead, we have also to note that the spherical harmonic vectors depend on time and we will write 

$$
\bb{f}(\bb{x},t) = \sum_n\sum_m \big(f^P_{n,m}(x,t)\,\bb{P}_n^m(\vartheta,\varphi,t)+f^B_{n,m}(x,t)\,\bb{B}_n^m(\vartheta,\varphi,t)+f^C_{n,m}(x,t)\,\bb{C}_n^m(\vartheta,\varphi,t)\big)
$$

where, according to eqs {eq}`EQ` and {eq}`O:39`, the time-dependent spherical harmonic vectors are

$$
\begin{align}
& \bb{P}_n^m(\vartheta,\varphi,t) = Y_n^m(\vartheta,\varphi)\,\bn{r}(\vartheta,\varphi,t)
\\
& \bb{B}_n^m(\vartheta,\varphi,t) = \frac{r}{\gamma_n}\,\grad Y_n^m(\vartheta,\varphi)
\\
& \bb{C}_n^m(\vartheta,\varphi,t) = 
\end{align}
$$

The integral of the time-dependent spherical harmonic vectors over the unit sphere yields

$$
\int_{\Omega}\mathbf{P}_n^m(\vartheta,\varphi,t)\,\mathrm{d}\Omega = \frac{4\,\pi}{\sqrt{3}}\,\mathbf{p}_m(t)\,\delta_{n,1}
\qquad
\int_{\Omega}\mathbf{B}_n^m(\vartheta,\varphi,t)\,\mathrm{d}\Omega = \frac{4\,\pi\,\sqrt{2}}{\sqrt{3}}\,\mathbf{p}_m(t)\,\delta_{n,1}
\qquad
\int_{\Omega}\mathbf{C}_n^m(\vartheta,\varphi,t)\,\mathrm{d}\Omega =\mathbf{0}
$$ (P:54)

where $\mathbf{p}_m$ being the folliwing unit vectors 

$$
\mathbf{p}_0(t) = \mathbf{e}_3(t)
\qquad
\mathbf{p}_{-1}(t) = \mathbf{e}_1(t)
\qquad
\mathbf{p}_1(t) = \mathbf{e}_2(t)
$$ (P:55)


The spherical harmonic expansion of scalar and vector fields in the Eulerian representation read

$$
\bar{f}(\bar{\bb{r}},t) = \bar{f}_{n,m}(\bar{r},t)\,Y_n^m(\bar{\vartheta},\bar{\varphi})
\qquad
\bb{f}(\bar{\bb{r}},t) = \sum_n\sum_m \big(\bar{f}^P_{n,m}(\bar{r},t)\,\bb{P}_n^m(\bar{\vartheta},\bar{\varphi},t)+\bar{f}^B_{n,m}(\bar{r},t)\,\bb{B}_n^m(\bar{\vartheta},\bar{\varphi},t)+\bar{f}^C_{n,m}(\bar{r},t)\,\bb{C}_n^m(\bar{\vartheta},\bar{\varphi},t)\big)
$$

with 

$$
\bar{\bb{r}} = \bar{r}\,\bn{r}(\bar{\vartheta},\bar{\varphi})
$$

We note that the spherical coordinates $(\bar{r},\bar{\vartheta},\bar{\varphi})$ indentifying the current position of the particle differs by first-order terms with respect to those of the initial position.
