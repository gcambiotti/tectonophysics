## Non-inertial reference frames

```{div} full-width

Let us denote with  $\bb{X}$ points of the three dimensional space and identify them with respect to an inertial reference frame. In this respect, we decompose $\bb{X}$ into its relative position $\bb{x}$ with respect to a time-dependent origin $\bb{X}_O$

$$
\bb{X}(x_1,x_2,x_3,t) = \bb{X}_O(t) + \bb{x}(x_1,x_2,x_3,t)
$$ (O:1)

where the relative position $\bb{x}$ is expressed with respect to a basis of rotating unit vectors $\bb{e}_i$

$$
\bb{x}(x_1,x_2,x_3,t) = x_i\,\bb{e}_i(t)
$$ (O:2)

In this respect, the Cartesian coordiantes $x_i$ refers to a non-inertial reference frame that, as will discuss in the next chapter, we will choose in such a way that can be considered *attached* to the deformable Earth system. For now, however, we do not specify how the translating origin $\bb{X}_O$ and the angular velocity $\sb{\omega}$ of the rotating unit vectors $\bb{e}_i$ can be choosen.

```

```{admonition} Rotating unit vectors
:class: full-width, dropdown

Starting from a basis of Cartesian unit vectors $\bn{x}_i$, the rotating unit vectors $\bb{e}_i$ can be defined by means of an orthonormal time-dependent rotation tensor $\bb{Q}$

$$
\bb{e}_i(t) = \bb{Q}(t)\cdot\bn{x}_i 
\qquad
\bb{Q}(t)\cdot\bb{Q}^\mathrm{T}(t) = \one
$$ (O:3)

By introducing the following tensor

$$
\bb{W}(t) = \dercla{\bb{Q}(t)}\cdot\bb{Q}^\mathrm{T}(t)
$$ (O:4)

we note that the time derivative of eq. {eq}`O:3` gives

$$
\dercla{\bb{e}_i(t)} = \dercla{\bb{Q}(t)}\cdot\bn{x}_i = \bb{W}(t)\cdot\bn{e}_i(t) 
\qquad
\bb{W}(t) + \bb{W}^\mathrm{T}(t) = \zero
$$ (O:5)

and, so, $\bb{W}$ can be used to compute the time derivative of the rotating unit vectors. Moreover, being anti-symmetric, it can be written as follows

$$
\begin{align}
\bb{W}(t) &= \omega_1(t)\,\big(\bb{e}_3(t)\,\bb{e}_2(t)-\bb{e}_2(t)\,\bb{e}_3(t)\big) +  \omega_2(t)\,\big(\bb{e}_1(t)\,\bb{e}_3(t)-\bb{e}_3(t)\,\bb{e}_1(t)\big)
\\
&+  \omega_3(t)\,\big(\bb{e}_2(t)\,\bb{e}_1(t)-\bb{e}_1(t)\,\bb{e}_2(t)\big) 
\end{align}
$$ (O:6)

where $\omega_i$ are the off-diagonal components with respect to the basis of rotating unit vectors. In light of this, it is straightforward to show that

$$
\dercla{\bb{e}_i(t)} = \sb{\omega}(t)\times\bb{e}_i(t) 
\qquad
\sb{\omega}(t) = \omega_i(t)\,\bb{e}_i(t)
$$ (O:7)

where we have introduced the angular velocity vector $\sb{\omega}$, the components of which are the off-diagonal components of the anti-symmetric tensor $\bb{W}$. 

```

### Lagrangian approach

```{div} full-width

Let us identify a particle of the deformable Earth with a set of Cartesian coordinates $(x_1,x_2,x_3)$ and define its current position at present time $t$ by introducing the displacement field $\bb{u}$ as follows

$$
\bb{R}(x_1,x_2,x_3,t) = \bb{X}(x_1,x_2,x_3,t) + \bb{u}(x_1,x_2,x_3,t) 
$$ (O:8)

This framework well adapts to a translating and rotating deformable body where the time dependent position $\bb{X}$ would describe the position of the particle in the case of no deformation, $\bb{u}=\zero$, and we can account for any deformation with respect to this time dependent position thanks to the displacement field $\bb{u}$. As done for the relative position $\bb{x}$, eq. {eq}`O:2`, we will express the displacement with respect to the same basis of rotating unit vectors 

$$
\bb{u}(x_1,x_2,x_3,t) = u_i(x_1,x_2,x_3,t)\,\bb{e}_i(t)
$$ (O:9)

Alternatively, eq. {eq}`O:8` can be decomposed as follows

$$
\bb{R}(x_1,x_2,x_3,t) = \bb{X}_O(t) + \bb{r}(x_1,x_2,x_3,t)
$$ (O:10)

where $\bb{r}$ denote the relative current position of the particle with respect to the time-dependent origin 

$$
\bb{r}(x_1,x_2,x_2,t) = r_i(x_1,x_2,x_3,t)\,\bb{e}_i(t)
\qquad
r_i(x_1,x_2,x_3,t) = x_i + u_i(x_1,x_2,x_3,t)
$$ (O:11)

This approach for describing the particle position is Lagrangian because the fields depend on the coordinates $x_i$ by means of which we identify the initial position of the particle. In this respect, we will refer to $\bb{u}$ as the Lagrangian displacement field.

```

### Eulerian approach

```{div} full-width

Let us now consider the Eulerian approach and identify the position of a particle at the current time as follows

$$
\bar{\bb{R}}(r_1,r_2,r_3,t) = \bb{X}_O(t) + \bar{\bb{r}}(r_1,r_2,r_3,t)
$$ (O:12)

where $\bar{\bb{r}}$ is the relative position with respect to the time-dependent origin 

$$
\bar{\bb{r}}(r_1,r_2,r_3,t) = r_i\,\bb{e}_i(t)
$$ (O:13)

We note that we use the bar symbol for denoting fields which depend on the current coordinates $r_i$ of the particle, i.e., in the Eulerian representation. 

By equating the Lagrangian and Eulerian representations of the current position, eqs. {eq}`O:10` and {eq}`O:12`, and defining the Eulerian displacement $\bar{\bb{u}}$ as follows

$$
\bar{\bb{u}}(r_1,r_2,r_3,t) = \bar{u}_i(r_1,r_2,r_3,t)\,\bb{e}_i(t)
\qquad
\bar{u}_i(r_1,r_2,r_3,t) = u_i(x_1,x_2,x_3,t)
$$ (O:14)

we obtain the Eulerian representation of the initial position of the particle

$$
\bar{\bb{x}}(r_1,r_2,r_3,t) = \bar{\bb{r}}(r_1,r_2,r_3,t) - \bar{\bb{u}}(r_1,r_2,r_3,t) \qquad \bar{x}_i(r_1,r_2,r_3,t) = r_i - \bar{u}_i(r_1,r_2,r_3,t) 
$$ (O:14a)

```

### Gradient operators 

```{div} full-width

By defining the gradient operators with respect to the (relative) initial and current coordinates of the particle

$$
\grad = \bb{e}_i(t)\,\frac{\partial }{\partial x_i}
\qquad
\bar{\grad} = \bb{e}_i(t)\,\frac{\partial }{\partial r_i}
$$ (O:15)

we can write

$$
\grad \bb{r}(x_1,x_2,x_3,t) = \one + \grad\bb{u}(x_1,x_2,x_3,t)
\qquad
\bar{\grad}\bar{\bb{x}}(r_1,r_2,r_3,t) = \one - \bar{\grad}\bar{\bb{u}}(r_1,r_2,r_3,t)
$$ (O:16)

Also, by considering that $\bar{\grad}\bar{\bb{r}}=\one$ and noting that

$$
\begin{align}
\bar{\grad}\bar{\bb{r}}(r_1,r_2,r_3,t) &= \frac{\partial r_j}{\partial r_i}\,\bb{e}_i(t)\,\bb{e}_j(t) = \frac{\partial \bar{x}_k(r_1,r_2,r_3,t)}{\partial r_i}\,\frac{\partial r_j(x_1,x_2,x_3,t)}{\partial x_k}\,\bb{e}_i(t)\,\bb{e}_j(t) 
\end{align}
$$ (O:17a)

we can obtain the following relation between the gradients of the Eulerian and Lagragian representation of the displacement field

$$
\begin{align}
& \big(\one - \bar{\grad}\bar{\bb{u}}(r_1,r_2,r_3,t)\big)\cdot\big(\one+\grad\bb{u}(x_1,x_2,x_3,t)\big) = \one 
\end{align}
$$ (O:17)

which means that

$$
 \big(\one - \bar{\grad}\bar{\bb{u}}(r_1,r_2,r_3,t)\big)^{-1} = \big(\one+\grad\bb{u}(x_1,x_2,x_3,t)\big) 
$$ (O:18)

```



```{admonition} Infinitesimal deformations
:class: full-width

In the following, we shall specify some results within the assumption of intinitesimal deformations. This means that the displacement field $\bb{u}$ and its gradient $\grad\bb{u}$ are assumed to be infinitesimal and it is straightforward to show that, at the first-order approximation (i.e, neglecting second order terms and higher),  eq. {eq}`O:18` gives

$$
\bar{\grad}\bar{\bb{u}}(r_1,r_2,r_3,t) \approx \grad\bb{u}(x_1,x_2,x_3,t)
$$ (O:19)

```

#### Change of variable

```{div} full-width

Let us denote with $\mathcal{V}_t$ a three-dimensional region deforming according to the displacement field $\bb{u}$ and consider the volume integral of a field

$$
F(t) = \int_{\mathcal{V}_t} \bar{f}(r_1,r_2,r_3,t)\,\md \bar{V}
$$ (OC:1)

with $\md \bar{V}=\md r_1\,\md r_2\,\md r_3$. After the change of variables from the current to the initial coordinates, $r_i\rightarrow x_i$, eq. {eq}`OC:1` becomes

$$
F(t) = \int_{\mathcal{V}_0} f(x_1,x_2,x_3,t)\,J(x_1,x_2,x_3,t)\,\md V
$$ (OC:2)

where we use the fact that

$$
\md \bar{V} = J(x_1,x_2,x_3,t)\,\md V 
$$ (OC:3)

with $\md V=\md x_1\,\md x_2\,\md x_3$ and $J$ being the Jacobian of the change of variables

$$
\qquad J(x_1,x_2,x_3,t) = \mathrm{det}\big[\grad\bb{r}(x_1,x_2,x_3,t)\big] = \mathrm{det}\big[\one+\grad\bb{u}(x_1,x_2,x_3,t)\big] 
$$ (OC:4)

We note that the relative volume variations can be expressed as follows

$$
\chi(x_1,x_2,x_3,t) = \frac{\md \bar{V}-\md V}{\md V} =  \mathrm{det}\big[\one+\grad\bb{u}(x_1,x_2,x_3,t)\big] - 1 \approx \grad\cdot\bb{u}(x_1,x_2,x_3,t)
$$ (OC:5)

where the approximation holds under the assumption of infinitesimal deformation.

```

(sec-notations)=
### Notations

```{div} full-width

In the following, for the sake of brevity, we will denote the spatial dependence on the initial and current coordinates of Lagrangian, $f$, and Eulerian, $\bar{f}$, fields as follows

$$
f(\bb{x},t) = f(x_1,x_2,x_3,t) = \bar{f}(r_1,r_2,r_3,t) = \bar{f}(\bb{r},t) 
$$ (O:20)

and we will omit to explicitate the spatial and temporal dependences on the initial, $\bb{x}$, and current, $\bb{r}$, coordinates and avoid to distinguish between $\bar{\bb{x}}$ and $\bb{x}$ or between $\bar{\bb{r}}$ and $\bb{r}$. In this respect, eqs. {eq}`O:1`, {eq}`O:8` and {eq}`O:12` become

$$
\bb{X}(\bb{x},t) = \bb{X}_O(t) + \bb{x} = \bar{\bb{X}}(\bb{r},t) \qquad \bb{R}(\bb{x},t) = \bb{X}_O(t) + \bb{r} = \bar{\bb{R}}(\bb{r},t) 
$$ (O:21)

with

$$
\bb{r} = \bb{x} + \bb{u}(\bb{x},t)
\qquad
\bb{x} = \bb{r}-\bar{\bb{u}}(\bb{r},t)
$$ (O:22)

Also, we can write the following gradients of the current and initial positions

$$
\begin{align}
& \grad\bb{r} = \grad\big(\bb{x}+\bb{u}(\bb{x},t)\big) = \one + \grad\bb{u}(\bb{x},t)
\qquad \grad\bb{x} = \one
\\
& \bar{\grad}\bb{x} = \bar{\grad}\big(\bb{r}-\bar{\bb{u}}(\bb{r},t)\big) = \one - \bar{\grad}\bar{\bb{u}}(\bb{r},t) \qquad \bar{\grad}\bb{r}=\one
\end{align}
$$ (O:23)

while eq. {eq}`O:18` becomes

$$
\bar{\grad}\bb{r} = \bar{\grad}\bb{x}\cdot\grad\bb{r} = \big(\one - \bar{\grad}\bar{\bb{u}}(\bb{r},t)\big)\cdot\big(\one + \grad\bb{u}(\bb{x},t)\big) = \one
$$ (O:24)

that, at the first-order, implies

$$
\bar{\grad}\bar{\bb{u}}(\bb{r},t) \approx \grad\bb{u}(\bb{x},t)
$$ (O:19a)

```

## Initial and incremental fields

```{div} full-width

Let us now consider a field $f$ and decompose it into its initial, $f_0$, and local, $\loc f$, or material, $\mat f$, increments

$$
\bar{f}(\bb{r},t) = f_0(\bb{r}) + \loc \bar{f}(\bb{r},t) = f_0(\bb{x}) + \mat f(\bb{x},t) =  f(\bb{x},t) 
$$ (O:25)

We note that it is natural to define the local incremental fields within the Eulerian framework, while the material incremental fields within the Lagrangian one. Nevertheless, we can always consider the Lagrangian description of a local increment or the Eulerian description of a material increment

$$
\loc\bar{f}(\bb{r},t) = \loc f(\bb{x},t) \qquad \mat\bar{f}(\bb{r},t) = \mat f(\bb{x},t) 
$$ (O:26)

From the comparison of the local or material incremental decompositions, we can obtain  the following relation between the local and material increments

$$
\loc\bar{f}(\bb{r},t) = \loc f(\bb{x},t) = f_0(\bb{x}) - f_0(\bb{r}) + \mat f(\bb{x},t) 
$$ (O:27)

that, at the first order, simplifies into

$$
\loc f(\bb{x},t) \approx  -\bb{u}(\bb{x},t)\cdot\grad f_0(\bb{x})  + \mat f(\bb{x},t) 
$$ (O:28)

```

### Gradients

```{div} full-width

Let us consider a scalar field $f$ and takes its gradient with respect to the current coordinates

$$
\begin{align}
\bar{\grad} \bar{f}(\bb{r},t) = \bar{\grad} f(\bb{x},t) &= \bb{e}_j(t)\,\frac{\partial x_k}{\partial r_j}\,\frac{\partial f(\bb{x},t)}{\partial x_k} = \big(\one-\bar{\grad}\bar{\bb{u}}(\bb{r},t)\big)\cdot\grad f(\bb{x},t) 
\\
&= \big(\one+\grad\bb{u}(\bb{x},t)\big)^{-1}\cdot\grad f(\bb{x},t)
\end{align}
$$ (O:29)

and note that, at the first order, it becomes

$$
\bar{\grad} \bar{f}(\bb{r},t) \approx \grad f(\bb{x},t) - \grad\bb{u}(\bb{x},t)\cdot \grad f(\bb{x},t)
$$ (O:30)

Moreover, whe the incremental field is already a first-order term, we note that also the second term in the right-hand side of the above equation can be neglected. For instance, gradients of (infinitesimal) local and material increments give

$$
\bar{\grad} \loc\bar{f}(\bb{r},t) \approx \grad \loc f(\bb{x},t) 
\qquad
\bar{\grad} \mat\bar{f}(\bb{r},t) \approx \grad \mat f(\bb{x},t) 
$$ (O:31)

A similar result holds also for the Laplacian or the divergence of incremental fields

$$
\begin{align}
& \bar{\nabla}^2 \loc\bar{f}(\bb{r},t) \approx \nabla^2 \loc f(\bb{x},t)  \qquad &
\bar{\grad}\cdot\loc\bar{\bb{f}}(\bb{r},t) \approx \grad\cdot\loc \bb{f}(\bb{x},t) 
\\
& \bar{\nabla}^2 \mat\bar{f}(\bb{r},t) \approx \nabla^2 \mat f(\bb{x},t)  
\qquad &
\bar{\grad}\cdot\mat\bar{\bb{f}}(\bb{r},t) \approx \grad\cdot\mat \bb{f}(\bb{x},t)  
\end{align}
$$ (O:32)

```

## Time derivatives

```{div} full-width

In the following we will consider Lagrangian and Eulerian vector fields expressed with respect to the rotating unit vectors

$$
\begin{align} &
\bb{f}(\bb{x},t) = f_i(\bb{x},t)\,\bb{e}_i(t) =
 \bar{f}_i(\bb{r},t)\,\bb{e}_i(t) = \bar{\bb{f}}(\bb{r},t) 
\end{align}
$$ (O:33)

In the perspective of differentiation with respect to time, we shall take into account the fact that the rotating unit vectors also depend on time and write

$$
\begin{align} &
\dercla{\bb{f}(\bb{x},t)} = \derpar{ f_i(\bb{x},t)}\,\bb{e}_i(t) + f_i(\bb{x},t)\,\sb{\omega}(t)\times\bb{e}_i(t) = \derpar{\bb{f}(\bb{x},t)} + \sb{\omega}(t)\times\bb{f}(\bb{x},t)
\\
& \dercla{\bar{\bb{f}}(\bb{r},t)} = \derpar{ \bar{f}_i(\bb{r},t)}\,\bb{e}_i(t) + \bar{f}_i(\bb{r},t)\,\sb{\omega}(t)\times\bb{e}_i(t) = \derpar{\bar{\bb{f}}(\bb{r},t)} + \sb{\omega}(t)\times\bar{\bb{f}}(\bb{r},t)
\end{align}
$$ (O:34)

where $\partial/\partial t$ is the time derivative operator acting only on the components of the vector field 

$$
\begin{align} &
\derpar{\bb{f}(\bb{x},t)} =  \derpar{ f_i(\bb{x},t)}\,\bb{e}_i(t) 
\\
& \derpar{\bar{\bb{f}}(\bb{r},t)} =  \derpar{ \bar{f}_i(\bb{r},t)}\,\bb{e}_i(t) 
\end{align}
$$ (O:35)

We note that this convenction about the time derivative, together with the notation introduced in  {numref}`Section %s <sec-notations>` - {ref}`sec-notations`, can arise some confusion when we consider the time derivative of the initial and current positions of the particle. In this respect, we note that, within the Lagrangian framework, setting $\bb{f}(\bb{x},t) = \bb{x}$ or $\bb{f}(\bb{x},t) = \bb{r}=\bb{x}+\bb{u}$, we have

$$
\begin{align}
& \dercla{\bb{x}} = \derpar{x_i}\,\bb{e}_i(t) + \sb{\omega}(t)\times\bb{x} = \sb{\omega}(t)\times\bb{x} 
\\
& \dercla{\bb{r}} = \derpar{\big(x_i+u_i(\bb{x},t)\big)}\,\bb{e}_i(t) + \sb{\omega}(t)\times\bb{r} = \derpar{\bb{u}(\bb{x},t)} + \sb{\omega}(t)\times\bb{x}  
\end{align}
$$ (O:36)

Within the Eulerian framework, instead, setting  $\bar{\bb{f}}(\bb{r},t) = \bb{x}=\bb{r}-\bar{\bb{u}}$ or $\bar{\bb{f}}(\bb{r},t) = \bb{r}$, we have

$$
\begin{align}
& \dercla{\bb{x}} = \derpar{\big(r_i-\bar{u}_i(\bb{r},t)\big)}\,\bb{e}_i(t) + \sb{\omega}(t)\times\bb{x} = -\derpar{\bar{\bb{u}}(\bb{r},t)} + \sb{\omega}(t)\times\bb{x} 
\\
& \dercla{\bb{r}} = \derpar{r_i}\,\bb{e}_i(t) + \sb{\omega}(t)\times\bb{r} = \sb{\omega}(t)\times\bb{r}   
\end{align}
$$ (O:37)

We note that $\md \bb{x}/\md t$ and $\md\bb{r}/\md t$ return different results depending whether they make sense within the Lagrangian or Eulerian frameworks and, so, we shall always remember which framework we are adopting before taking time derivaties. From a physical point of view, this difference is correct because the time derivative takes different meaning in the two frameworks. In the Lagrangian one, it gives the rate of change of the field value associated to a given particle. In the Eulerian framework, instead, it gives the rate of change of the field value at a given observation point (that is occupied by different particles through time). In particular, focusing on $\md\bb{r}/\md t$ in the Lagrangian framework, we can understand that it describes the velocity $\bb{v}^R$ of the particle in the non-inertial translating frame and that this velocity can be seen as the velocity due to the rotation of the rotating reference frame, $\sb{\omega}\times\bb{r}$, and the relative velocity with respect to the rotating reference frame, $\bb{v}=\partial \bb{u}/\partial t$,

$$
\bb{v}^R(\bb{x},t) = \dercla{\bb{r}} = \bb{v}(\bb{x},t) + \sb{\omega}(t)\times\bb{x}  
\qquad
\mathrm{with}
\qquad
\bb{v}(\bb{x},t) = \derpar{\bb{u}(\bb{x},t)}
$$ (O:38)

In the end, we note that the above distinction between the time derivative operators $\md/\md t$ and $\partial/\partial t$ is relevant only for vector (or tensor) fields. For scalar fields, instead, the two operators coincide because there are no rotating unit vectors to be derived and we can write

$$
\dercla{f(\bb{x},t)} = \derpar{f(\bb{x},t)}
\qquad
\dercla{\bar{f}(\bb{r},t)} = \derpar{\bar{f}(\bb{r},t)}
$$ (O:39)

A similar coincidence occurs, as special case, for the time derivative of the angular velocity vector which gives

$$
\dercla{ \sb{\omega}(t)} = \derpar{\sb{\omega}(t)} + \sb{\omega}(t)\times\sb{\omega}(t) = \derpar{\sb{\omega}(t)}
$$ (O:40)

```

### Material time derivatives

```{div} full-width

Let us now introduce the material time derivative as the rate of change of the field value associated to a given particle. Within the Lagrangian framework, it corresponds to the classical time derivative

$$
\dermat{f(\bb{x},t)} = \dercla{f(\bb{x},t)}
$$ (O:41)

Within the Eulerian framework, instead, we have to consider that the time dependence of the current coordinates of the particle and, so, the material time derivative gives

$$
\dermat{\bar{f}(\bb{r},t)} = \dercla {\bar{f}(\bb{r},t)} + \derpar{ r_j}\,\rpar{f(\bb{r},t)}{j} = \dercla {\bar{f}(\bb{r},t)} + \bar{\bb{v}}(\bb{r},t)\cdot\bar{\grad} f(\bb{r},t)
$$ (O:42)

where $\bar{\bb{v}}$ is the Eulerian representation of the relative velocity with respect to the rotating reference frame 

$$
\bar{\bb{v}}(\bb{r},t) =  \bb{v}(\bb{x},t) = \derpar{\bb{u}(\bb{x},t)}
$$ (O:43)

Obviously, by definition of material time derivative, we have

$$
\dermat{\bar{f}(\bb{r},t)} = \dermat{f(\bb{x},t)} 
$$ (O:44)

In order to understand the consistency and the physical meaning of the material time derivative in the Eulerian framework, let us consider the Eulerian vector field $\bar{\bb{f}}(\bb{r},t)=\bb{r}$ and its material time derivatives

$$
\dermat{\bb{r}} = \frac{\md \bb{r}}{\md t} + \bar{\bb{v}}(\bb{r},t)\cdot\bar{\grad}\bb{r} = \sb{\omega}(t)\times \bb{r} + \bar{\bb{v}}(\bb{r},t)\cdot\one =  \sb{\omega}(t)\times \bb{r} + \bar{\bb{v}}(\bb{r},t)
$$ (O:45)

As it concerns the Lagrangian framework, by considering $\bb{f}(\bb{x},t) = \bb{r}$, we obtain

$$
\dermat{\bb{r}} = \dercla{\bb{r}} = \sb{\omega}(t)\times \bb{r} + \bb{v}(\bb{x},t)
$$ (O:46)

and, so, by comparing it with eq. {eq}`O:45` , we correctly obtain that $\bar{\bb{v}}$ is the Eulerian representation of the Lagrangian field $\bb{v}$, as stated by eq. {eq}`O:43`.

```

### Velocity and acceleration

```{div} full-width

Let us define the velocity and acceleration in the Lagrangian framework

$$
\begin{align}
& \bb{V}(\bb{x},t) = \dermat{\bb{R}(\bb{x},t)} =  \dercla{\bb{X}_O(t)} + \sb{\omega}(t)\times\bb{r} + \bb{v}(\bb{x},t)
\\
& \bb{A}(\bb{x},t) = \dermat{\bb{V}(\bb{x},t)} =  \ddercla{\bb{X}_O(t)} + \sb{\omega}(t)\times\sb{\omega}(t)\times\bb{r} +2\,\sb{\omega}(t)\times\bb{v}(\bb{x},t) + \derpar{\sb{\omega}(t)}\times\bb{r} + \bb{a}(\bb{x},t)
\end{align}
$$ (O:47)

where $\bb{a}$ is the relative acceleration

$$
\bar{\bb{a}}(\bb{r},t) = \bb{a}(\bb{x},t) = \derpar{\bb{v}(\bb{x},t)} = \dderpar{\bb{u}(\bb{x},t)}
$$ (O:48)

while the first term in the right-hand side is the acceleration of the translating origin, and the second, third and fourth terms are the centrifugal, Coriolis and Euler acceleration.

Let us consider also their representations in the Eulerian framework

$$
\begin{align}
& \bar{\bb{V}}(\bb{r},t) =  \dercla{\bb{X}_O(t)} + \sb{\omega}(t)\times\bb{r} + \bar{\bb{v}}(\bb{r},t)
\\
& \bar{\bb{A}}(\bb{r},t) = \ddercla{\bb{X}_O(t)} + \sb{\omega}(t)\times\sb{\omega}(t)\times\bb{r} +2\,\sb{\omega}(t)\times\bar{\bb{v}}(\bb{r},t) + \derpar{\sb{\omega}(t)}\times\bb{r} + \bar{\bb{a}}(\bb{r},t)
\end{align}
$$ (O:49)


In order to better understand the meaning of the relative acceleration with respect to rotating unit vectors in the Eulerian representation, $\bar{\bb{a}}$, we now consider the material time derivative of the Eulerian velocity field

$$
\begin{align} 
\bar{\bb{A}}(\bb{r},t) &= \ddermat{ \bar{\bb{V}}(\bb{r},t)} = \ddercla{ \bb{X}_O(t)} + \dermat{ \big(\sb{\omega}(t)\times\bb{r}\big)} + \dermat{\bar{\bb{v}}(\bb{r},t)}
\\
&= \ddercla{ \bb{X}_O(t)} + \dercla{ \big(\sb{\omega}(t)\times\bb{r}\big)} +\bar{\bb{v}}(\bb{r},t)\cdot\bar{\grad}\big(\sb{\omega}(t)\times\bb{r}\big)+ \dercla{\bar{\bb{v}}(\bb{r},t)} +\bar{\bb{v}}(\bb{r},t)\cdot\bar{\grad}\bar{\bb{v}}(\bb{r},t)
\\
&= \ddercla{ \bb{X}_O(t)} + \derpar{\sb{\omega}(t)}\times\bb{r} + \sb{\omega}(t)\times\sb{\omega}(t)\times\bb{r} +2\,\sb{\omega}(t)\times\bar{\bb{v}}(\bb{r},t) 
\\
&+\derpar{\bar{\bb{v}}(\bb{r},t)} +\bar{\bb{v}}(\bb{r},t)\cdot\bar{\grad}\bar{\bb{v}}(\bb{r},t)
\end{align}
$$ (O:50)

where we have used the following identity

$$
\bar{\bb{v}}(\bb{r},t)\cdot\bar{\grad}\big(\sb{\omega}(t)\times\bb{r}\big) = \sb{\omega}(t)\times\bar{\bb{v}}(\bb{r},t)
$$ (O:50a)

Then, from the comparison with eq. {eq}`O:49` , we understand that


$$
\bar{\bb{a}}(\bb{r},t)= \derpar{\bar{\bb{v}}(\bb{r},t)} +\bar{\bb{v}}(\bb{r},t)\cdot\bar{\grad}\bar{\bb{v}}(\bb{r},t)
$$ (O:51)

which corresponds to the material time derivative under the assumption that the rotating unit vectors do not rotate.

```

#### Centrifugal potential

```{div} full-width

Let us define the centrifugal potential within the Eulerian framework as follows

$$
\bar{\psi}^C(\bb{r},t) = \frac{(\sb{\omega}(t)\cdot\bb{r})^2 - \omega^2(t)\,r^2}{2}
$$ (O:52)

in such a way that its gradient yields

$$
\bar{\grad}\bar{\psi}^C(\bb{r},t) = \sb{\omega}(t)\times\sb{\omega}(t)\times\bb{r}
$$ (O:53)

In this respect, eq. {eq}`O:49` becomes

$$
\begin{align} 
\bar{\bb{A}}(\bb{r},t) 
&=  \ddercla{ \bb{X}_O(t)} + \bar{\grad}\bar{\psi}^C(\bb{r},t) +  \bar{\bb{a}}(\bb{r},t) + \derpar{\sb{\omega}(t)}\times\bb{r}  +2\,\sb{\omega}(t)\times\bar{\bb{v}}(\bb{r},t)  
\end{align}
$$ (O:54)

```

(sec-constitutive)=
## Consitutive equations

```{div} full-width

In the perspective of definining consitutive equation relating the Cauchy stress tensor to deformations, let us first characterize in details what the deformations are. In this respect, within the assumption of infinitesimal deformation ($\bb{u}\ll 1$ and $\grad\bb{u}\ll 1$), we first introduce the strain tensor 

$$
\bar{\sb{\varepsilon}}(\bb{r},t) = \sb{\varepsilon}(\bb{x},t) = \frac{\grad\bb{u}(\bb{x},t)+\big(\grad\bb{u}(\bb{x},t)\big)^\mathrm{T}}{2}
$$ (O:55)

and the strain rate tensor

$$
\bar{\bb{d}}(\bb{r},t) = \bb{d}(\bb{x},t) = \derpar{\sb{\varepsilon}(\bb{x},t)} = \frac{\grad\bb{v}(\bb{x},t)+\big(\grad\bb{v}(\bb{x},t)\big)^\mathrm{T}}{2}
$$ (O:56)

We note that, in defining the strain rate tensor, we made use of the time derivative $\partial/\partial t$ acting only on the components of the (vector or tensor) fields with respect to the rotating unit vector. This is due to the fact, as discussed in {download}`Chapter 5 of Martinec (2019)<https://unimibox.unimi.it/index.php/s/zNEPDyefFrdW7tT>`, consitutive equations have to be frame-indifferent, which means that they cannot depend on time dependent translations and rotations. In particular, they cannot depend on the choice of the translating origin $\bb{X}_O$ and of the angular velocity vector $\sb{\omega}$. 

In the following we shall also consider the isotropic and deviatoric decompositions of the strain and strain rate tensors

$$
\sb{\varepsilon}(\bb{x},t) = \chi(\bb{x},t)\,\one + \sb{\varepsilon}^d(\bb{x},t)
\qquad
\bb{d}(\bb{x},t) = \derpar{\chi(\bb{x},t)}\,\one + \bb{d}^d(\bb{x},t)
$$ (O:57)

where, according to eq. {eq}`OC:5`, $\chi$ is the relative volume variation (or the trace of the strain tensor) 

$$
\chi(\bb{x},t) = \grad\cdot\bb{u}(\bb{x},t) \qquad \derpar{\chi(\bb{x},t)} = \grad\cdot\bb{v}(\bb{x},t)
$$ (O:58)

```



### The Hooke law

```{div} full-width

For an isotropic elastic material, assuming an initial state of hydrostatic equilibrium, the constituive equation relating the Cauchy stress to the deformation reads

$$
\stress(\bb{x},t) = -p_0(\bb{x})\,\one + \mat\stress(\bb{x},t)
$$ (O:59)

where $p_0$ is the initial hydrostatic pressure and $\mat\stress$ is the material incremental Cauchy stress given by the Hooke law

$$
\mat\stress(t) = \lambda\,\chi(t)\,\one+2\,\mu(\bb{x})\,\sb{\varepsilon}(\bb{x},t) = \kappa\,\chi(t)\,\one+2\,\mu\,\sb{\varepsilon}^d(t)
$$ (O:60)

Here, for the sake of brevity and clarity, we omit the dependence on the spatial coordinates $\bb{x}$. Also, $\lambda$ and $\mu$ are the two Lam√® parameters and $\kappa$ is the bulk modulus

$$
\kappa = \lambda + \frac{2}{3}\,\mu
$$ (O:61)

We note that these elasic parameters can also depend on the spatial coordinates, in the case of heterogeneous materials. 

```

(sec-linear-viscoelastic)=
### Linear viscoelastic materials

```{div} full-width

The Cauchy stress tensor for a linear viscoelastic material takes the same form of eq. {eq}`O:59` but for the fact that the material increment is now defined thorough the time convolution of the so called shear relaxation function $q$ and the past deviatoric strain rate history 

$$
\mat\stress(t) = \kappa\,\chi(t)\,\one + 2\,\mu\,\int_{-\infty}^\infty q(\tau)\,\bb{d}^d(t-\tau)\,\md \tau
$$ (O:62)

with the requirement that the shear relaxation function is causal and decays increasing time (thus limiting the memory of the viscoelastic material to just the near past history)

$$
q(t<0) = 0 \qquad \lim_{t\rightarrow \infty} q(t) = 0
$$ (O:63)

As an examaple, we consider the shear relaxation function for the so-called Maxwell viscoelastic body

$$
q(t) = e^{-t/\tau_m}
$$ (O:64)

with $\tau_m$ being the Maxwell time, which can depend on the spatial coordinates $\bb{x}$.

```

```{warning} 
:class: full-width

We note that the meaning of time convolution in eq. {eq}`O:62` can be misleading, especially in the case in which the deviatoric strain rate tensor is expressed with respect to the rotating unit vectors

$$
\bb{d}^d(t) = d_{ij}^d(t)\,\bb{e}_i(t)\,\bb{e}_j(t)
$$ (O:65)

Indeed, we could think to take into account the time dependence of the rotating unit vectors and write eq. {eq}`O:62`

$$
\mat\stress(t) = \kappa\,\chi(t)\,\one + 2\,\mu\,\int_{-\infty}^\infty q(\tau)\,d_{ij}^d(t-\tau)\,\bb{e}_i(t-\tau)\,\bb{e}_j(t-\tau)\,\md \tau
$$ (O:66)

Different, in view of the requirement of frame indifference, the time convolution enetring eq. {eq}`O:62` must be understood in the following sense

$$
\mat\stress(t) = \kappa\,\chi(t)\,\one + 2\,\mu\,\left(\int_{-\infty}^\infty q(\tau)\,d_{ij}^d(t-\tau)\,\md \tau\right)\,\bb{e}_i(t)\,\bb{e}_j(t)
$$ (O:67)

for the same reason why we consider the gradient of relative velocity, $\bb{v}$ in defining the strain rate tensor, without including the contribution due to rotation velocity $\sb{\omega}\times\bb{r}$.

```




#### The fluid limit

```{div} full-width

Let us now assume that the deviatoric strain rate is constant (or change slowly though time on time scales much larger than the Maxwell time). In this case, eq. {eq}`O:62` gives

$$
\mat\stress(t) = \kappa\,\chi(t)\,\one + 2\,\mu\,\bb{d}^d(t)\,\int_{-\infty}^\infty q(\tau)\,\md \tau = \kappa\,\chi(t)\,\one + 2\,\eta\,\bb{d}^d(t)
$$ (O:68)

where $\eta$ is the steady-state viscosity

$$
\eta = \mu\,\tau_m
$$ (O:69)

This example show how, under certain assumptions on the deviatoric strain history, the viscoelastic body behaves as a fluid, where the deviatoric stress is indeed related to the instantaneous deviatoric strain rate multiplied by the steady-state viscotity. We note that this reasoning holds under the assumptions that $\grad \bb{u}\ll 1$, which implies both infinitesimal deformation and rotation.

```

## Correspondece principle

```{div} full-width

After Laplace transform, the momentum and Poisson equations for a linear viscoelastic solid become formally equivalent to those for an elastic solid, but for the fact that the displacement and potential fields are substituted by their Laplace transform while the shear modulus is substitued by the equivalent shear modulus. After inverse Laplace transform of the viscoelastic solution in the Laplace domain (formally equivalent to the elastic solution), we will obtain the solution to the viscoelastic problem in the time domain.

In order to better understand this possibility that we shall exploit later in the development of the theoretical framework, we now focus on the Laplace transform of the material incremental Cauchy stress tensor.

```


(sec-equivalent-shear)=
### Equivalent shear modulus


```{div} full-width

After Laplace transform and making use of eqs. {eq}`L:9` and {eq}`L:11`, the viscoelastic constitutive equation {eq}`O:62` can be recast as follows

$$
\begin{align}
\delta\tilde{\boldsymbol{\sigma}}(s) = \kappa\,\tilde{\chi}(s)\,\mathbf{1} + 2\,\hat{\mu}(s)\,\tilde{\boldsymbol{\varepsilon}}^d(s)
\end{align}
$$ (O:70)

where we made use of the fact that $\tilde{\bb{d}}(s) = s\,\tilde{\sb{\varepsilon}}$, and 
$\hat{\mu}$ is the equivalent shear modulus defined as function of the Laplace variable

$$
\hat{\mu}(s) = \mu\,s\,\tilde{q}(s)
$$  (O:71)


We note that eq. {eq}`O:70` is formally equivalent to the Hooke law for an elastic material, eq. {eq}`O:60`, although it is defined in terms of the Laplace transforms of the stress and strain tensors. This equivalence will allows us to obtain the solution of a (linear) viscoelastic problem in the Laplace domain by solving an equivalent elastic problem, with the shear modulus $\mu$ substituted by its equivalent $\hat{\mu}$. 

In the case of a Maxwell body, for which the relaxation function takes the form given by eq. {eq}`O:64` and in view of eq. {eq}`L:31`, the Laplace transform of the shear relaxation function gives

$$
\tilde{q}(s) = \frac{1}{s+\frac{1}{\tau_\mathrm{m}}}
$$  (O:72)

The equivalent shear modulus thus becomes

$$
\hat{\mu}(s) = \frac{\mu\,s}{s+\frac{1}{\tau_\mathrm{m}}}
$$  (O:73)


```


### Relataxtion experiment

```{div} full-width

The relaxation experiment assumes that the deviatoric strain tensor is constant, $\boldsymbol{\varepsilon}_0^d$, starting from the initial time $t=0$

$$
\begin{align}
\boldsymbol{\varepsilon}^d(t) = \boldsymbol{\varepsilon}_0^d\,H(t) 
\end{align}
$$  (O:74)

The deviatoric strain rate is thus impulsive

$$
\begin{align}
\bb{d}(t) = \frac{\partial \boldsymbol{\varepsilon}^d(t)}{\partial t} =  \boldsymbol{\varepsilon}_0^d\,\delta(t) 
\end{align}
$$  (O:75)

and, from the Maxwell body, the incremental stress becomes

$$
\begin{align}
\delta\boldsymbol{\sigma}(t) = \kappa\,\chi(t)\,\mathbf{1} + 2\,\mu\,\boldsymbol{\varepsilon}_0^d\,(\delta\times q)(t) = \kappa\,\chi(t)\,\mathbf{1} + 2\,\mu\,\boldsymbol{\varepsilon}_0^d\,q(t) 
\end{align}
$$  (O:76)

We thus learn that the deviatoric stress decays with time as the shear relaxation function.

```

### Creep experiment

```{div} full-width

The creep experiment assumes that the deviatoric strain tensor grows linearly with time starting from the initial time $t=0$

$$
\begin{align}
\boldsymbol{\varepsilon}^d(t) =  t\,\dot{\boldsymbol{\varepsilon}}_0^d\,H(t) 
\end{align}
$$  (O:77)

It thus corresponds to the case in which the deviatoric strain rate is constant, $\dot{\boldsymbol{\varepsilon}}_0^d$, after the initial time

$$
\begin{align}
\bb{d}(t) = \dot{\boldsymbol{\varepsilon}}^d(t) = \dot{\boldsymbol{\varepsilon}}_0^d\,H(t)
\end{align}
$$  (O:78)

and, from the viscoelastic constitutive equation, the incremental stress  becomes

$$
\begin{align}
\delta\boldsymbol{\sigma}(t) = \kappa\,\chi(t)\,\mathbf{1} + 2\,\mu\,\dot{\boldsymbol{\varepsilon}}_0^d\,(H\star q)(t)
\end{align}
$$  (O:79)

```

#### Maxwell body and steady-state viscosity

```{div} full-width

By considering that

$$
(H\star q)(t) = \int_0^t e^{-(t-\tau)/\tau_\mathrm{m}}\,d\tau =e^{-t/\tau_\mathrm{m}}\,\int_0^t e^{\tau/\tau_\mathrm{m}}\,d\tau = \tau_\mathrm{m}\,\left(1-e^{-t/\tau_\mathrm{m}}\right)
$$  (O:80)

we can specify eq. {eq}`O:79` as follows

$$
\begin{align}
\delta\boldsymbol{\sigma}(t) = \kappa\,\chi(t)\,\mathbf{1} + 2\,\eta\,\dot{\boldsymbol{\varepsilon}}_0^d\,\left(1-e^{-t/\tau_\mathrm{m}}\right)
\end{align}
$$  (L:81)

with $\eta$ being the steady-state viscosity already defined by eq. {eq}`O:69`. We thus learn that, at large times, in the limit for  $t\rightarrow\infty$, the deviatoric stress converges to a finite value proportional to the constant deviatoric strain rate. In this respect, $\eta$ can be regarded as the steady-state viscosity.

The following figure shown the dependence on time of the deviatoric stress for unitary deviatoric strain and strain rate in the case of the relaxation and creep experiment
```

````python
import numpy as np
import matplotlib.pyplot as plt

maxwell_time = 1  # [kyr]
q = lambda t: np.exp(-t/maxwell_time)*(t>=0)
creep_experiment = lambda t: (1-np.exp(-t/maxwell_time))*(t>=0)/maxwell_time

ts = np.linspace(-1,10,6001)

fig, axes = plt.subplots(1,2,figsize=(10,3),sharex=True,sharey=True)
axes[0].plot(ts,q(ts));
axes[0].set_title("Relaxation experiment")
axes[1].plot(ts,creep_experiment(ts))
axes[1].set_title("Creep experiment")

axes[0].set_ylabel(r"Deviatoric stress $\sigma^d\,/\,(2\,\mu)$")
axes[1].set_ylabel(r"Deviatoric stress $\sigma^d\,/\,(2\,\eta)$")
for ax in axes:
    ax.set_xlabel("Time [kyr]")
    ax.axhline(0,color="black",linewidth=0.5)
    ax.axhline(1,color="black",linewidth=0.5)
    ax.axvspan(0,maxwell_time,alpha=0.3,color="red")
    ax.axvspan(0,2*maxwell_time,alpha=0.3,color="red")
````
