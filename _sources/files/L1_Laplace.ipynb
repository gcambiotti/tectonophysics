
$\newcommand{\md}{\mathrm{d}\hspace{0cm}}$
$\newcommand{\mdd}{\mathrm{d}\hspace{0cm}^2}$
$\newcommand{\grad}{\boldsymbol{\nabla}\hspace{-0.01cm}}$
$\newcommand{\stress}{\boldsymbol{\sigma}}$
$\newcommand{\bn}[1]{\hat{\mathbf{#1}}}$
$\newcommand{\bb}[1]{\mathbf{#1}}$
$\newcommand{\sn}[1]{\hat{\boldsymbol{#1}}}$
$\newcommand{\sb}[1]{\boldsymbol{#1}}$
$\newcommand{\zero}{\mathbf{0}}$
$\newcommand{\one}{\mathbf{1}}$
$\newcommand{\loc}{\Delta\hspace{0cm}}$
$\newcommand{\mat}{\delta\hspace{0cm}}$
$\newcommand{\det}{\mathrm{\det}\hspace{0cm}}$
$\newcommand{\cm}[1]{\bar{\bb{#1}}}$
$\newcommand{\lmsum}{\sum_{n=0}^\infty\sum_{m=-n}^n}$
$\newcommand{\kqsum}{\sum_{k=0}^\infty\sum_{q=-k}^k}$
$\newcommand{\dif}[1]{#1\hspace{0cm}\prime}$
$\newcommand{\lap}{\nabla^2\hspace{0cm}}$
$\newcommand{\dermat}[1]{\frac{\mathrm{D}\hspace{0.02cm}#1}{\mathrm{D}\hspace{0.02cm} t}}$
$\newcommand{\dercla}[1]{\frac{\mathrm{d}\hspace{0.02cm}#1}{\mathrm{d}\hspace{0.02cm} t}}$
$\newcommand{\derpar}[1]{\frac{\partial\hspace{0.02cm}#1}{\partial\hspace{0.02cm} t}}$
$\newcommand{\rpar}[2]{\frac{\partial\hspace{0.02cm}#1}{\partial\hspace{0.02cm} r_{#2}}}$
$\newcommand{\xpar}[2]{\frac{\partial\hspace{0.02cm}#1}{\partial\hspace{0.02cm} x_{#2}}}$
$\newcommand{\ddermat}[1]{\frac{\mathrm{D}^2\hspace{0.02cm}#1}{\mathrm{D}\hspace{0.02cm} t^2}}$
$\newcommand{\ddercla}[1]{\frac{\mathrm{d}^2\hspace{0.02cm}#1}{\mathrm{d}\hspace{0.02cm} t^2}}$
$\newcommand{\dderpar}[1]{\frac{\partial^2\hspace{0.02cm}#1}{\partial\hspace{0.02cm} t^2}}$
$\newcommand{\drpar}[2]{\frac{\partial^2\hspace{0.02cm}#1}{\partial\hspace{0.02cm} r_{#2}^2}}$
$\newcommand{\dxpar}[2]{\frac{\partial^2\hspace{0.02cm}#1}{\partial\hspace{0.02cm} x_{#2}^2}}$

## Complex functions

```{div} full-width

Let us considering a complex-valued function $f$ of a complex variable $s$

$$\begin{align}
f:\mathbb{C}\rightarrow\mathbb{C}
\end{align}
$$ (L:14)

By considering the real, $x$, and imaginary, $y$, parts of the complex variable $s=x+i\,y$, we can look at the real, $u$, and imaginary, $v$, parts of the function $f=u+i\,v$ as a function of both $x$ and $y$

$$\begin{align}
f(s) = f(x+i\,y) = u(x,y) + i\,v(x,y)
\end{align}
$$ (L:15)

Let us now define the derivative of $f$ at a point $s_0$ s the incremental ratio

$$\begin{align}
\frac{d f(s_0)}{d s} = \lim_{s\rightarrow s_0}\frac{f(s)-f(s_0)}{s-s_0}
\end{align}
$$ (L:16)

and consider the partial derivatives with respect to $x$ and $y$

$$\begin{align}
\frac{\partial f(x+i\,y)}{\partial x} = \frac{\partial u(x,y)}{\partial x} + i\,\frac{\partial v(x,y)}{\partial y}
\qquad
\frac{\partial f(x+i\,y)}{\partial y} = \frac{\partial u(x,y)}{\partial y} + i\,\frac{\partial v(x,y)}{\partial y}
\end{align}
$$ (L:17)

In view of the fact that 

$$\begin{align}
\frac{\partial f(x+i\,y)}{\partial x} = \frac{1}{i}\,\frac{\partial f(x+i\,y)}{\partial y}
\end{align}
$$ (L:18)

we thus obtain

$$\begin{align}
\frac{\partial u(x,y)}{\partial x} = \frac{\partial v(x,y)}{\partial y}
\qquad
\frac{\partial v(x,y)}{\partial y} = -\frac{\partial u(x,y)}{\partial x}
\end{align}
$$ (L:19)

In this respect, differentiable complex function have specific constraints on how the partial derivative of its real and complex parts behave. 

```

### Analyticity of holomorphic functions

```{div} full-width

In complex analysis, a complex-valued function f of a complex variable s:

- is said to be holomorphic at a point $s_0\in\mathbb{C}$ if it is differentiable at every point within a neighborhood of $s_0$
- is said to be analytic at $s_0$ if in a neighborhood centered at $s_0$ it can be expanded as a convergent power series

$$\begin{align}
f (s) = \sum_{n=0}^\infty f_n ( s âˆ’ s_0 )^n 
\end{align}
$$  (L:20)
    
with $f_n$ being complex coefficients.

One of the most important theorems of complex analysis is that holomorphic functions are analytic and vice versa. From this theorem follows that holomorphic (differentiable) function in an open set $\mathcal{U}\subseteq \mathbb{C}$ are infinitely differentiable and that the coefficients $f_n$ are given by

$$\begin{align}
f_n = \frac{d^n f(s_0)}{d s^n}
\end{align}
$$ (L:21)
```

#### Residue theorem

```{div} full-width

When the holes are just a series of point $s_k\in\mathcal{U}$, the closed contour $\gamma_k$ can be chosen as a small circle $c_k$ of radius $r$ centred at $s_k$ and, according to the residue theorem, one gets

$$\begin{align}
\oint_{\gamma_k} f(s)\,ds = \oint_{c_k} f(s)\,ds = 2\,\pi\,i\,\mathrm{Res}(f,s_k)
\end{align}
$$  (L:23)

with $\mathrm{Res}(f,s_k)$ being the residue of $f$ at $s_k$ defined as 

$$\begin{align}
\mathrm{Res}(f,s_k) = \frac{1}{2\,\pi\,i}\,\oint_{c_k} f(s)\,ds = \frac{1}{2\,\pi}\,\int_{0}^{2\,\pi} f\big(s_k+z(\theta)\big)\,z(\theta)\,d\theta
\end{align}
$$  (L:24)

with $z(\theta) = r\,e^{i\,\theta}$. 

For first-order poles, it can be shown that 

$$\begin{align}
\mathrm{Res}(f,s_k) = \lim_{s\rightarrow s_k} (s-s_k)\,f(s)
\end{align}
$$ (L:5)

while, for general $n$-order poles, with $n\geq0$, we have

$$\begin{align}
\mathrm{Res}(f,s_k) = \frac{1}{(n-1)!}\,\lim_{s\rightarrow s_k} \frac{d^{n-1} (s-s_k)^n\,f(s)}{d s^{n-1}}
\end{align}
$$  (L:25)

```

### Analytic continuation

```{div} full-width

Analytic continuation is a technique to extend the domain of definition of a given analytic (or holomorphic) function. Suppose $f$ is an analytic function defined on a non-empty open subset $\mathcal{U}\subset\mathbb{C}$. If $\mathcal{V}$ is a larger open subset containg the former, $\mathcal{U}\subset\mathbb{V}$ and $F$ is an analytic (or holomorphic) function defined on $\mathcal{V}$ such that

$$\begin{align}
F( s ) = f ( s ) \qquad \forall\,s\in\mathcal{U} 
\end{align}
$$  (L:26)

then $F$ is called an analytic continuation of $f$. In other words, the restriction of $F$ to $\mathcal{U}$ is the function $f$ we started with. The import asdpect of analytic continuation (which follows directly from the identiy theorem for holomorphic function) is that is unique, in the sense that there exists only one holomorphic function $F$ which, once restricted to $\mathcal{U}$, coincides with $f$. 

```

## Laplace transform

```{div} full-width

The  Laplace transform is defined as follows

$$\begin{align}
\tilde{f}(s) = \int_{-\infty}^\infty f(t)\,e^{-s\,t}\,dt
\end{align}
$$ (L:1)

where $s$ is a complex angular function (or Laplace variable)

$$\begin{align}
s = \sigma+i\,\omega
\end{align}
$$  (L:27)

and it is assumed that the function is zero at negative times

$$\begin{align}
f(t) = 0 \qquad \forall t<0
\end{align}
$$ (L:7)

A necessary condition for existence of the integral in eq. {eq}`L:1` is that $f$ must be locally integrable on $[0, \infty)$. Also, the integral can converge only within some subset $\mathcal{U}$ of the complex plane. In this case, however, the complex function, as far as it is differetiable in $\mathcal{U}$ is also holomorphic (or analytic) and, so, by analytic continuation can be extended to the whole complex plane. 

As example, let us consider the Laplace transform of the following function

$$\begin{align}
f(t) = H(t)\,e^{\alpha\,t}
\end{align}
$$  (L:28)

In particular, we obtain that the integral in eq. {eq}`L:1` converges  

$$\begin{align}
\tilde{f}(s) = \int_{0}^\infty e^{(\alpha-s)\,t}\,dt = \left. \frac{e^{(\alpha-s)\,t}}{\alpha-s} \right|_{t=0}^\infty = \frac{1}{s-\alpha}  \qquad \Leftrightarrow\qquad \Re[s] > \alpha
\end{align}
$$  (L:29)

only assuming that $\Re[s] > \alpha$, i.e, in the open subset 

$$\begin{align}
\mathcal{U}=\big\{s\in\mathbb{C}\,\mid \Re[s]>\alpha\big\}
\end{align}
$$  (L:30)

By analytic continuation, however, we can set

$$\begin{align}
\tilde{f}(s) =  \frac{1}{s-\alpha}  \qquad \forall s \in\mathbb{C}
\end{align}
$$  (L:31)


Another import example is the Laplace transform of the Dirac delta which yields

$$\begin{align}
\tilde{\delta}(s) =  \int_{-\infty}^\infty \delta(t)\,e^{-s\,t}\,dt = 1
\end{align}
$$ (L:3)

The definition of Laplace transform given by eq. {eq}`L:1` is the bilateral one, which differs by the definition of unilateral Laplace transform where the time integration is restricted to non-negative times, from $0$ to $\infty$, instead of the whole real axis, from $-\infty$ to $\infty$. Among the bilateral and unilateral definitions, the former has the advantage that it is strictly related to the Fourier transform and allows us to deal with the Dirac delta as well, without making further considerations. We have to keep in mind, however, that we shall apply the bilateral Laplace transform to functions which are zero for negative time, eq. {eq}`L:7`.

```

### Properties of the Laplace transform

```{div} full-width

Hereinafter we list the main properties of the Laplace transform which are closely related to those of the Fourier transform

```

#### Shift in time

```{div} full-width

The Laplace transform of a function shifted in time $f(t+t_0)$ yields the Lalace transform $\tilde{f}(s)$ multiplied by $e^{s\,t_0}$

$$
\begin{align}
\int_{-\infty}^\infty f(t+t_0)\,e^{-s\,t}\,\mathrm{d}t 
= e^{s\,t_0}\,\tilde{f}(s)
\end{align}
$$ (L:8a)

```

#### Derivatives

```{div} full-width

The Laplace transform of a $n$-th order time derivative of a function $f(t)$ yields the Laplace transform $\tilde{f}(\omega)$ multiplied by $s^n$

$$
\begin{align}
\int_{-\infty}^\infty \frac{\mathrm{d}^n f(t)}{\mathrm{d} t^n}\,e^{-s\,t}\,\mathrm{d}t = s^n\,\tilde{f}(\omega)
\end{align}
$$ (L:9)

```

#### Convolution

```{div} full-width

The Laplace transform of the convolution between two functions $f(t)$ and $g(t)$ 

$$
\begin{align}
h(t) = \int_{-\infty}^\infty f(\tau)\,g(t-\tau)\,\mathrm{d}\tau 
\end{align}
$$ (L:10)

yields the product of the two Laplace transforms $\tilde{f}(\omega)$ and $\tilde{g}(\omega)$ 

$$
\begin{align}
\int_{-\infty}^\infty h(t)\,e^{-s\,t}\,\mathrm{d}t = \tilde{f}(s)\,\tilde{g}(s)
\end{align}
$$ (L:11)

```

#### Real functions

```{div} full-width

The Laplace transform of a real-valued function $f(t)$ has the following property

$$
\begin{align}
\tilde{f}^*(s) = \tilde{f}(s^*)
\end{align}
$$ (L:12)

with the asterisk $*$ stands for the complex conjugate, as proved hereinafter

$$
\begin{align}
& \tilde{f}^*(t) = \int_{-\infty}^\infty f(t)\,e^{s^*\,t}\,d t = \tilde{f}(s^*)
\end{align}
$$ (L:13)

By denoting the real and imaginary parts of the Laplace variable $s$ with $x$ and $y$, and the real and imaginary parts of the Laplace transform $\tilde{f}(s)$ with $u$ and $v$

$$
s=x+i\,y \qquad \tilde{f}(s) = u(x,y)+i\,v(x,y)
$$  (L:32)

we can write eq. {eq}`L:12` as follows

$$
\begin{align}
u(x,y) = u(x,-y) \qquad v(x,y) = -v(x,-y)
\end{align}
$$  (L:33)

which shows how the real part $u$ is symmetric with respect to the real axis, while the imaginary part $v$ is anti-symmetric. In this respect, the Laplace transform of a real-valued function is real on the real axis of the complex plane, i.e., for $s\in\mathbb{R}$.

---

```

```{admonition} Jordan lemma
:class: full-width, dropdown

The Jordan lemma states that

$$
\begin{align}
\bigg\|\int_{C_R} f(\omega)\,e^{i\,\omega\,t}\,d \omega \bigg\| \leq \frac{\pi}{\|t\|}\,M_R\qquad M_R = \max_{\omega\in C_R} \big\| f(\omega)\,\big\| 
\end{align}
$$  (L:34)

where the contour path of integration $C_R$ is chosen depending whether the time $t$ is positive or negative

$$
\begin{align}
& C_R = \begin{cases}
\Big\{\omega=R\,e^{i\,\theta}\,\mid\,\forall\,\theta\in[0,\pi] \Big\} 
 & t>0
\\ 
\Big\{\omega=R\,e^{i\,\theta}\,\mid\,\forall\,\theta\in[\pi,2\,\pi] \Big\} 
& t<0 
\end{cases}
\end{align}
$$  (L:35)

To be applied to the inverse Laplace transform, it can be recast in the following form

$$
\begin{align}
\bigg\|\int_{C_R} f(s)\,e^{s\,t}\,d s \bigg\| \leq \frac{\pi}{a}\,M_R\qquad M_R = \max_{\omega\in C_R} \big\| f(s)\,\big\| 
\end{align}
$$  (L:36)

$$
\begin{align}
& C_R = \begin{cases}
\Big\{s=c+R\,e^{i\,\theta}\,\mid\,\forall\,\theta\in\big[\tfrac{1}{2}\pi,\tfrac{3}{2}\pi\big] \Big\} 
 & t>0
\\ 
\Big\{s=c+R\,e^{i\,\theta}\,\mid\,\forall\,\theta\in\big[-\tfrac{1}{2}\pi,\tfrac{1}{2}\pi\big] \Big\} 
& t<0 
\end{cases}
\end{align}
$$  (L:37)

```

### Inverse Laplace transform

```{div} full-width

![start](../images/Bromwich.png)


The inverse Laplace transform is defined as follows

$$
\begin{align}
f(t) = \frac{1}{2\,\pi\,i}\,\lim_{R\rightarrow\infty} \int_{c-i\,R}^{c+i\,R} \tilde{f}(s)\,e^{s\,t}\,d s
\end{align}
$$ (L:4)

where $c$ is a real number so that the contour path of integration is in the region of convergence of $\tilde{f}(s)$. In light of this, we note that the inverse Laplace transform is obtained by integrating a path parallel to the imaginary axis of the complex domain, the so called Bromwich path.

A common case  is when the Laplace transform $\tilde{f}(s)$ converges to a (same) constant value in the limit of $s$ going to infinity (in all the directions of the complex plane)

$$
\begin{align}
\lim_{\|s\|\rightarrow \infty} \tilde{f}(s) = f_E
\end{align}
$$  (L:38)

We will refer to this contant value as the *elastic limit*. In this case, eq. {eq}`L:4` can be written as follows

$$
\begin{align}
f(t) = f_E\,\delta(t) + \frac{1}{2\,\pi\,i}\,\lim_{R\rightarrow\infty} \int_{c-i\,R}^{c+i\,R} \tilde{f}(s)\,e^{s\,t}\,d s
\end{align}
$$  (L:39)

where we make use of the fact that, according to eq. {eq}`L:3`, the inverse Laplace transform of $1$ is the Dirac delta $\delta(t)$. After that,  as depicted in Figure X for positive times, $t>0$, the integral along the Bromwich path is subsituted by an integral along a closed contour $\Gamma_R$ minus the integral along the half circle $C_R$ of radius $R$ used to closed the Bromwich path.

$$
\begin{align}
f(t) = f_E\,\delta(t) + \frac{1}{2\,\pi\,i}\,\lim_{T\rightarrow\infty} \left(\int_{\Gamma_R} \big(\tilde{f}(s)-f_E)\,e^{s\,t}\,d s - \int_{C_R} \big(\tilde{f}(s)-f_E\big)\,e^{s\,t}\,d s\right)
\end{align}
$$  (L:40)

and the two integrals are performed making use of the residue theorem and the Jordan lemma discussed above. In particular, in the limit of $R\rightarrow\infty$ and thanks to eq. {eq}`L:2`, the first contribution becomes the summation over closed contour path enclosing the the regions where $\tilde{f}(s)$ is not analytic, i.e. the holes of the open subset where $\tilde{f}(s)$ is not analytic after analytic continuation. The Jordan lemma, instead, guarantees that the second contribution yields zero because

$$
\begin{align}
\lim_{R\rightarrow\infty} M_R = \lim_{\|s\,\|\rightarrow\infty} \big(\tilde{f}(s)-f_E\big) = 0
\end{align}
$$  (L:41)

$$
\begin{align}
f(t) = f_E\,\delta(t) + \sum_k \int_{\gamma_k} \big(f(s)-f_E\big)\,e^{s\,t}\,ds
\end{align}
$$  (L:42)

Assuming that $\tilde{f}(s)$ has only first-order poles $s_k$ and making use of the Residue theorem and of eq. {eq}`L:5`, we obtain

$$
\begin{align}
f(t) = f_E\,\delta(t) + \sum_k f_k\,e^{s_k\,t}\,H(t)
\end{align}
$$ (L:6)

with $f_k$ being the residues associated to the pole $s_k$

$$
\begin{align}
f_k = \lim_{s\rightarrow s_k} (s-s_k)\,\tilde{f}(s)
\end{align}
$$  (L:43)

```
