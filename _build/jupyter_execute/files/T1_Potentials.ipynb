## Poisson equation

```{div} full-width

Let us consider the gravitational potential $\phi$ due to a continuum body with density distribution $\rho$ at the present time $t$. In classical physics, the link between these two fields is instantaneous and does not depend on the motion of the particles, but just on their current positions. In this respect, the spatial coordinates used for  identify the position in the three-dimensional space can be arbitrarly choosen, even adopting non-inertial reference frame with rotating unit vectors referred to a time-dependent translating origin. We will come back on this issue in section {numref}`Chapter %s <chap-kinematics>` - {ref}`chap-kinematics` and, for now, we will simply assume that the points of the three-dimensional space can be identified by a vector position 

$$ 
\bb{x} = x_i\,\bn{x}_i = x\,\bn{r}(\vartheta,\varphi)
$$ (G:0)

with $x_i$ and $\bn{x}_i$ being the Cartesian coordinates and unit vectors, and $(x,\vartheta,\varphi)$ being the spherical coordinates and $\bn{r}$ being the radial unit vector pointing towards $\bb{x}$.

The definition of the gravitational potential in terms of the density distribution reads

$$
\phi(\bb{x}) = -G\,\int_{\mathbb{R}^3}\frac{\rho(\dif{\bb{x}})}{\|\bb{x}-\dif{\bb{x}}\|}\,\md \dif{V}
$$ (G:1)

where $G$ is the universal gravitational constant. By introducing the gradient and the Laplacian operators

$$
\grad = \bn{x}_i\,\frac{\partial}{\partial x_i} \qquad \lap  = \grad\cdot\grad = \frac{\partial^2}{\partial x_i\,\partial x_i}
$$ (G:2)

and applying the Laplacian operator to eq. {eq}`G:1`, we obtain the Poisson equation

$$
\lap\phi(\bb{x}) = C\,\rho(\bb{x})
$$ (G:3)

where we have made use of the following identity

$$
\lap\frac{1}{\|\bb{x}-\dif{\bb{x}}\|} = - 4\,\pi\,\delta(\bb{x}-\dif{\bb{x}})
$$ (G:4)

and

$$
C=4\,\pi\,G
$$ (G:4a)

```



### Spherical harmonic expansion 

```{div} full-width

Let us now adopt consider the spherical coordinates $(x,\vartheta,\varphi)$ and introduce the following spherical harmonic expansions

$$
\phi(\bb{x}) = \sum_n\sum_m \phi_{n,m}(x)\,Y_n^m(\vartheta,\varphi) \qquad \rho(\bb{x}) = \sum_n\sum_m \rho_{n,m}(x)\,Y_n^m(\vartheta,\varphi) 
$$ (G:5)

By noting that

$$
\lap \big(\phi_{n,m}(x)\,Y_n^m(\vartheta,\varphi)\big) = \big(\nabla_n^2\phi_{n,m}(x)\big)\,Y_n^m(\vartheta,\varphi)
$$ (G:6)

with $\nabla_n^2$ being the degree-$n$ radial Laplacian operator

$$
\nabla_n^2 = \frac{\partial^2}{\partial x^2}+\frac{2}{x}\,\frac{\partial x}{\partial x} - \frac{n\,(n+1)}{x^2}
$$ (G:7)

and considering that the spherical harmonics are orthogonal to each other, eq. {eq}`P:33` the Poisson equation {eq}`G:3` becomes

$$
\nabla_n^2\phi_{n,m}(r) = c\,\rho_{n,m}(x)
$$ (G:8)

```




### Laplace equation

```{div} full-width

We note the Laplace equation (i.e., the Possion equation with the density set to zero) reads

$$
\nabla\phi(\bb{x}) = 0 \qquad \nabla_r^2\phi_{n,m}(x) = 0
$$ (G:9)

and it admits the following two independent solutions

$$
\phi_{n,m}(x) = A^+_{n,m}\,x^n + A^-_{n,m}\,x^{-(n+1)}
$$ (G:10)

```

## Solid Earth (and load) system

```{div} full-width

Let us now assume that the density distribution $\rho$ describes mass of the solid Earth and loads and that differs from zero only with a sphere of radius $a$ containing them

$$
\rho(x>a,\vartheta,\varphi) = 0
$$ (G:11)

In this case, the Poisson equation becomes the Laplace equation outside the sphere of radius $a$

$$
\nabla_x^2\phi_{n,m}(x) = 0 \qquad\mathrm{with}\qquad x > a
$$ (G:12)

In view of eqs. {eq}`G:10`  and requiring the regularity of the solution at the infinity, $r\rightarrow \infty$, we thus obtain

$$
\phi_{n,m}(x) = A^-_{n,m}\,x^{-(n+1)} \qquad\mathrm{with}\qquad x > a
$$ (G:13)

Furthermore, because the gravitational potential is continuous, we can evaluate eq. {eq}`G:13` over the sphere of radius $a$

$$
\phi_{n,m}(a) = A^-_{n,m}\,a^{-(n+1)}
$$ (G:14)
 
and rewrite eq. {eq}`G:13` as follows

$$
\phi_{n,m}(x) = \phi_{n,m}(a)\,\left(\frac{x}{a}\right)^{-(n+1)} \qquad\mathrm{with}\qquad x\geq a
$$ (G:15)

```

### MacCullagh theorem

```{div} full-width

Let us now specify eq. {eq}`G:1` to the present case and perform the volume integral in spherical coordinates

$$
\phi(\bb{x}) = -G\,\int_{0}^{a^+}\Bigg(\int_{\Omega}\frac{\rho(\dif{\bb{x}})}{\|\bb{x}-\dif{\bb{x}}\|}\,\md\dif{\Omega}\Bigg)\,\dif{x}^2\,\md \dif{x}
$$ (G:16)

By considering that the inverse of the distance is the generating function of the Laplacian operator and making use of the spherical harmonic addition theorem

$$
\begin{align}
\frac{1}{\|\bb{x}-\dif{\bb{x}}\|} &= \frac{1}{\sqrt{x^2+\dif{x}^2+\cos\gamma\,x\,\dif{x} }} =  \frac{1}{x}\,\sum_{n=0}^\infty \left(\frac{\dif{x}}{x}\right)^n\,P_n(\cos\gamma) 
\\
&= \frac{1}{x}\,\sum_n\sum_m\left(\frac{\dif{x}}{x}\right)^n\,\frac{Y_n^m(\dif{\vartheta},\dif{\varphi})\,Y_n^m(\vartheta,\varphi)}{2\,n+1}  \qquad\mathrm{with}\qquad x\geq a
\end{align}
$$ (G:17)

with

$$
\cos\gamma = \bn{x}(\vartheta,\varphi)\cdot\bn{r}(\dif{\vartheta},\dif{\varphi}) = \cos\vartheta\,\cos\dif{\vartheta} +  \sin\vartheta\,\sin\dif{\vartheta}\,\cos(\varphi-\dif{\varphi})
$$ (G:18)

eq. {eq}`G:16` gives

$$
\begin{align}
\phi(\bb{x}) &= -\frac{G}{x}\,\sum_n\sum_m \frac{Y_n^m(\vartheta,\varphi)}{2\,n+1}\,\int_{0}^{a^+}\Bigg(\int_{\Omega} \rho(\dif{\bb{x}})\,Y_n^m(\dif{\vartheta},\dif{\varphi})\,\md\Omega\Bigg)\,\left(\frac{\dif{x}}{x}\right)^n\,\dif{x}^2\,\md \dif{x}
\\
&= -\frac{C}{x}\,\sum_n\sum_m \frac{Y_n^m(\vartheta,\varphi)}{2\,n+1}\,\int_{0}^{a^+}\rho_{n,m}(\dif{x})\,\left(\frac{\dif{x}}{x}\right)^n\,\dif{x}^2\,\md \dif{x}    \qquad\mathrm{with}\qquad x\geq a
\end{align}
$$ (G:19)

From the comparison with eq. {eq}`G:5` we thus obtain the so called MacCullagh theorem linking $\phi_{n,m}$ to $\rho_{n,m}$

$$
\begin{align}
\phi_{n,m}(x) &= -\frac{C}{x^{n+1}\,(2\,n+1)}\,\int_{0}^{a^+}\rho_{n,m}(\dif{x})\,\dif{x}^{n+2}\,\md \dif{x}    \qquad\mathrm{with}\qquad x\geq a
\end{align}
$$ (G:21)

```

```{admonition} Surface density distribution
:class: full-width, dropdown

Let us now consider the case in which the density distribution $\rho$ describe a load localized over a sphere of radius $a$

$$
\rho(\bb{x}) = \sigma(\vartheta,\varphi)\,\delta(x-a) \qquad\mathrm{with}\qquad \sigma(\vartheta,\varphi) = \lmsum\sigma_{n,m}\,Y_n^m(\vartheta,\varphi)
$$ (G:22)

where $\sigma$ is the surface density distribution which we can expand in spherical harmonics. This case fits the requirement of the MacCullagh theorem and, so, from eq. {eq}`G:21` we obtain

$$
\begin{align}
\phi_{n,m}(x) &= -\frac{C\,\sigma_{n,m}\,a}{2\,n+1}\,\left(\frac{a}{x}\right)^{n+1}  \qquad\mathrm{with}\qquad x\geq a
\end{align}
$$ (G:23)

Furthermore, by considering that the surface density distribution is zero also within the sphere of radius $a$, the gravitational potential solve the Laplace equation within this sphere and, for the regularity at the Earth centre, $r=0$, we can write

$$
\phi_{n,m}(x) = A^+_{n,m}\,x^n = \phi_{n,m}(a)\,\left(\frac{x}{a}\right)^n = -\frac{C\,\sigma_{n,m}\,a}{2\,n+1}\,\left(\frac{x}{a}\right)^n \qquad\mathrm{with}\qquad x\leq a
$$ (G:25)

In light of eqs. {eq}`G:23` and {eq}`G:25`, we thus can write

$$
\phi_{n,m}(x) = \phi_{n,m}(a)\,\begin{cases} \left(\frac{x}{a}\right)^n & x\leq a \\ 
\left(\frac{x}{a}\right)^{-(n+1)} & x\geq a
\end{cases} 
$$ (G:26)

with

$$
\phi_{n,m}(a) = -\frac{C\,\sigma_{n,m}\,a}{2\,n+1}
$$ (G:24)

```


```{admonition} Mass and centre of mass
:class: full-width, dropdown

Let us now calculate the first moment of the Earth density distribution

$$
M\,\bb{Z} = \int \rho(\bb{x})\,\bb{x}\,\md V \qquad\mathrm{with}\qquad M = \int \rho(\bb{x})\,\md V 
$$ (G:27)

with $M$ and $\bb{Z}$ being the total mass and the centre of mass. After spherical harmonic expansion and noting that 

$$
\bb{x}\,Y_{n,m}=x\,\bb{P}_n^m
$$ (G:27a)

we obtain

$$
\begin{align}
& M = \int \rho(\bb{x})\,\md V = 4\,\pi\,\int \rho_{0,0}(x)\,x^2\,\md x = -\frac{a}{G}\,\phi_{0,0}(a)
\\
& M\,\bb{Z} = \sum_n\sum_m \left(\int\rho_{n,m}(x)\,x^{3}\,\md x\right)\,\left(\int_\Omega \bb{P}_n^m(\vartheta,\varphi)\,\md \Omega\right) = \frac{4\,\pi}{\sqrt{3}}\sum_m \bb{p}_m\,\int\rho_{1,m}(x)\,x^3\,\md x 
\\
&\qquad\qquad = -\frac{\sqrt{3}\,a^2}{G}\sum_m \bb{p}_m\,\phi_{1,m}(a) 
\end{align}
$$ (G:28)

where we made use of the MacCullagh theorem for the degrees $n=0$ and $1$

$$
\begin{align}
& 4\,\pi\,\int_{0}^{a^+}\rho_{0,0}(x)\,x^{2}\,\md x  =- \frac{a}{G}\,\phi_{0,0}(a) 
\qquad
4\,\pi\,\int_{0}^{a^+}\rho_{1,m}(x)\,x^{3}\,\md x  =- \frac{3\,a^{2}}{G}\,\phi_{1,m}(a) 
\end{align}
$$ (G:29)

and of the following identity (from eq. {eq}`V:12`)

$$
\begin{align}
& \int_{\Omega}\mathbf{P}_n^m(\vartheta,\varphi,t)\,\mathrm{d}\Omega = \frac{4\,\pi}{\sqrt{3}}\,\mathbf{p}_m(t)\,\delta_{n,1}
\end{align}
$$ (G:30)

In this respect, the degree-$0$ coefficient $\phi_{0,0}(a)$ describe the total mass of the density distribution, while the degree-$1$ coefficients $\phi_{1,m}(a)$ describe the position of the centre of mass. In the following lectures, we will assume that the origin of the reference system coincides with the centre of mass of the solid Earth and load system and, so, the degree-$1$ spherical harmonic coefficients of the gravitational potential will be constrained to zero, $\phi_{1,m}(a) =0$.

```

```{admonition} Inertia tensor
:class: full-width, dropdown


When we will discuss the dynamics of the rotation of the Earth and load system, we will introduce the inertia tensor $\bb{I}$ defined as follows

$$
\bb{I} = \int \rho(\bb{x})\,\big(x^2\,\one-\bb{x}\,\bb{x}\big)\,\md V
$$ (G:31)

Let us now introduce the following deviatoric tensor and consider its spherical harmonic expansion which only involves degree-$2$ terms

$$
\bb{Q}(\vartheta,\varphi) = \frac{1}{3}\,\one - \bn{r}(\vartheta,\varphi)\,\bn{r}(\vartheta,\varphi) = \sum_m \bb{Q}_m\,Y_2^m(\vartheta,\varphi)
$$ (G:32)

and note that

$$
x^2\,\one-\bb{x}\,\bb{x} = x^2\,\left(\frac{2}{3}\,\one + \bb{Q}(\vartheta,\varphi)\right) 
$$ (G:33)

Here, $\bb{Q}_m$ are the following constant deviatoric tensors


$$
\begin{align}
& \bb{Q}_0 = \frac{\bn{x}_1\,\bn{x}_1+\bn{x}_2\,\bn{x}_2-2\,\bn{x}_3\,\bn{x}_3}{3\,\sqrt{5}}
& \bb{Q}_{\pm1} = -\frac{\bn{x}_3\,\bb{p}_{\pm1} + \bb{p}_{\pm1}\,\bn{x}_3}{\sqrt{15}} \\
& \bb{Q}_{-2} = -\frac{\bn{x}_2\,\bn{x}_2 -\bn{x}_1\,\bn{x}_1}{\sqrt{15}}
& \bb{Q}_{2} =-\frac{ \bn{x}_1\,\bn{x}_2 + \bn{x}_2\,\bn{x}_1}{\sqrt{15}}
\end{align}
$$ (G:34)

or, in matrix notation,

$$
\begin{align}
& \bb{Q}_0 = \frac{1}{3\,\sqrt{5}}\left(\begin{matrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & -2\end{matrix}\right) \\
& \bb{Q}_{-1} = -\frac{1}{\sqrt{15}}\left(\begin{matrix} 0 & 0 & 1 \\ 0 & 0 & 0 \\ 1 & 0 & 0\end{matrix}\right)
\qquad  \bb{Q}_1 = -\frac{1}{\sqrt{15}}\left(\begin{matrix} 0 & 0 & 0 \\ 0 & 0 & 1 \\ 0 & 1 & 0\end{matrix}\right)
\\
&\bb{Q}_{-2} = -\frac{1}{\sqrt{15}}\left(\begin{matrix} 1 & 0 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & 0\end{matrix}\right)
\qquad  \bb{Q}_2 = -\frac{1}{\sqrt{15}}\left(\begin{matrix} 0 & 1 & 0 \\ 1 & 0 & 0 \\ 0 & 0 & 0 \end{matrix}\right)
\end{align}
$$ (G:34a)

Thanks to this decomposition, after spherical harmonic expansion and making use of the MacCullagh theorem for the degree $n=2$, eq. {eq}`G:31` becomes

$$
\begin{align}
\bb{I} &= \sum_n\sum_m\Bigg(\int \rho_{n,m}(x)\,x^4\Bigg)\,\int\Bigg(\frac{2}{3}\,\one + \bb{Q}(\vartheta,\varphi)\Bigg)\,Y_{n,m}(\vartheta,\varphi)\,\md \Omega 
\\
&= I\,\one + 4\,\pi\,\sum_m \Bigg(\int \rho_{2,m}(x)\,x^4\Bigg)\,\bb{Q}_m = I\,\one + \bb{D}
\end{align}
$$ (G:35)

with $I$ and $\bb{D}$ being the scalar inertia moment and deviatoric inertia tensor of the solid Earth and load system

$$
I = \frac{8\,\pi}{3}\,\int \rho_{0,0}(x)\,x^4\,\md x \qquad \bb{D} = - \frac{5\,a^3}{G}\,\sum_m \phi_{2,m}(a)\,\bb{Q}_m
$$ (G:36)

```


```{admonition} Gravity acceleration
:class: full-width, dropdown

The gravitational acceleration $\bb{g}$ corresponds to minus the gradient of the gravitational potential 

$$
\bb{g}(\bb{x}) = -\grad\phi(\bb{x}) 
$$ (G:37)

that, after expansion in spherical harmonics, reads

$$
\begin{align}
\bb{g}(\bb{x}) &= -\sum_n\sum_m\left(\frac{\partial \phi_{n,m}(x)}{\partial x}\,\bn{r}(\vartheta,\varphi)\,Y_n^m(\vartheta,\varphi) + \phi_{n,m}(x)\,\grad Y_n^m(\vartheta,\varphi)\right)
\\
&= -\sum_n\sum_m\left(\frac{\partial \phi_{n,m}(x)}{\partial x}\,\bb{P}_n^m(\vartheta,\varphi) + \frac{\gamma_n}{x}\,\phi_{n,m}(x)\,\bb{B}_n^m(\vartheta,\varphi)\right)
\end{align}
$$ (G:38)

where $\bb{P}_n^m$ and $\bb{B}_n^m$ are the first two spherical harmonic vectors and $\gamma_n=\sqrt{n\,(n+1)}$. For the case of the solid Earth and load system and focusing only outside the sphere of radius $a$, in view of eqs. {eq}`V:14` and {eq}`G:15`, we can write

$$
\begin{align}
\bb{g}(\bb{x}) 
&= \frac{a}{x^2}\,\sum_n\sum_m\left(\frac{a}{x}\right)^{n}\,\phi_{n,m}(a)\,\left((n+1)\,\bb{P}_n^m(\vartheta,\varphi) - \gamma_n\,\bb{B}_n^m(\vartheta,\varphi)\right)
\\
&= -\frac{G\,M}{x^2}\,\bn{r}(\vartheta,\varphi) + \frac{a}{x^2}\,\sum_{n\geq 1}\sum_m\left(\frac{a}{x}\right)^{n}\,\phi_{n,m}(a)\,\left((n+1)\,\bb{P}_n^m(\vartheta,\varphi) - \gamma_n\,\bb{B}_n^m(\vartheta,\varphi)\right)
\end{align}
$$ (G:39)

This result shows how the gravity acceleration for a general density distribution is characterized by both radial and tangential components. In particular, by definition, the degree-$0$ contribution is along the radial direction while those from the higher harmonic degrees $n\geq1$ have also the tangential component. We note also that, within the assumption that the origin of the reference system coincide with the centre of mass, $\phi_{1,m}(a)=0$, the summation of the second term in the right-hand side involves degrees $n\geq 2$.
```

(sec-celestial)=
## Celestial bodies

```{div} full-width

Let us now assume that the density distribution $\rho^T$ describes the celestial body mass distribution and that differs from zero inside the sphere of radius $R$ containing the Earth and load system, $a<R$ 

$$
\rho^T(x\leq R,\vartheta,\varphi) = 0 
$$ (G:40)

In this case, the gravitational potential $\phi^T$ solves the Laplace equation inside the sphere of radius $R$ and, so, following similar arguments of the previous sections, we can write

$$
\phi^T_{n,m}(x) = \phi^T_{n,m}(a)\,\left(\frac{x}{a}\right)^n  \qquad \mathrm{with}\qquad x < R
$$ (G:41)

```

```{admonition} Point-like celestial bodies
:class: full-width, dropdown

Let us approximate the celestial body to a point-like mass $M_T$ centred at the position $\bb{x}_T$ which corresponds to its  (time-dependent) centre of mass

$$
\rho^T(\bb{x}) = M_T\,\delta(\bb{x}-\bb{x}_T)  \qquad\mathrm{with}\qquad \bb{x}_T = x_T\,\bn{r}(\vartheta_T,\varphi_T)
$$ (G:42)

and define its gravitational potential

$$
\phi^T(\bb{x}) = -G\,\int\frac{\rho^T(\dif{\bb{x}})}{\|\bb{x}-\dif{\bb{x}}\|}\,\md \dif{V} = -\frac{G\,M_T}{\left\|\bb{x}-\bb{x}_T\right\|}
$$ (G:43)

By considering that the inverse of the distance is the generating function of the Legendre polynomials and making use of the spherical harmonic addition theorem, we obtain

$$
\frac{1}{\left\|\bb{x}-\bb{x}_T\right\|} = \frac{1}{x_T}\,\sum_n\left(\frac{x}{x_T}\right)^n\,P_n(\cos\gamma) = \frac{1}{x_T}\,\sum_n\sum_m \left(\frac{x}{x_T}\right)^n\,\frac{Y_n^m(\vartheta_T,\varphi_T)\,Y_n^m(\vartheta,\varphi)}{2\,n+1}
$$ (G:44)

and, so, the spherical harmonic expansion of gravitation potential of the celestial bodies reads

$$
\phi^T(\bb{x}) = \sum_n\sum_m \phi_{n,m}^T(x)\,Y_n^m(\vartheta,\varphi)
\qquad
\mathrm{with}
\qquad \phi^T_{n,m}(x) = -\frac{G\,M_T}{x_T}\,\left(\frac{x}{x_T}\right)^n\,\frac{Y_n^m(\vartheta_T,\varphi_T)}{2\,n+1}
$$ (G:45)

Thanks to this result (which holds also in the case of a spherically syemmtric celestial body, without requiring that it is a point-like mass) we can understand the gravitational potential of the celestial body acting on the Earth, i.e., for $x \leq a<R$, decrease as $\left(x/x_T\right)^n$ and, so, it becomes negligible already starting from the first few harmonic degrees, considering that $x/x_T\ll 1$.

```


```{admonition} General celestial bodies (optional)
:class: full-width, dropdown

Let us express the density distribution $\rho^T$ of the celestial body with respect to a (time-dependent) origin $\bb{x}_T$. In order to fix the ideas, we can consider the origin $\bb{x}=\zero$ of the spherical coordinate system as the centre of mass of the solid Earth, while $\bb{x}_T$ is the time-dependent centre of mass of the celestial body relative to it. Within this framework, we denote with a breve accent the relative observation point $\breve{\bb{x}}$ with respect to $\bb{x}_T$ 

$$
\bb{x} = \bb{x}_T+\breve{\bb{x}}  \qquad\mathrm{with}\qquad \breve{\bb{x}} = \breve{x}\,\bn{r}(\breve{\vartheta},\breve{\varphi})
$$ (G:46) 

and we consider the following spherical harmonic expansion of the density distribution and of the associated gravitational potential 

$$
\begin{align}
& \rho^T(\bb{x}) = \breve{\rho}^T(\breve{\bb{x}}) = \sum_n\sum_m \breve{\rho}_{n,m}^T(\breve{x})\,Y_n^m(\breve{\vartheta},\breve{\varphi})
\\
& \phi^T(\bb{x}) = \breve{\phi}^T(\breve{\bb{x}}) = \sum_n\sum_m \breve{\phi}_{n,m}^T(\breve{x})\,Y_n^m(\breve{\vartheta},\breve{\varphi})
\end{align}
$$ (G:47)

Here and hereinafter, we will use the breve accent to denote quantities with respect to $\bb{x}_T$, including the spherical coordinate $(\breve{x},\breve{\vartheta},\breve{\varphi})$. Within this framework, we can write

$$
\bb{x}_T = x_T\,\bn{r}(\breve{\vartheta}_T,\breve{\varphi}_T) \qquad \mathrm{with} \qquad  \breve{\vartheta}_T=-\vartheta_T\qquad \breve{\varphi}_T =-\varphi_T
$$ (G:48)

By requiring that the total mass $M_T$ conserves through time and that the centre of mass of the celestial body coincides with $\bb{x}_T$

$$
M_T = \int \rho^T(\bb{x})\,\md V 
\qquad
M_T\,\bb{x}_T = \int \rho^T(\bb{x})\,\bb{x}\,\md V 
$$ (G:49)

we thus obtain the following constraints on the degree-0 and -1 spherical harmonic coefficients of the density distribution

$$
4\,\pi\,\int \breve{\rho}_{0,0}(\breve{x},t)\,\breve{x}^2\,\md \breve{x}= M_T
\qquad
\int \breve{\rho}_{1,m}(\breve{x},t)\,\breve{x}^3\,\md \breve{x} = 0
$$ (G:51)



Let us now consider the gravitational potential due to the celestial body and expand it in spherical harmonics

$$
\begin{align}
\phi^T(\bb{x},t) =  \breve{\phi}^T(\breve{\bb{x}},t) &= - G\,\int\frac{\rho^T(\dif{\bb{x}},t)}{\|\bb{x}-\dif{\bb{x}}\|}\,\md \dif{V} = - G\,\int\frac{\breve{\rho}^T(\dif{\breve{\bb{x}}},t)}{\|\breve{\bb{x}}-\dif{\breve{\bb{x}}}\|}\,\md \dif{\breve{V}} 
\\
& = - \sum_n\sum_m \frac{G\,Y_{n,m}(\breve{\vartheta},\breve{\varphi})}{(2\,n+1)\,\breve{x}^{n+1}}\int Y_{n,m}(\dif{\breve{\vartheta}},\dif{\breve{\varphi}})\,\breve{\rho}^T(\dif{\breve{\bb{x}}},t)\,\dif{\breve{x}}^n\,\md \dif{\breve{V}}
\\
&= - \sum_n\sum_m \frac{4\,\pi\,G\,Y_{n,m}(\breve{\vartheta},\breve{\varphi})}{(2\,n+1)\,\breve{x}^{n+1}}\int \breve{\rho}_{n,m}^T(\dif{\breve{x}},t)\,\dif{\breve{x}}^{n+2}\,\md\dif{\breve{x}}
\end{align}
$$ (G:52)

Within the assumption that $\breve{x}>\breve{a}_T$, with $\breve{a}_T$ being the radius of the sphere centred in $\bb{x}_T$ and containing the celestial body, we can write

$$
\breve{\phi}_{n,m}^T(\breve{x},t) = \breve{\phi}_{n,m}(\breve{x}_T,t)\,\left(\frac{\breve{x}}{\breve{a}_T}\right)^{-(n+1)}
$$ (G:53)

Then, making use of the MacCullach theorem

$$
4\,\pi\,G\,\int \breve{\rho}_{n,m}^T(\dif{\breve{x}},t)\,\dif{\breve{x}}^{n+2}\,\md \dif{\breve{x}} = -(2\,n+1)\,\breve{a}_T^{n+1}\,\breve{\phi}_{n,m}^T(\breve{a}_T,t)
$$ (G:54)

eq. {eq}`G:52` becomes

$$
\begin{align}
\phi^T(\bb{x},t) 
&= \breve{\phi}_{0,0}(\breve{a}_T,t)\,\left(\frac{\breve{a}_T}{\breve{r}}\right) + \sum_{n\geq 2}\sum_m Y_{n,m}(\breve{\vartheta},\breve{\varphi})\,\breve{\phi}_{n,m}(\breve{a}_T,t)\,\left(\frac{\breve{a}_T}{\breve{r}}\right)^{n+1}
\end{align}
$$ (G:55a)

where the degree-1 term does not contribute in view of eq. {eq}`G:51`. This result show that the departure from the spherical symmetry are a minor correction to the gravitational potential of the celestial body because $\breve{\phi}_{n,m}(\breve{a}_T,t)\ll \breve{\phi}_{0,0}(\breve{x},t)$ and the dependence on $(\breve{a}_T/\breve{x})^{n+1}$, with $\breve{a}_T\ll \breve{x}$. In this respect, we can neglect the contributions from harmonic degrees $n\geq 2$. Also, by considering that

$$
\breve{\phi}_{0,0}(\breve{a}_T) = -\frac{G\,M_T}{\breve{a}_T}
$$ (G:55b)

eq. {eq}`G:55b` becomes

$$
\breve{\phi}_{0,0}(\breve{x}) = -\frac{G\,M_T}{\breve{x}}
$$ (G:55c)

which coincides with the result already discussed for the point-like celestal body, eq. {eq}`G:43`, because $\breve{x} = \left\|\bb{x}-\bb{x}_T\right\|$.


```


```{div} full-width

For later convenience, let us decompose this gravitational potential into the degree-$0$ and -$1$ contributions and the so-called tidal potential $\psi^T$ as follows

$$
\phi^T(\bb{x}) = \phi^T_0(\bb{x}) + \phi^T_1(\bb{x}) + \psi^T(\bb{x})
$$ (G:56)

where, according to eq. {eq}`G:41`, $\phi_0^T$, $\phi_1^T$ and $\psi^T$ are given by

$$
\begin{align} &
\phi^T_0(\bb{x}) = \phi^T_{0,0}(a)
\qquad
\phi^T_1(\bb{x}) = \sum_m \frac{x}{a}\,\phi^T_{1,m}(a)\,Y_1^m(\vartheta,\varphi) 
\\
& \psi^T(\bb{x}) = \sum_{n\geq 2}\sum_m \left(\frac{x}{a}\right)^n\,\psi^T_{n,m}(a)\,Y_n^m(\vartheta,\varphi)
 \qquad\mathrm{with}\qquad x\leq R
 \end{align}
$$ (G:57)

According to this decomposition, the spherical harmonic coefficients of the tidal potential are given by

$$
\psi^T_{n,m}(x) = \begin{cases} 0 & n=0,1 \\ \phi^T_{n,m}(x) & n\geq 2 \end{cases}
$$ (G:58)

The gravity acceleration due to celestial body inside the sphere of radius $R$ yields

$$
\bb{g}^T(\bb{x})= -\grad\phi^T(\bb{x}) = \bb{g}^T_1 + \bb{f}^T(\bb{x})  \qquad\mathrm{with}\qquad x\leq R
$$ (G:59)

Here, $\bb{f}^T$ is the tidal acceleration

$$
\bb{f}^T(\bb{x}) = -\grad\psi^T(\bb{x}) = -\sum_{n\geq 2}\sum_m \frac{\psi^T_{n,m}(a)}{a}\,\left(\frac{x}{a}\right)^{n-1}\,\bb{V}_n^m(\vartheta,\varphi)
$$ (G:60)

and $\bb{g}_1^T$ is gravity acceleration due to the degree-$1$ contribution

$$
\begin{align}
\bb{g}_1^T = -\grad\phi^T_1(\bb{x}) = -\sum_m \frac{\phi^T_{1,m}(a)}{a}\,\bb{V}_1^m(\vartheta,\varphi) = -\sqrt{3}\,\sum_m \frac{\phi^T_{1,m}(a)}{a}\,\bb{p}_m
\end{align}
$$ (G:61)

Here, $\bb{V}_n^m$ is given by

$$
\bb{V}_n^m(\vartheta,\varphi) = n\,\bb{P}_{n,m}(\vartheta,\varphi)+\gamma_n\,\bb{B}_{n,m}(\vartheta,\varphi)
$$

and we have made use of the following identity

$$
\bb{V}_1^m(\vartheta,\varphi) = \bb{P}_1^m(\vartheta,\varphi) + \gamma_1\,\bb{B}_1^m(\vartheta,\varphi) = \sqrt{3}\,\bb{p}_m
$$ (G:62)

It is notewothy that $\bb{g}_1^T$ is just a constant (in space) vector which does not depend on the spatial coordinates. This decomposition of the gravitational potential of the celestial body into the degree-$0$ and -$1$ contributions and the tidal potential is usefull when we will consider the forces acting on the Earth system. Indeed, as we will show, the tidal potential $\psi^T$ is responsible for deformations of the Earth and, so, they can be only of degree $n\geq 2$. Different, the degree-1 contribution $\phi_1^T$ is responsible for the orbital motion of the Earth around the celestial body (or viceversa).

We note that, within the (reasonable) approximation of a point-like (or spherically symmetric) celestial body, the constant (in space) vector $\bb{g}_1^T$ becomes

$$
\bb{g}_1^T  = \frac{G\,M_T}{x_T^2}\,\frac{1}{\sqrt{3}}\,\sum_m  Y_1^m(\vartheta_T,\varphi_T)\,\bb{p}_m  = \frac{G\,M_T}{x_T^2}\,\bn{r}(\vartheta_T,\varphi_T)
$$ (G:63)

```

(sec-total-force)=
### Total force
```{div} full-width

Let us now evaluate the total force acting on the Earth due to the celestial body

$$
\bb{F}^T = -\int \rho(\bb{x})\,\grad\phi^T(\bb{x})\,\md V = M\,\bb{g}_1^T + \int \rho(\bb{x})\,\bb{f}^T(\bb{x})\,\md V
$$ (G:64)

where it is understood that $\rho$ is the density distribution of the Earth system and $M$ is its total mass. After spherical harmonic expansion and making use of the decomposition given by eq. {eq}`G:60`, eq. {eq}`G:64` becomes

$$
\begin{align}
\bb{F}^T &= M\,\bb{g}_1^T -\sum_{n\geq 2}\sum_m \sum_k\sum_q \frac{\psi^T_{n,m}(a)}{a^n}\,\Bigg(\int \rho_{k,q}(x)\,x^{n+1}\,\md x \Bigg)
\Bigg(\int_\Omega \bb{V}_{n,m}(\vartheta,\varphi)\,Y_{k,q}(\vartheta,\varphi)\,\md \Omega\Bigg)
\\
&= M\,\bb{g}_1^T + \sum_{n\geq 2}\sum_m \sum_q \frac{\psi^T_{n,m}(a)\,\phi_{n-1,q}(a)}{G}\,\bb{p}^n_{m,q}
\end{align}
$$ (G:65)

where we made use of the MacCullagh theorem for the degree $n-1$ 

$$
4\,\pi\,\int \rho_{n-1,m}(x)\,x^{n+1}\,\md x = - \frac{(2\,n-1)\,a^{n}\,\phi_{n-1,m}(a)}{G}
$$ (G:66)

and of the following identity

$$
\int_\Omega \bb{V}_{n,m}(\vartheta,\varphi)\,Y_{k,q}(\vartheta,\varphi)\,\md \Omega = \frac{4\,\pi}{2\,n-1}\,\bb{p}^n_{m,q}\,\delta_{n,k+1}
$$ (G:67)

with $\bb{p}^{n}_{m,q}$ being constant vectors.

We note that, once fixed the origin of the reference system with the centre of mass of the Earth system, we have that $\phi_{1,m}(a)=0$ and, so, the summation over the harmonic degrees $n\geq 2$ in the right-hand side of eq. {eq}`G:65` will start actually from $n\geq 3$. Moreover, within the (reasonable) approximation of a point-like (or spherically symmetric) celestial body, we can also write


$$
\begin{align}
\bb{F}^T &= \frac{G\,M\,M_T}{x_T^2}\,\bn{r}(\vartheta_T,\varphi_T) + \sum_{n\geq 3}\sum_m \sum_q (2\,n-1)\,\frac{\psi^T_{n,m}(a)\,\phi_{n-1,q}(a)}{G}\,\bb{p}^n_{m,q}
\end{align}
$$ (G:68)

The first term in the right-hand side is a simple statement of the gravitational force between two (spherically symmetric) massive objects at the distance $x_T$. The second term, instead, show how departures from the spherical symmetry of the Earth starting from the degree $n\geq 2$ can be responsible for some corrections to the first term. On the other hand, when these departures and the tidal potential are considered infinitesimal first-order terms, these corrections can be neglected and, so, correct to the first-order, we can write

$$
\begin{align}
\bb{F}^T &\approx M\,\bb{g}_1^T = \frac{G\,M\,M_T}{x_T^2}\,\bn{r}(\vartheta_T,\varphi_T) 
\end{align}
$$ (G:69)


```


(sec-total-torque)=
### Total torque

```{div} full-width

Let us now evaluate the total force acting on the Earth due to the celestial body

$$
\begin{align}
\bb{K}^T &= -\int \rho(\bb{x})\,\bb{r}\times\grad\phi^T(\bb{x})\,\md V = \Bigg(\int \rho(\bb{x})\,\bb{r}\,\md V\Bigg)\times\bb{g}_1^T + \int \rho(\bb{x})\,\bb{r}\times\bb{f}^T(\bb{x})\,\md V \\
&= M\,\bb{Z}\times\bb{g}_1^T + \int \rho(\bb{x})\,\bb{r}\times\bb{f}^T(\bb{x})\,\md V
\end{align}
$$ (G:70)

After spherical harmonic expansion and making use of the decomposition given by eq. {eq}`G:60`, eq. {eq}`G:70` becomes

$$
\begin{align}
\bb{K}^T &= M\,\bb{Z}\times\bb{g}_1^T -\sum_{n\geq 2}\sum_m \sum_k\sum_q \frac{\psi^T_{n,m}(a)}{a^n}\,\Bigg(\int \rho_{k,q}(x)\,x^{n+2}\,\md x \Bigg)\Bigg(\int_\Omega\bn{r}(\vartheta,\varphi)\times\bb{V}_{n,m}(\vartheta,\varphi)\,Y_{k,q}(\vartheta,\varphi)\,\md \Omega\Bigg)
\\
&= M\,\bb{Z}\times\bb{g}_1^T  + \sum_{n\geq 2}\sum_m \sum_k\sum_q \frac{\psi^T_{n,m}(a)}{a^n}\,\Bigg(\int \rho_{k,q}(x)\,x^{n+2}\,\md x \Bigg)\,\Bigg(\gamma_n\,\int_\Omega \bb{C}_{n,m}(\vartheta,\varphi)\,Y_{k,q}(\vartheta,\varphi)\,\md \Omega\Bigg)
\\
&= M\,\bb{Z}\times\bb{g}_1^T  - \sum_{n\geq 2}\sum_m \sum_k\sum_q a\,\frac{\psi^T_{n,m}(a)\,\phi_{n,q}(a)}{G}\,\bb{q}_{m,q}^n
\end{align}
$$ (G:70a)

where we made use of the MacCullagh theorem for the degree $n$ 

$$
4\,\pi\,\int \rho_{n,m}(x)\,x^{n+2}\,\md x = - \frac{(2\,n+1)\,a^{n+1}\,\phi_{n,m}(a)}{G}
$$ (G:71)

and of the following identities

$$
\begin{align}
\bn{r}(\vartheta,\varphi)\times\bb{V}_{n,m}(\vartheta,\varphi) = -\gamma_n\,\bb{C}_{n,m}(\vartheta,\varphi)
\qquad
\gamma_n\,\int_\Omega \bb{C}_{n,m}(\vartheta,\varphi)\,Y_{k,q}(\vartheta,\varphi)\,\md \Omega = \frac{4\,\pi}{(2\,n+1)}\,\bb{q}^n_{m,q}\,\delta_{n,k}
\end{align}
$$ (G:72)

with $\bb{q}^{n}_{m,q}$ being constant vectors.

We note that, once fixed the origin of the reference system with the centre of mass of the Earth, we have that $\bb{Z}=\zero$ and $\phi_{1,m}(a)=0$ and, so, the first term in the right-hand side of eq. {eq}`G:70a` is zero. The second term, instead, show how departures from the spherical symmetry of the Earth starting from the degree $n\geq 2$ can be responsible for some corrections to the first term. On the other hand, when these departures and the tidal potential are considered infinitesimal first-order terms, these corrections can be neglected and, so, correct to the first-order, we can write

$$
\begin{align}
\bb{K}^T &\approx \zero
\end{align}
$$ (G:73)

In this respect, within a first-order approximation theory, the total torque acting on the Earth is zero and there is no exchange of angular momentum between the Earth and the celestial bodies. In order to describe such a process, we shall at least develop a second-order perturbation theory where the degree-2 potential of the Earth related to its equatorial bulge and its deviatoric inertia tensor is taken into account.
```
