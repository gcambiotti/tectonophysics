```{admonition} Proof of 
:class: note, full-width

Let us consider the integration by parts of two function $f(x)$ and $g(x)$

$$\begin{align}
\int_a^b f(x)\,\dot{g}(x)\,\mathrm{d}x = \left. f(x)\,g(x)\,\right|_a^b - \int_a^b \dot{f}(x)\,g(x)\,\mathrm{d}x
\end{align}$$ (A0:6)

and apply it to the case in which $g(x)=H(x)$ and $a<0<b$

$$\begin{align}
\int_a^b f(x)\,\delta(x)\,\mathrm{d}x &= \left.f(x)\,H(x)\,\right|_a^b - \int_a^b \dot{f}(x)\,H(x)\,\mathrm{d}x \\
&= f(b) - \int_0^b \dot{f}(x)\,\mathrm{d}x = f(b) - \left. f(x)\,\right|_0^b = f(0)
\end{align}$$ (A0:7)



## Dirac delta derivative

```{div} full-width

The Dirac delta derivative is defined as follows

$$\begin{align}
\int_a^b f(x)\,\delta'(x)\,\mathrm{d}x &= \left. f(x)\,\delta(x)\,\right|_a^b - \int_a^b \dot{f}(x)\,\delta(x)\,\mathrm{d}x =  -\dot{f}(0)
\end{align}$$ (A0:8)

For higher-order derivatives one obtians

$$\begin{align}
\int_a^b f(x)\,\delta^{(n)}(x)\,\mathrm{d}x &= (-1)^n\,f^{(n)}(0)
\end{align}$$ (A0:9)

```

## Basic propetries 

```{div} full-width

The Dirac delta is symmetric

$$\begin{align}
\delta(-x) = \delta(x)
\end{align}$$ (A0:10)

and the product of the Dirac delta with $x$ is equal to zero

$$\begin{align}
x\,\delta(x) = 0
\end{align}$$ (A0:11)

The Dirac delta derivative satisfies a number of basic properties

$$\begin{align}
& \dot{\delta}(-x) = -\dot{\delta}(x) \\
& x\,\dot{\delta}(x) = -\delta(x)
\end{align}$$ (A0:12)

```

## Three-dimensional Dirac delta

The three-dimensional Dirac delta can be defined in the Cartesian reference frame as 

$$
\begin{align}
\delta(\mathbf{x}) = \delta(x_1)\,\delta(x_2)\,\delta(x_3)
\end{align}$$ (A0:13)

where $\mathbf{x}$ is the vector position and $x_j$ are its Cartesian coordinates

$$
\begin{align}
\mathbf{x} = x_j\,\hat{\mathbf{x}}_j
\end{align}$$ (A0:14)

with $\hat{\mathbf{x}}_j$ being the Cartesian unit vectors.

Similar to the one-dimensional case, eq. {eq}`A0:3`, it satify the following identity

$$\begin{align}
\int_{\mathcal{V}} f(\mathbf{x})\,\delta(\mathbf{x})\,\mathrm{d}V = 
\begin{cases} f(\mathbf{0}) & \mathbf{0}\in\mathcal{V} \\ 
0 & \mathbf{0}\notin\mathcal{V}
\end{cases}
\end{align}$$ (A0:15)

for any (scalar, vector or tensor) field $f$ that is continuous and has continuous first order partial derivatives. Here, $\mathcal{V}\subseteq\mathbb{R}^3$ is a open region of the three-dimensional space and $dV=dx_1\,dx_2\,dx_3$ is the infinitesimal volume element.


### The Laplacian of the inverse of the radial distance

```{tip} 
:class: full-width

To understand the content of this section, you should have been studied the Prerequisits

```

Let us consider the radial distance from the origin


$$
\begin{align}
r = \sqrt{\mathbf{x}\cdot\mathbf{x}}
\end{align}$$ (A0:16)

It can be shown that the Laplacian of the inverse of the radial distance is proportional to the three-dimensional Dirac delta. In particular, we can write

$$
\begin{align}
\nabla^2 a(r) = \delta(\mathbf{x})    
\end{align}$$ (A0:17)

with $a$ being the following function proportional to the inverse of the radial distance

$$
\begin{align}
a(r) = -\frac{1}{4\,\pi\,r}
\end{align}$$ (A0:18)



First of all, let us consider that

$$
\begin{align}
\nabla^2 a(r)= 0
\end{align}$$

everywhere but at the origin $\mathbf{x}=\mathbf{0}$, where the function $a$ is singular. In this respect, as far as the region $\mathcal{V}$ does not contain the origin, the following integral yields always zero

$$
\int_\mathcal{V} f(\mathbf{x})\,\nabla^2 a(r)\,d V = 0 \qquad \mathbf{0}\notin\mathcal{V}
$$

as expected from eq. {eq}`A0:15`.

In order to better understand the behaviour  at the origin, let us now consider the following integral over an open volume $\mathcal{V}$ containing the origin, $\mathbf{0}\in\mathcal{V}$,

$$
\begin{align}
\int_{\mathcal{V}} \nabla^2 a(r)\,d V &= -\frac{1}{4\,\pi}\,\int_{\mathcal{B}}\nabla^2\frac{1}{r}\,d V  
= -\frac{1}{4\,\pi}\,\int_{\partial\mathcal{B}}\hat{\mathbf{r}}\cdot\boldsymbol{\nabla}\frac{1}{r}\,d S \\
&= \frac{1}{4\,\pi\,R^2}\,\int_{\partial \mathcal{B}} d S = 1 &
\end{align}
$$

Here, $\mathcal{B}\subset \mathcal{V}$ is a sphere of radius $R$ centered at the origin, $\partial \mathcal{B}$ is its boundary of area $4\,\pi\,R^2$ and we made use of the divergence theorem to relate the volume and surface integrals. As expected setting $f(\mathbf{x})=1$ in eq. {eq}`A0:15`, the integral yields $1$.

In order to prove eq. {eq}`A0:17`, we have still to verify it for the case of $\mathbf{0}\in\mathcal{V}$ and for any (scalar, vector or tensor) field $f$. By making use of the following general identity

$$
\begin{align}
\boldsymbol{\nabla}\cdot\big((\boldsymbol{\nabla}a(\mathbf{x}))\,f(\mathbf{x})\big) &= (\nabla^2 a(\mathbf{x}))\,f(\mathbf{x}) + \big(\boldsymbol{\nabla} a(\mathbf{x})\big)\cdot\big(\boldsymbol{\nabla} f(\mathbf{x}) \big) &
\end{align}$$

for $a$ given by eq. {eq}`A0:18`, we can write

$$
\begin{align}
(\nabla^2 a(r))\,f(\mathbf{x}) &=  \frac{1}{4\,\pi}\,\left[\boldsymbol{\nabla}\cdot\left(\frac{\hat{\mathbf{r}}}{r^2}\,f(\mathbf{x})\right) -  \frac{1}{r^2}\,\frac{\partial f(\mathbf{x})}{\partial r}\right] &
\end{align}$$

Then, by considering the desired integral and proceed as follows

$$
\begin{align}
\int_\mathcal{V} f(\mathbf{x})\,\nabla^2 a(r)\,d V &= \int_\mathcal{B} f(\mathbf{x})\,\nabla^2 a(r)\,d V  = \frac{1}{4\,\pi}\,\left(\int_{\partial\mathcal{B}} \frac{f(\mathbf{x})}{r^2}\,d S - \int_{\mathcal{B}} \frac{1}{r^2}\,\frac{\partial f(\mathbf{x})}{\partial r}\,d V \right)&
\\
& = \frac{1}{4\,\pi}\,\left(\int_{\Omega}\,f(\mathbf{x})|_{r=R}\,d \Omega - \int_{\Omega}\int_0^R \frac{\partial f(\mathbf{x})}{\partial r}\,d r\,d \Omega \right)
\\
&= \frac{1}{4\,\pi}\,\left(\int_{\Omega}\,f(\mathbf{x})|_{r=R}\,d \Omega - \int_{\Omega}\Big(f(\mathbf{x})|_{r=R}-f(\mathbf{x})|_{r=0}\Big)\,d \Omega \right)
\\
& = \frac{1}{4\,\pi}\,\int_{\Omega}f(\mathbf{0})\,d \Omega = f(\mathbf{0})  \qquad \mathbf{0}\in\mathcal{V}
\end{align}
$$

we can see that, as expected, it yields the field evaluated at the origin when $\mathbf{0}\in\mathcal{V}$.


<p style="page-break-after:always;"></p>
