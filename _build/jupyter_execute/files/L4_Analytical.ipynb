

$\newcommand{\md}{\mathrm{d} \hspace{0cm}}$
$\newcommand{\mdd}{\mathrm{d}\hspace{0cm}^2}$
$\newcommand{\grad}{\sb{\nabla}\hspace{-0.01cm}}$
$\newcommand{\stress}{\sb{\sigma}}$
$\newcommand{\bn}[1]{\hat{\bb{#1}}}$
$\newcommand{\bb}[1]{\mathbf{#1}}$
$\newcommand{\lap}{\nabla^2\hspace{0cm}}$
$\newcommand{\nlap}{\nabla_n^2\hspace{0cm}}$
$\newcommand{\sn}[1]{\hat{\sb{#1}}}$
$\newcommand{\sb}[1]{\boldsymbol{#1}}$
$\newcommand{\zero}{\bb{0}}$
$\newcommand{\one}{\bb{1}}$
$\newcommand{\loc}{\Delta \hspace{0cm}}$
$\newcommand{\mat}{\delta \hspace{0cm}}$
$\newcommand{\det}{\mathrm{\det}\hspace{0cm}}$
$\newcommand{\cm}[1]{\bar{\bb{#1}}}$
$\newcommand{\lmsum}{\sum_{n=0}^\infty\sum_{m=-n}^n}$
$\newcommand{\kqsum}{\sum_{k=0}^\infty\sum_{q=-k}^k}$
$\newcommand{\dif}[1]{#1\hspace{0cm}\prime}$
$\newcommand{\dermat}[1]{\frac{\mathrm{D}\hspace{0.02cm}#1}{\mathrm{D}\hspace{0.02cm} t}}$
$\newcommand{\dercla}[1]{\frac{\md \hspace{0.02cm}#1}{\md \hspace{0.02cm} t}}$
$\newcommand{\derpar}[1]{\frac{\partial\hspace{0.02cm}#1}{\partial\hspace{0.02cm} t}}$
$\newcommand{\rpar}[2]{\frac{\partial\hspace{0.02cm}#1}{\partial\hspace{0.02cm} r_{#2}}}$
$\newcommand{\xpar}[2]{\frac{\partial\hspace{0.02cm}#1}{\partial\hspace{0.02cm} x_{#2}}}$
$\newcommand{\ddermat}[1]{\frac{\mathrm{D}^2\hspace{0.02cm}#1}{\mathrm{D}\hspace{0.02cm} t^2}}$
$\newcommand{\ddercla}[1]{\frac{\md ^2\hspace{0.02cm}#1}{\md \hspace{0.02cm} t^2}}$
$\newcommand{\dderpar}[1]{\frac{\partial^2\hspace{0.02cm}#1}{\partial\hspace{0.02cm} t^2}}$
$\newcommand{\drpar}[2]{\frac{\partial^2\hspace{0.02cm}#1}{\partial\hspace{0.02cm} r_{#2}^2}}$
$\newcommand{\dxpar}[2]{\frac{\partial^2\hspace{0.02cm}#1}{\partial\hspace{0.02cm} x_{#2}^2}}$


## Layered Earth models

```{div} full-width

In the perspective of obtaining the propagator matrix, let us consider the homogeneous incremental Poisson and momentum equations (i.e., we set to zero the forcing terms $\rho_{n,m}^I$ and $f_{n,m}^X$, with $X=P,\,B$ and $C$) within the assumption that the Lam√® parameters are constant within each layer of the Earth, i.e,

$$
\dot{\mu} = \dot{\lambda} = 0
$$ (A:1)

from one interface to another. In this case, the incremental Poisson and momentum equations, eq. {eq}`M:25`, simplify as follows

$$
\begin{align}
& \nlap \loc \Psi_{n,m} + C\,\big(\varrho_0\,\chi_{n,m}+\dot{\varrho}_0\,U_{n,m}\big) = 0
\\
& \beta\,\dot{\chi}_{n,m} +\frac{\mu\,\gamma_n\,H_{n,m}}{x}
 -\varrho_0\,\frac{\partial \big(g_0\,U_{n,m}+ \loc \Psi_{n,m}\big)}{\partial x} +\varrho_0\,\chi_{n,m}\,g_0= 0
\\
& \beta\,\gamma_n\,\chi_{n,m} + \mu\,\big(H_{n,m}+x\,\dot{H}_{n,m}\big) -\gamma_n\,\varrho_0\,\big(g_0\,U_{n,m}+\loc \Psi_{n,m}\big)  = 0
\\
& \nlap  W_{n,m} = 0
\end{align}
$$ (A:2)

For the sake of brevity, we have multiplied by $r$ the $B$-component of the momentum equation and divided by $\mu$ the $C$-component.

```

### Toroidal analytical solutions

```{div} full-width

We note that, in this case, the analytical solution of the toroidal component of the displacement reads

$$
W_{n,m} = C_+\,x^n + C_-\,x^{-(n+1)} 
$$ (A:3)

and, according to eq. {eq}`M:20`, the incremental traction gives

$$
T_{n,m} = C_+\,(n-1)\,x^{n-1}  - C_{-}\,(n+2)\,x^{-(n+2)}
$$ (A:4)

We thus can write the homogeneous toroidal vector solution as follows

$$
\bb{y}_{n,m}^h(x) = \bb{Y}_n(x)\,\bb{C}
$$ (A:5)

where $\bb{C}$ is the $2$-vector collecting the integration constants and $\bb{Y}_n$ is the so-called toroidal fundamental $2\times 2$-matrix  

$$
\bb{C} = \left(\begin{matrix} C_+ \\ C_- \end{matrix}\right) 
\qquad 
\bb{Y}_{n,m}(x) = \left(\begin{matrix} x^n & x^{-(n+1)} \\ \mu\,(n-1)\,x^{n-1} & -\mu\,(n+2)\,x^{-(n+2)}\end{matrix}\right)
$$ (A:6)

Once determined the fundamental matrix, the propagator matrix can be obtained as follows

$$
\sb{\Pi}_n(x,x_0) = \bb{Y}_n(x)\,\bb{Y}_n^{-1}(x_0)
$$ (A:7)

Some times, it can be convinient to recast the fundamental matrix in the following form

$$
\bb{Y}_{n,m}(x) = \bar{\bb{Y}}_n(x)\,\bb{D}_n(x)
$$ (A:8a)

with

$$
\bar{\bb{Y}}_n(x) = \left(\begin{matrix} 1 & 1 \\ \mu\,\frac{n-1}{x}\, & -\mu\,\frac{n+2}{x} \end{matrix}\right)\qquad\mathrm{and}\qquad \bb{D}_n(x) = \mathrm{Diag}\left(\begin{matrix} x^n & x^{-(n+1)} \end{matrix}\right)
$$ (A:8)

and cast the propagator matrix as follows

$$
\sb{\Pi}_n(x,x_0) = \bar{\bb{Y}}_n(x)\,\bb{D}(x/x_0)\,\bar{\bb{Y}}_n^{-1}(x_0)
$$ (A:7a)

```

### Spheroidal analytical solutions

```{div} full-width

So far, the spheroidal analytical solution for layered Earth models have been obtained only in a few cases: incompressible layared Earth model and the homogeneous compressible Earth model and the homogeneous self-compressed compressible Earth model. These special cases are discussed in the next sections.

```

## Incompressible layered Earth models

```{div} full-width

Let us now assume material incompressibility. In this case the bulk modulus $\kappa$ goes to infinity and the relative volume variations $\chi$ goes to zero in such a way that their product converge to a finite value which can be seen as minus the material incremental pressure

$$
\kappa\,\chi\rightarrow -\mat  p
$$ (A:9)


In this respect, we also have

$$
\begin{align}
& \chi_{n,m} \rightarrow 0 
\qquad
&\kappa\,\chi_{n,m}\rightarrow -\mat  p_{n,m}
\qquad 
&\lambda\,\chi_{n,m}\rightarrow -\mat  p_{n,m}
\end{align}
$$ (A:10)

where $p_{n,m}$ are the spherical harmonic coefficients of the material incremental pressure

$$
\mat  p = \sum_{n=0}^\infty\sum_{m=-n}^n \mat  p_{n,m}\,Y_n^m
$$ (A:11)

In view of eq. {eq}`A:10`, we can also write

$$
\begin{align}
& V_{n,m} = \frac{r\,\dot{U}_{n,m}+2\,U_{n,m}}{\gamma_n}
\qquad
& H_{n,m} = \frac{1}{\gamma_n}\,\nlap (x\,U_{n,m})
\qquad
& R_{n,m} = -\mat  p_{n,m} + 2\,\mu\,\dot{U}_{n,m}
\end{align}
$$ (A:12)


Let us also assume that the shear modulus $\mu$ and the initial density $\varrho_0$ are constant within each layer of the Earth so that we can write

$$
\dot{\mu} = 0 \qquad \dot{\varrho}_0 = 0
$$ (A:13)

but at the interface between the different layers of the Earth, where they can be discontinuous.

Within these assumptions, the homogeneous Poisson equation and the spheroidal components of the momentum equation become

$$
\begin{align}
& \nlap \loc \Psi_{n,m}  = 0 
\\
&\frac{\partial \big(p_{n,m} + \varrho_0\,g_0\,U_{n,m} +\varrho_0\,\loc \Psi_{n,m}\big)}{\partial x} = \frac{\mu}{x}\,\gamma_n\,H_{n,m}  
\\
& \gamma_n\,\left(\mat  p_{n,m}+\varrho_0\,g_0\,U_{n,m}+\varrho_0\,\loc \Psi_{n,m}\right) = \mu\,\left(H_{n,m}+x\,\dot{H}_{n,m}\right)  
\end{align}
$$ (A:14)


```

### Spheroidal analytical solutions

```{div} full-width

By defining 

$$
\begin{align}
\Gamma_{n,m} = \mat  p_{n,m}+\varrho_0\,g_0\,U_{n,m}+\varrho_0\,\loc \Psi_{n,m}
\end{align}
$$ (A:15)

eq. {eq}`A:14` can be recast as follows

$$
\begin{align}
& \nlap \loc \Psi_{n,m}  = 0
\qquad
&\dot{\Gamma}_{n,m} = \frac{\mu}{r}\,\gamma_n\,H_{n,m} 
\qquad
&\gamma_n\,\Gamma_{n,m} = \mu\,\left(H_{n,m}+x\,\dot{H}_{n,m}\right)
\end{align}
$$ (A:16)


Let us now apply the operator $-\partial/\partial x - 2/x$ to the $P$ component of the momentum equation 


$$
\begin{align}
\ddot{\Gamma}_{n,m} +\frac{2}{x}\,\dot{\Gamma}_{n,m} = \frac{\mu}{x^2}\,\gamma_n\,\big(H_{n,m}+x\,\dot{H}_{n,m}\big) 
\end{align}
$$ (A:17)

and add the $B$-component of the momentum equation multiplied by $\gamma_n/x^2$

$$
\begin{align}
\frac{\gamma_n^2}{x^2}\,\Gamma_{n,m} = \frac{\mu}{x^2}\,\gamma_n\,\left(H_{n,m}+x\,\dot{H}_{n,m}\right) 
\end{align}
$$ (A:18)

in order to obtain the following differential equation 

$$
\begin{align}
\nlap  \Gamma_{n,m} = 0
\end{align}
$$ (A:19)

In this respect and making use of eq. {eq}`A:12` for substituting $H_{n,m}$, rather than solving eq. {eq}`A:16`, we can solve the following one


$$
\begin{align}
& \nlap \loc \Psi_{n,m}  = 0
\qquad
&\nlap \Gamma_{n,m} = 0 
\qquad
&\nlap  (x\,U_{n,m}) =  \frac{x}{\mu}\,\dot{\Gamma}_{n,m} 
\end{align}
$$ (A:20)

For $\loc  \Psi_{n,m}$ and $\Gamma_{n,m}$ we thus can write

$$
\begin{align}
\loc \Psi_{n,m} = C_+\,x^n + C_-\,x^{-(n+1)}
\qquad
\Gamma_{n,m} = D_+\,x^n +D_-\,x^{-(n+1)}
\end{align}
$$ (A:21)

and the third differential equation of the differential system given by eq. {eq}`A:20` becomes

$$
\begin{align}
\nlap (x\,U_{n,m}) = \frac{x}{\mu}\,\dot{\Gamma}_{n,m} = D_+\,\frac{n}{\mu}\,x^{n} - D_-\,\frac{n+1}{\mu}\,x^{-(n+1)}
\end{align}
$$ (A:22)

In this respect, the solution for $U_{n,m}$ must be composed by the two indepepent solution of the homogeneous differential equation (i.e., eq. {eq}`A:22` with $D_\pm=0$), and of two particular solutions. We thus write

$$
\begin{align}
U_{n,m} =  A_+\,x^{n+1}+ A_-\,x^{-n}  + B_+\,x^{n-1}+B_-\,x^{-(n+2)}
\end{align}
$$ (A:23)

with

$$
\begin{align}
D_+ = \mu\,\frac{2\,(2\,n+3)}{n}\,A_+
\qquad
D_- = \mu\,\frac{2\,(2\,n-1)}{n+1}\,A_-
\end{align}
$$ (A:24)

At this stage, we dispose of all the quantities that we need for obtaining all the components of the spheroidal vector solution. For instance, from eq.  {eq}`A:15`, we obtain the material incremental pressure

$$
\begin{align}
 \mat  p_{n,m} &= \Gamma_{n,m} - \varrho_0\,g_0\,U_{n,m} - \varrho_0\,\loc \Psi_{n,m}
 =  \left(\frac{D_+}{x} - A_+\,\varrho_0\,g_0\right)\,x^{n+1} +  \left(\frac{D_-}{x} - A_-\,\varrho_0\,g_0\right)\,x^{-n}
 \\
 & -B_+\,\varrho_0\,g_0\,x^{n-1} -B_-\,\varrho_0\,g_0\,x^{-(n+2)} - C_+\,\varrho_0\,x^n - C_-\,\varrho_0\,x^{-(n+1)}
\end{align}
$$ (A:26)

Similar to the toroidal case discussed in {numref}`Section %s <sec-A-toro>` {ref}`sec-A-toro` and after some straightfoward algebra, we can write the homogeneous spheroidal vector solution as follows

$$
\bb{y}_{n,m}^h(x,t) = \bb{Y}_{n}(x)\,\bb{C}(t)
$$ (A:27)

where $\bb{Y}_n$ is the spheroidal fundamental $6\times 6$-matrix 

$$
\bb{Y}_n(x) = \bar{\bb{Y}}_n(x)\,\bb{D}_n(x)
$$ (A:28)

with

$$
\begin{align}
\bar{\bb{Y}}_n(x) = \left(\begin{matrix}
1 & 1 & 0 & 1 & 1 & 0 \\
\frac{n+3}{\gamma} & \frac{n+1}{\gamma} & 0 & \frac{2-n}{\gamma} & -\frac{n}{\gamma} & 0 \\
0 & 0 & 1 & 0 & 0 & 1 \\
\varrho_0\,g_0 + \frac{2\,(n\,(n-1)-3)\,\mu}{n\,x} & \varrho_0\,g_0 + \frac{2\,(n-1)\,\mu}{x}   & \varrho_0 & \varrho_0\,g_0-\frac{2\,(n\,(n+3)-1)\,\mu}{(n+1)\,x} & \varrho_0\,g_0-\frac{2\,(n+2)\,\mu}{x}   & \varrho_0 \\
\frac{2\,n\,(n+2)\,\mu}{x\,\gamma} & \frac{2\,(n^2-1)\,\mu}{x\,\gamma} & 0 & \frac{2\,(n^2-1),\mu}{x\,\gamma} & \frac{2\,n\,(n+2)\,\mu}{x\,\gamma} & 0 \\
C\,\varrho_0 & C\,\varrho_0 & \frac{2\,n+1}{x} & C\,\varrho_0 & C\,\varrho_0 & 0 
\end{matrix}\right)
\end{align}
$$ (A:29)

and

$$
\begin{align}
& \bb{C}(t) = \left(\begin{matrix} A_+(t) & B_+(t) & C_+(t) & A_-(t) & B_-(t) & C_-(t) \end{matrix}\right)^\mathrm{T}
\\
& \bb{D}_n(x) = \mathrm{Diag}\big(\begin{matrix} x^{n+1} & x^{n-1} & x^n & x^{-n} & x^{-(n+2)} & x^{-(n+1)}\end{matrix}\big)
\end{align}
$$ (A:30)

```

### Homogeneous incompressible Earth model

```{div} full-width

Let us now consider an homogeneous incompressible model extending from the Earth centre to the Earth surface, without the presence of a inviscid fluid core. In this case, the initial gravity reads

$$
g_0(r) = \frac{C\,\varrho_0\,x}{3}
$$ (A:31)

and we have to consider the only regual solutions at the Earth centre. In this respect, the homogeneous spheroidal vector solution can be written as follows

$$
\bb{y}_{n,m}^h(x,t) = \bar{\bb{Y}}_{n}^+(x)\,\bb{D}_n^+(x)\,\bb{C}_+(t)
$$ (A:32)

with $\bar{\bb{Y}}_{n}^+$ being the $6\times 3$-matrix consisting of the first three column of the matrix $\bar{\bb{Y}}_n$

$$
\begin{align}
\bar{\bb{Y}}^+_n(x) = \left(\begin{matrix}
1 & 1 & 0 \\
\frac{n+3}{\gamma} & \frac{n+1}{\gamma} & 0  \\
0 & 0 & 1   \\
\varrho_0\,g_0 + \frac{2\,(n\,(n-1)-3)\,\mu}{n\,x}& \varrho_0\,g_0  + \frac{2\,(n-1)\,\mu}{x}   & \varrho_0  \\
\frac{2\,n\,(n+2)\,\mu}{x\,\gamma} & \frac{2\,(n^2-1)\,\mu}{x\,\gamma} & 0 \\
C\,\varrho_0 & C\,\varrho_0 & \frac{2\,n+1}{x} 
\end{matrix}\right)
\end{align}
$$ (A:33)

with

$$
\begin{align}
& \bb{C}_+(t) = \left(\begin{matrix} A_+(t) & B_+(t) & C_+(t) \end{matrix}\right)^\mathrm{T}
\\
& \bb{D}_n^+(r) = \mathrm{Diag}\big(\begin{matrix} r^{n+1} & r^{n-1} & r^n \end{matrix}\big)
\end{align}
$$ (A:34)

By considering only surface loads and tidal and centrifugal forcings, we can establish the only three constants of integration $\bb{C}_+$ 

$$
\bb{C}_+(t) = \big[\bb{P}_2\,\bar{\bb{Y}}_n^+(a)\,\bb{D}_n^+(a)\big]^{-1}\,\bb{b}_{n,m}(t)
$$ (A:35)

and obtain the spheroidal vector soltion 

$$
\bb{y}_{n,m}(r,t) = \bar{\bb{Y}}_{n}^+(r)\,\bb{D}_n^+\big(r/a\big)\,\big[\bb{P}_2\,\bar{\bb{Y}}_n^+(a)\big]^{-1}\,\bb{b}_{n,m}(t)
$$ (A:36)

After some straightforward algebra and defining the following parameter

$$
\begin{align}
\alpha_n = 1 + \frac{(2\,n^2+4\,n+3)\,\mu}{n\,a\,\varrho_0\,g_0(a)}
\end{align}
$$ (A:37)

we can write the radial and tangential spheroidal displacement and the total potential pertubation at the Earth surface as follows


$$
\begin{align}
\left(\begin{matrix} 
U_{n,m}(a,t)
\\
V_{n,m} (a,t)
\\
\loc G_{n,m}(a,t)
\end{matrix}\right) 
&=\bb{k}_n^{L,E}(a)\,\loc G_{n,m}^S(a,t) + \bb{k}_n^{T,E}(a)\,\big(\loc G_{n,m}^{T}(a,t) +\loc G_{n,m}^{C}(a,t) \big)
\end{align}
$$ (A:38)


where, according to eq. {eq}`MM:8sol`, the elastic radial, $h_n^{X,E}$, tangential, $l_n^{X,E}$, and gravitational, $k_n^{X,E}$, components of the Love numbers are given by 

$$
\begin{align}
& \bb{k}_n^{L,E}(a) = \left(\begin{matrix} h_n^{L,E}(a) \\ l_n^{L,E}(a) \\ 1+k_n^{L,E}(a)\end{matrix}\right) = \left(\begin{matrix}
0 \\ 0 \\ 1
\end{matrix}
\right) + \mat\bb{k}_n^L(a)
\\
& \bb{k}_n^{T,E}(a) = \left(\begin{matrix} h_n^{T,E}(a) \\ l_n^{T,E}(a) \\ 1+k_n^{T,E}(a)\end{matrix}\right) = \left(\begin{matrix}0 \\ 0 \\ 1
\end{matrix}
\right) + \mat\bb{k}_n^T(a)
\end{align}
$$ (A:39)

with

$$
\begin{align}
& \mat\bb{k}_n^L(a) = -\frac{1}{\alpha_n}\,\left(\begin{matrix}
\frac{2\,n+1}{3} \\ \frac{1+n}{\gamma_n} \\  1
\end{matrix}\right)
\qquad \mat\bb{k}_n^T(a) = \frac{1}{\alpha_n\,(n-1)}\,\left(\begin{matrix}
\frac{2\,n+1}{2} \\ \frac{3\,(n+1)}{2\,\gamma_n} \\ \frac{3}{2} 
\end{matrix}\right)
\end{align}
$$ (A:39a)

```

### Viscoelastic response

```{div} full-width

According to the Corresponde Principle discussed in  {numref}`Section %s <sec-linear-viscoelastic>` {ref}`sec-linear-viscoelastic` the viscoelastic response in the Laplace domain reads

$$
\begin{align}
& \left(\begin{matrix} \tilde{U}_{n,m}(a,s) 
\\
\tilde{V}_{n,m}(a,s)
\\
 \loc \tilde{G}_{n,m}(a,s)\end{matrix}\right) = \tilde{\bb{k}}_n^L(a,s)\,\,\loc\tilde{G}_{n,m}^S(a,s) +   \tilde{\bb{k}}_n^T(a,s)\,\big(\loc\tilde{G}_{n,m}^T(a,s)+\loc\tilde{G}_{n,m}^C(a,s)\big)
\end{align}
$$ (A:40)

where $\tilde{\bb{k}}_n^X$ are given by eqs. {eq}`A:39` and {eq}`A:39a` with the substitution $\alpha_n\rightarrow \hat{\alpha}_n$, with

$$
\hat{\alpha}_n(s) = 1 + \frac{(2\,n^2+4\,n+3)\,\hat{\mu}(s)}{n\,a\,\varrho_0\,g_0(a)} = \frac{\alpha_n\,s+\frac{1}{\tau_m}}{s+\frac{1}{\tau_m}}
$$ (A:41)

In particular, we note that

$$
\begin{align}
& \tilde{\bb{k}}_n^X(a,s) = \bb{k}_n^{X,E}(a) + \tilde{f}_n(s)\,\mat\bb{k}_n^X(a)
\\
& \bb{k}_n^X(a,t) = \delta(t)\,\bb{k}_n^{X,E}(a) + f_n(t)\,\mat\bb{k}_n^X(a)
\end{align}
$$ (A:42)

with

$$
\begin{align}
& \tilde{f}_n(s) = \frac{\alpha_n}{\tilde{\alpha}_n(s)}-1 = \frac{s+\frac{1}{\tau}}{s+\frac{1}{\tau_n}}-1 = \frac{1}{\tau_m}\,\frac{\alpha_n-1}{\alpha_n\,s+\frac{1}{\tau_m}}
\\
& f_n(t) = \frac{1}{\tau_m}\,\left(1-\frac{1}{\alpha_n}\right)\,e^{-\frac{t}{\alpha_n\,\tau_m}}\,H(t)
\end{align}
$$ (A:43)

In this respect, we note that $\alpha_n\,\tau_m$ is the caracteristic relaxation time of the viscoelastic relaxation of the homogeneous incompressible Earth model.

Then, making use of eq. {eq}`A:42` into eq. {eq}`A:40`, we obtain

$$
\begin{align}
\left(\begin{matrix} 
U_{n,m}(a,t)
\\
V_{n,m} (a,t)
\\
\loc G_{n,m}(a,t)
\end{matrix}\right) 
&=\bb{k}_n^{L,E}(a)\,\loc G_{n,m}^S(a,t) + \bb{k}_n^{T,E}(a)\,\big(\loc G_{n,m}^{T}(a,t) +\loc G_{n,m}^{C}(a,t) \big)
\\
&+ \mat\bb{k}_n^{L}\,\int_{-\infty}^\infty f(\tau)\,\loc  G_{n,m}^S(a,t-\tau)\,\md \tau +  \mat\bb{k}_n^{T}\,\int_{-\infty}^\infty f(\tau)\,\big(\loc  G_{n,m}^{C}(a,t-\tau)+\loc  G_{n,m}^{T}(a,t-\tau)\big)\,\md \tau
\end{align}
$$ (A:44)

```

#### Step-like time history

```{div} full-width

In the simple case in which the time history of the forcings is step-like

$$
\loc  G^X(a,t) = \loc  \bar{G}^X(a)\,H(t)
$$ (A:45)

and considering that the time convolution between $f$ and $H$ reads

$$
F(t) = \int_{-\infty}^\infty f(\tau)\,H(t-\tau)\,\md  t = \bigg(\alpha_n-1\bigg)\,\bigg(1-e^{-\frac{t}{\alpha_n\,\tau}}\bigg)\,H(t)
$$ (A:46)

eq. {eq}`A:44` becomes

$$
\begin{align}
\left(\begin{matrix} 
U_{n,m}(a,t)
\\
V_{n,m} (a,t)
\\
\loc G_{n,m}(a,t)
\end{matrix}\right) 
&=\big(\bb{k}_n^{L,E}(a)\,H(t)+\mat\bb{k}_n^{L}\,F(t)\big)\,\loc \bar{G}_{n,m}^S(a) + \big(\bb{k}_n^{T,E}(a)\,H(t)+\mat\bb{k}_n^{T}\,F(t)\big)\,\big(\loc \bar{G}_{n,m}^{T}(a) +\loc \bar{G}_{n,m}^{C}(a) \big)
\end{align}
$$ (A:47)

By considering the following elastic ($t\rightarrow 0^+$) and fluid ($t\rightarrow \infty$) limits

$$
\lim_{t\rightarrow 0^+} F(t) = 0 \qquad \lim_{t\rightarrow \infty} F(t) = \alpha_n-1
$$ (A:48)

we can verify that the elastic and fluid responses coincide with our previous results. Indeed, in the elastic limit, eq. {eq}`A:47` coincides with eq. {eq}`A:38`, while in the fluid limit, we obtain the same results discussed in {numref}`Section %s <sec-fluid-earth>` {ref}`sec-fluid-earth`, eq. {eq}`MM:5flua`. In particular, the radial displacement and total potential perturbation in the fluid limit yields

$$
\begin{align}
& \lim_{t\rightarrow\infty} U_{n,m}(a,t) = -\frac{2\,n+1}{3}\,\loc  G_{n,m}^S(a)+  \frac{2\,n+1}{2\,(n-1)}\,\big(\loc  G_{n,m}^{C}(a)+\loc  G_{n,m}^{T}(a)\big)
\\
&  \lim_{t\rightarrow\infty} \loc G_{n,m}(a,t) =  \frac{2\,n+1}{2\,(n-1)}\,\big(\loc \bar{G}_{n,m}^{C}(a)+\loc \bar{G}_{n,m}^{T}(a)\big)
\end{align}
$$ (A:49)
```
