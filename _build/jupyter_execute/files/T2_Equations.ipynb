
$\newcommand{\md}{\mathrm{d}\hspace{0cm}}$
$\newcommand{\mdd}{\mathrm{d}\hspace{0cm}^2}$
$\newcommand{\grad}{\boldsymbol{\nabla}\hspace{-0.01cm}}$
$\newcommand{\stress}{\boldsymbol{\sigma}}$
$\newcommand{\bn}[1]{\hat{\mathbf{#1}}}$
$\newcommand{\bb}[1]{\mathbf{#1}}$
$\newcommand{\sn}[1]{\hat{\boldsymbol{#1}}}$
$\newcommand{\sb}[1]{\boldsymbol{#1}}$
$\newcommand{\zero}{\mathbf{0}}$
$\newcommand{\one}{\mathbf{1}}$
$\newcommand{\loc}{\Delta\hspace{0cm}}$
$\newcommand{\mat}{\delta\hspace{0cm}}$
$\newcommand{\det}{\mathrm{\det}\hspace{0cm}}$
$\newcommand{\cm}[1]{\bar{\bb{#1}}}$
$\newcommand{\lmsum}{\sum_{n=0}^\infty\sum_{m=-n}^n}$
$\newcommand{\kqsum}{\sum_{k=0}^\infty\sum_{q=-k}^k}$
$\newcommand{\dif}[1]{#1\hspace{0cm}\prime}$
$\newcommand{\dermat}[1]{\frac{\mathrm{D}\hspace{0.02cm}#1}{\mathrm{D}\hspace{0.02cm} t}}$
$\newcommand{\dercla}[1]{\frac{\mathrm{d}\hspace{0.02cm}#1}{\mathrm{d}\hspace{0.02cm} t}}$
$\newcommand{\derpar}[1]{\frac{\partial\hspace{0.02cm}#1}{\partial\hspace{0.02cm} t}}$
$\newcommand{\rpar}[2]{\frac{\partial\hspace{0.02cm}#1}{\partial\hspace{0.02cm} r_{#2}}}$
$\newcommand{\xpar}[2]{\frac{\partial\hspace{0.02cm}#1}{\partial\hspace{0.02cm} x_{#2}}}$
$\newcommand{\ddermat}[1]{\frac{\mathrm{D}^2\hspace{0.02cm}#1}{\mathrm{D}\hspace{0.02cm} t^2}}$
$\newcommand{\ddercla}[1]{\frac{\mathrm{d}^2\hspace{0.02cm}#1}{\mathrm{d}\hspace{0.02cm} t^2}}$
$\newcommand{\dderpar}[1]{\frac{\partial^2\hspace{0.02cm}#1}{\partial\hspace{0.02cm} t^2}}$
$\newcommand{\drpar}[2]{\frac{\partial^2\hspace{0.02cm}#1}{\partial\hspace{0.02cm} r_{#2}^2}}$
$\newcommand{\dxpar}[2]{\frac{\partial^2\hspace{0.02cm}#1}{\partial\hspace{0.02cm} x_{#2}^2}}$

```{div} full-width

In this lecture, we will discuss the solid Earth and adopt the translating and rotating reference system discussed in {numref}`Chapter %s <chap-kinematics>` - {ref}`chap-kinematics` and further specified in {numref}`Chapter %s <chap-reference>` - {ref}`chap-reference`. In this respect, we will adopt a spherical reference frame centred at Earth's centre of mass $\bb{Z}$ and defined with respect to the rotating unit vectors $\bb{e}_i$

$$
\bb{x} = x_i\,\bn{e}_i(t) = x\,\bn{r}(\vartheta,\varphi,t)
$$ (M:1)

with

$$
\bn{r}(\vartheta,\varphi,t) = \sin\vartheta\,\big(\cos\varphi\,\bn{e}_1(t)+\sin\varphi\,\bn{e}_2(t)\big) + \sin\vartheta\,\bn{e}_3(t)
$$ (M:2)

The spherical unit vectors along the meridians, $\sn{\vartheta}$, and parallels, $\sn{\varphi}$, will be also defined with respect to the rotating unit vectors, as well as the spherical harmonic vectors $\bb{P}_n^m$, $\bb{B}_n^m$ and $\bb{C}_n^m$.

Due to the instantaneous relatation between the density distribution and the gravitational potential, all the results discussed in {numref}`Chapter %s <chap-potentials>` - {ref}`chap-potentials` can be written by (i) adding the time depencence to these two fields, (ii) replacing the Cartesian unit vectors with the rotating ones and (iii) taking into account the translating origin of the reference system. In particular, the unit vectors $\bb{p}_m$ in eq. {eq}`V:13` now  depend on time and read

$$
\bb{p}_0(t) = \bb{e}_3(t) \qquad \bb{p}_{-1}(t) = \bb{e}_1(t) \qquad \bb{p}_1 = \bb{e}_2(t)
$$ (M:2a)

as well as the tensor $\bb{Q}_m$ given by eq. {eq}`G:34`.

Furthermore, will denote with the dot $\cdot$ the derivative with respect to the radial distance 

$$
\dof{f}(x,t) = \frac{\partial f(x,t)}{\partial t}
$$ (M:2b)


```

## Initial and incremental fields

### Density and gravitational potential

```{div} full-width

In the following, as discussed in  {numref}`Section %s <sec-incremental-dec>` - {ref}`sec-incremental-dec`, it will be usefull to introduce the local and material decomposition of the (Eulerian and Lagrangian) density distribution of the Earth

$$
\bar{\rho}^E(\bb{r},t) = \varrho_0(\bb{r}) + \loc\bar{\rho}^E(\bb{r},t)
\qquad
\rho^E(\bb{x},t) = \varrho_0(\bb{x}) + \mat\rho^E(\bb{x},t)
$$ (M:3)

Here and afterwards, the superscript (or subscript) $E$ will be used for denoting fields related to the only Earth system, but for the initial Earth density that will be denote with $\varrho_0$

Then, from eqs. {eq}`E:5` and {eq}`E:8`, we obtain the following expression for the material increment

$$
\mat\rho^E(\bb{x},t) = \varrho_0(\bb{x})\,\left(\frac{1}{J(\bb{x},t)}-1\right) \approx - \varrho_0(\bb{x})\,\grad\cdot\bb{u}(\bb{x},t)
$$ (M:4)

and, from eqs. {eq}`O:27`  and {eq}`O:28`, the following relation between the local and material increments

$$
\loc\bar{\rho}^E(\bb{r},t) = \loc\rho^E(\bb{x},t) =  \varrho_0(\bb{x}) -  \varrho_0(\bb{r}) + \mat \rho^E(\bb{x},t) \approx -\grad\cdot(\varrho_0(\bb{x})\,\bb{u}(\bb{x},t)\big)
$$ (M:5)

As it concern the solid Earth gravitational potential, its decomposition into its initial and local increment read

$$
\bar{\phi}^E(\bb{r},t) = \phi_0^E(\bb{r}) + \loc\bar{\phi}^E(\bb{r},t)
$$ (M:6)

```

### Cauchy stress tensor

```{div} full-width

In addition to the decomposition of the Earth's density and gravitational potentials and according to the discussion in  {numref}`Section %s <sec-constitutive>` - {ref}`sec-constitutive`, we also consider a similar decomposition for the Cauchy stress tensor

$$
\bar{\sb{\sigma}}(\bb{r},t) = \sb{\sigma}_0(\bb{r}) + \loc\bar{\sb{\sigma}}(\bb{r},t) 
$$ (M:7)

with

$$
\loc\bar{\sb{\sigma}}(\bb{r},t) = \loc\sb{\sigma}(\bb{x},t) \approx  -\bb{u}(\bb{x},t)\cdot\grad\sb{\sigma}_0(\bb{x}) + \mat\sb{\sigma}(\bb{x},t)
$$ (M:8)

We recall that both elastic and viscoelastic constitutive equations, eqs. {eq}`O:60` and {eq}`O:62`, define the material (rather than the local) incremental stress and, so, we shall make use of eq. {eq}`M:8` into eq. {eq}`M:7`.

Furthermore, within the assumption that the initial state is in hydrostatic equilibrium, the initial stress takes the following form

$$
\sb{\sigma}_0(\bb{r}) = -p_0(\bb{r})\,\one \qquad \sb{\sigma}_0(\bb{x}) = -p_0(\bb{x})\,\one
$$ (M:10)

with $p_0$ being the initial hydrostatic pressure.

```

### Spherical symmetry

```{div} full-width

As first approximation, we will assume that the initial state of the Earth is spherically symmetric. This means that

$$
\varrho_0(\bb{x}) = \varrho_0(x) \qquad \phi_0^E(\bb{x}) = \phi^E_0(x) \qquad \qquad p_0(\bb{x}) = p_0(x)
$$ (M:11)

and that the local incremental Cauchy stress given by eq. {eq}`M:8` becomes

$$
\begin{align}
\loc\bar{\sb{\sigma}}(\bb{r},t) &= \loc\sb{\sigma}(\bb{x},t)   
\approx \dot{p}_0(x)\,u(\bb{x},t)\,\one + \mat\sb{\sigma}(\bb{x},t)
\end{align}
$$ (M:12)

with $u$ being the radial displacement

$$
u(\bb{x},t) = \bb{u}(\bb{x},t)\cdot\bn{r}(\vartheta,\varphi)
$$ (M:13)

```

### Spherical harmonic expansion

```{div} full-width

In the following, in order to take advantage of the spherical symmetry, it will be convinient to consider the spherical harmonic expansions of the density disttibution and the gravitational potential  

$$
\begin{align}
& \rho^E(\bb{x},t) = \sum_n\sum_m \rho^E_{n,m}(x,t)\,Y_n^m(\vartheta,\varphi) \qquad \loc\rho^E(\bb{x},t) = \sum_n\sum_m \loc\rho^E_{n,m}(x,t)\,Y_n^m(\vartheta,\varphi)
\\
& \phi^E(\bb{x},t) = \sum_n\sum_m \phi^E_{n,m}(x,t)\,Y_n^m(\vartheta,\varphi) \qquad \loc\phi^E(\bb{x},t) = \sum_n\sum_m \loc\phi^E_{n,m}(x,t)\,Y_n^m(\vartheta,\varphi)
\end{align}
$$ (M:14)

In this respect, the spherical harmonic coefficients $\rho_{n,m}^E$ and $\phi_{n,m}^E$ are given by

$$
\begin{align}
& \rho_{n,m}^E(x,t) = \begin{cases}
\varrho_0(x) + \loc\rho_{0,0}^E(x,t) & n=0 \\ \loc\rho^E_{n,m}(x,t) & n\geq 1 \end{cases}
\qquad
& \phi^E_{n,m}(x,t) = \begin{cases}
\phi^E_0(x) + \loc\phi^E_{0,0}(x,t) & n=0 \\ \loc\phi^E_{n,m}(x,t) & n\geq 1 \end{cases}
\end{align}
$$ (M:15)

In addition to it, we will also consider the spherical harmonic expansion of the (Lagrangian) displacement field

$$
\bb{u}(\bb{x},t) = \sum_n\sum_m\big(U_{n,m}(x,t)\,\bb{P}_n^m(\vartheta,\varphi)+V_{n,m}(x,t)\,\bb{B}_n^m(\vartheta,\varphi)+W_{n,m}(x,t)\,\bb{C}_n^m(\vartheta,\varphi)\big)
$$ (M:16)

where the first term in the summation describe the radial displacement

$$
u(\bb{x},t) =  \sum_n\sum_m U_{n,m}(x,t)\,Y_n^m(\vartheta,\varphi)
$$ (M:17)

while the second and last terms describe tangential displacements. As we will better discuss in CHAPTER, the second term is related to the tangential spheroidal displacement, while the third term to the tangential toroidal displacement.

```

```{admonition} Constraint on toroidal displacement
:class: full-width, dropdown

After the change of variable from the current to the initial coordinates, $\bb{r}\rightarrow\bb{x}$, the relative angular momentum becomes

$$
\bb{h}_E(t) =  \int \varrho_0(x)\,\bb{r}\times\bb{v}(\bb{x},t)\,\md V  \approx   \int \varrho_0(x)\,\bb{x}\times\bb{v}(\bb{x},t)\,\md V 
$$ (M:18)

Let us now make use of the spherical harmonic expansion of the relative velocity

$$
\bb{v}(\bb{x},t) = \derpar{\bb{u}(\bb{x},t)} = \sum_n\sum_m\left(\derpar{U_{n,m}(x,t)}\,\bb{P}_n^m(\vartheta,\varphi)+\derpar{V_{n,m}(x,t)}\,\bb{B}_n^m(\vartheta,\varphi)+\derpar{W_{n,m}(x,t)}\,\bb{C}_n^m(\vartheta,\varphi)\right)
$$ (M:19)

and note that

$$
\begin{align}
\bb{h}_E(t) &=  \sum_n\sum_m \int \varrho_0(x)\,\left(-\derpar{V_{n,m}(x,t)}\,\bb{C}_n^m(\vartheta,\varphi)+\derpar{W_{n,m}(x,t)}\,\bb{B}_n^m(\vartheta,\varphi)\right)\,\md V 
\\
&= \frac{4\,\pi\,\sqrt{2}}{\sqrt{3}}\sum_m \left(\int \varrho_0(x)\,\derpar{W_{1,m}(x,t)}\,x^3\,\md x\right)\,\bb{p}_m(t) = \frac{4\,\pi\,\sqrt{2}}{\sqrt{3}}\sum_m \derpar{w_m(t)}\,\bb{p}_m(t)
\end{align}
$$ (M:20)

with

$$
w_m(t) = \int \varrho_0(x)\,W_{1,m}(x,t)\,x^3\,\md x
$$ (M:21)

In this respect and due to the fact that the relative angular momentum of the solid Earth is zero, we can set the following constraints

$$
w_m(t) = 0
$$ (M:22)

```

## Poisson equation

```{div} full-width

By definition, the Poisson equation governing the initial gravitational potential read

$$
\bar{\nabla}^2 \phi_0^E(r) = C\,\varrho_0(r) 
\qquad \nabla^2 \phi_0^E(x) = C\,\varrho_0(x)
$$ (M:23)

For the local incremental gravitational potential, instead, we have the following incremental Poisson equations

$$
\begin{align}
& \bar{\nabla}^2 \loc\bar{\phi}^E(\bb{r},t) = C\,\loc\bar{\rho}^E(\bb{r},t) \qquad \nabla^2 \loc\phi^E(\bb{x},t) \approx C\,\loc\rho^E(\bb{x},t)  
\end{align}
$$ (M:24)

Here we have made use of the fact that the gradient or the Laplacian of first-order terms in the Eulerian and Lagrangian frameworks differ only by second-order terms, eqs. {eq}`O:31` and {eq}`O:32`.

```

### Initial gravity acceleration

```{div} full-width

Let us now write the initial Poisson equation {eq}`M:23` in terms of the initial gravity due to the solid Earth

$$
\begin{align}
& g_0(x) = \dot{\phi}_0^E(x)
\end{align}
$$ (M:25)

as follows

$$
\begin{align}
& \dot{g}_0(x)+\frac{2}{x}\,g_0(x) = C\,\varrho_0(x)
\end{align}
$$ (M:26)

In particular, in agreement with the MacCullagh theorem, eq. {eq}`G:21`, outside the solid Earth $x\geq a$, we have

$$
g_0(x\geq a) = \frac{G\,M_E}{x^2}
$$ (M:27)

with $M_E$ being the solid Earth mass

$$
M_E = \int\varrho_0(r)\,r^2\,\md \bar{V} = \int\varrho_0(x)\,x^2\,\md V = 4\,\pi\,\int\varrho_0(x)\,x^2\,\md x
$$ (M:28)

Inside the solid Earth, instead, we can write

$$
g_0(x\leq a) = \frac{C}{x^2}\,\int_0^x \varrho_0(\dif{x})\,\dif{x}^2\,\md x
$$ (M:29)

Furthermore, it will be convinient to define the solid Earth geoid perturbation as follows

$$
\loc G^E(\bb{x},t) = -\frac{\loc\phi^E(\bb{x},t)}{g_0(x)}
\qquad
\loc G_{n,m}^E(x,t) = -\frac{\loc\phi_{n,m}^E(x,t)}{g_0(x)}
$$ (M:30)

```

### Density and volume change

```{div} full.width

By making use of eq. {eq}`V:11`, the relative volume variation given by eq. {eq}`OC:5` can be expanded in spherical harmonics as follows

$$
\chi(\bb{x},t) = \grad\cdot\bb{u}(\bb{x},t) = \sum_n\sum_m \chi(x,t) Y_{n,m}(\vartheta,\varphi)
$$ (M:31)

with

$$
\chi_{n,m}(x,t) = \dot{U}_{n,m}(x,t)+\frac{2\,U_{n,m}(x,t) - \gamma_n\,V_{n,m}(x,t)}{x}
$$ (M:32)

In this respect, the material and local incremental densities read

$$
\mat\rho_{n,m}^E(x,t) = -\varrho_0(x)\,\chi_{n,m}(x,t) 
\qquad
\loc\rho_{n,m}^E(x,t) = -\varrho_0(x)\,\chi_{n,m}(x,t) - \dot{\varrho}_0(x)\,U_{n,m}(x,t)
$$ (M:33)

```

```{admonition} Mass and centre of mass
:class: full-width, dropdown

Within the present formalism and writing eq. {eq}`E:9` for the solid Earth system in following way

$$
M_E = \int \rho^E(\bb{r},t)\,\md \bar{V} = \int \varrho_0(r)\,\md \bar{V} + \int \loc\rho^E(\bb{r},t)\,\md \bar{V} =  M_E + \int \loc\rho^E(\bb{r},t)\,\md \bar{V}
$$ (M:34)

we obtain

$$
\begin{align}
& \int \loc\rho^E(\bb{r},t)\,\md \bar{V} \approx \int \loc\rho^E(\bb{x},t)\,\md V = 0
\end{align}
$$ (M:35)

In particular, after spherical harmonic expansion, eq. {eq}`M:35` becomes
 
$$
\int \loc\rho_{0,0}^E(x)\,x^2\,\md x = 0 \qquad \loc\phi_{0,0}^E(x\geq a,t) = 0
$$ (M:36)

The centre of mass, instead, becomes

$$
M_E\,\bb{Z}_E(t) = M_E\,\bb{Z}(t) + \bb{C}_E(t)
$$ (M:37)

with

$$
\begin{align}
\bb{C}_E(t) &= \int \bar{\rho}^E(\bb{r},t)\,\bb{r}\,\md \bar{V} = \int \loc\bar{\rho}^E(\bb{r},t)\,\bb{r}\,\md \bar{V} \approx \int \rho^E(\bb{x},t)\,\bb{x}\,\md V 
\\
& = -\frac{\sqrt{3}\,a}{G}\,\sum_m\,\phi^E_{1,m}(a,t)\,\bb{p}_m(t) = \sqrt{3}\,M_E\,\sum_m \loc G^E_{1,m}(a,t)\,\bb{p}_m(t)
\end{align}
$$ (M:38)

```

```{admonition}  Inertia tensor
:class: full-width, dropdown

Within the present formalism and from eqs. {eq}`G:35` and {eq}`E:28`, we can write the solid Earth inertia tensor as follows

$$
\bb{I}_E(t) = \int \bar{\rho}^E(\bb{r},t)\,\big(r^2\,\one - \bb{r}\,\bb{r}\big)\,\md \bar{V}
= \int \varrho_0(r)\,\big(r^2\,\one - \bb{r}\,\bb{r}\big)\,\md \bar{V} + \int \loc\bar{\rho}^E(\bb{r},t)\,\big(r^2\,\one - \bb{r}\,\bb{r}\big)\,\md \bar{V}  = \big(I_0+\Delta I_E(t))\,\one + \bb{D}_E(t)
$$ (M:39)

with $I_0$ and $\loc I$ being the solid Earth initial scalar inertia and its perturbation

$$
\begin{align}
& I_0 = \frac{2}{3}\,\int \varrho_0(r)\,r^2\,\md \bar{V} = \frac{2}{3}\,\int \varrho_0(x)\,x^2\,\md V \approx \frac{8\,\pi}{3}\,\int \varrho_0(x)\,x^4\,\md x
\\
& \loc I = \frac{2}{3}\,\int \loc\bar{\rho}^E(\bb{r},t) r^2\,\md \bar{V} = \frac{2}{3}\,\int \loc\rho^E(\bb{x},t) x^2\,\md V = \frac{8\,\pi}{3}\,\int \loc\rho^E_{0,0}(x)\,x^4\,\md x
\end{align}
$$ (M:40)

while $\bb{D}_E$ is the solid Earth deviatoric inertia

$$
\bb{D}_E(t) = -\frac{5\,a^3}{G}\,\sum_m \loc\phi^E_{2,m}(a,t)\,\bb{Q}_m(t) = 5\,M_E\,a\,\sum_m \loc G^E_{2,m}(a,t)\,\bb{Q}_m(t)
$$ (M:41)

```

## Momentum equation

```{div} full-width

In the perspective of obtaining the momentum equation for the deformable Earth, it is convinient to introduce the total gravitational potential $\Phi$ including the contribution from the Earth, $\phi^E$,  athmosphere, $\phi^A$, and the celestial bodies, $\phi^T$, 

$$
\Phi(\bb{x},t) = \phi^E(\bb{x},t) +  \phi^A(\bb{x},t) + \phi_0^T(\bb{x},t) + \phi_1^T(\bb{x},t) + \psi^T(\bb{x},t)
$$ (M:42)

where we use the decomposition given by eq. {eq}`G:56`, $\phi^T = \phi_0^T +\phi_1^T + \psi^T$ with $\psi^T$ being the tidal potential. In addition to it, we shall define  also the total (generalized) potential

$$
\Psi(\bb{x},t) = \phi^E(\bb{x},t)  + \phi^A(\bb{x},t)  + \psi^T(\bb{x},t) + \psi^C(\bb{x},t)
$$ (M:43)

where we have added the centrifugal potential and removed the degree-$0$ and -$1$ contributions of the gravitational potential of the celestial bodies. Its local decomposition reads

$$
\bar{\Psi}(\bb{r},t) = \phi_0^E(\bb{r}) +\loc\bar{\Psi}(\bb{r},t)
$$ (M:44)

with

$$
\loc\bar{\Psi}(\bb{r},t) = \loc\bar{\phi}^E(\bb{r},t) + \bar{\phi}^A(\bb{r},t)  +\bar{\psi}^T(\bb{r},t) +\bar{\psi}^C(\bb{r},t) 
$$ (M:45)

where the gravitational potentials due to the athmosphere and the celestial bodies, as well as the centrifugal potential, have been considered as first-order perturbations.

By taking the Laplacian of the local incremental total potential $\loc\Psi$, we note that it satisfy the following modified incremental Poisson equation

$$
\nabla^2\loc\Psi(\bb{x},t) = C\,\big(\loc\rho^E(\bb{x},t) + \rho^A(\bb{x},t)\big) - 2\,\omega(t)^2
$$ (M:46)

where we omit the celestial body density distribution because we will solve it just in the surrounding of the solid Earth, $\|\bb{x}\|=x \leq a^+$.

```


### Second newtown law

```{div} full-width

Let us now obtain the momentum equation describing the balance between surface and volume forces acting on an infinitesimal element of the Earth system. Let us consider a time-dependent region $\mathcal{V}_t$ evolving according to the displacement field $\bb{u}$ and the time derivative of the linear momentum to the total surface and volume forces

$$
\begin{align}
\int_{\mathcal{V}_t} \bar{\rho}^E(\bb{r},t)\,\bar{\bb{A}}(\bb{r},t)\,\md V &= \int_{\partial \mathcal{V}_t} \bar{\bb{n}}(\bb{r},t)\cdot\bar{\sb{\sigma}}(\bb{r},t)\,\md \bar{S} - \int_{\mathcal{V}_t}\bar{\rho}^E(\bb{r},t)\,\bar{\grad}\bar{\Phi}(\bb{r},t)\,\md \bar{V}  
\\
&= \int_{ \mathcal{V}_t} \big(\bar{\grad}\bar{\sb{\sigma}}(\bb{r},t)-\bar{\rho}^E(\bb{r},t)\,\bar{\grad}\bar{\Phi}(\bb{r},t)\big)\,\md \bar{V} 
\end{align}
$$ (M:47)

where $\bar{\sb{\sigma}}$ is the Eulerian Cauchy stress tensor, $\partial \mathcal{V}_t$ is the surface of the deforming region $\mathcal{V}_t$ and $\bar{\bb{n}}$ is its outward normal unit vector. By the arbitraryness of the region $\mathcal{V}_t$, the above equation holds at the local level and it yields the so called momentum equation

$$
\begin{align}
\bar{\grad}\cdot\bar{\sb{\sigma}}(\bb{r},t) -\bar{\rho}^E(\bb{r},t)\,\bar{\grad}\bar{\Phi}(\bb{r},t) = \bar{\rho}^E(\bb{r},t)\,\bar{\bb{A}}(\bb{r},t)
\end{align}
$$ (M:48)

By making use of eqs. {eq}`O:54` and {eq}`E:14`, it can be recasted as follows

$$
\begin{align}
 \bar{\grad}\cdot\bar{\sb{\sigma}}(\bb{r},t) - \bar{\rho}^E(\bb{r},t)\,\bar{\grad} \bar{\Phi}(\bb{r},t)
&= \bar{\rho}^E(\bb{r},t)\,\bar{\bb{a}}(\bb{r},t) +  \bar{\rho}^E(\bb{r},t)\,\ddercla{ \bb{Z}(t)}   + \bar{\rho}^E(\bb{r},t)\,\bar{\grad}\bar{\psi}^C 
\\
&+ \bar{\rho}^E(\bb{r},t)\,\left(\derpar{\sb{\omega}(t)}\times\bb{r} + 2\,\sb{\omega}(t)\times\bar{\bb{v}}(\bb{r},t) \right)
\end{align}
$$ (M:49)

that, in view of eq. {eq}`M:45`, can be rewritten in terms of the total (generalized) potential $\Psi$

$$
\begin{align}
\bar{\grad}\cdot\bar{\sb{\sigma}}(\bb{r},t) - \bar{\rho}^E(\bb{r},t)\,\bar{\grad}\bar{\Psi}
&= \bar{\rho}^E(\bb{r},t)\,\bar{\bb{a}}(\bb{r},t)  + \bar{\rho}^E(\bb{r},t)\,\left(\ddercla{\bb{Z}(t)}-\bb{g}_1^T(t)\right) \\
&+ \bar{\rho}^E(\bb{r},t)\,\left(\derpar{\sb{\omega}(t)}\times\bb{r} + 2\,\sb{\omega}(t)\times\bar{\bb{v}}(\bb{r},t)\right)
\end{align}
$$ (M:50)

In view of eq. {eq}`E:29`, the second term in the right-hand side yields zero. Furthermore,  the Euler and Coriolis accelerations can be neglected for studying solid Earth deformation and, so, eq. {eq}`M:50` simplifies into

$$
\begin{align}
& \bar{\grad}\cdot\bar{\sb{\sigma}}(\bb{r},t) - \bar{\rho}^E(\bb{r},t)\,\bar{\grad}\bar{\Psi}(\bb{r},t)
= \bar{\rho}^E(\bb{r},t)\,\bar{\bb{a}}(\bb{r},t) 
\end{align}
$$ (M:51)

that is the momentum equation that we shall solve.

```

### Initial momentum equation

```{div} full-width

We define the initial state in such a way that the following initial momentum equation holds

$$
\begin{align}
& \bar{\grad}\cdot\sb{\sigma}_0(\bb{r}) - \varrho_0(\bb{r})\,\bar{\grad}\phi_0^E(\bb{r})
=\zero
\end{align}
$$ (M:52)

or, equivalently,

$$
\begin{align}
& \grad\cdot\sb{\sigma}_0(\bb{x}) - \varrho_0(\bb{x})\,\grad\phi_0^E(\bb{x})
=\zero
\end{align}
$$ (M:53)

where $\sb{\sigma}_0$ is the initial Cauchy stress tensor. In the present case, assuming that the initial state is in hydrostatic equilibrium, eq. {eq}`M:53` becomes

$$
\begin{align}
& -\grad p_0(\bb{x}) - \varrho_0(\bb{x})\,\grad\phi_0^E(\bb{x})
=\zero
\end{align}
$$ (M:54)

Then, further assuing that the initial state is spherically symmetric, we obtain

$$
-\dot{p}_0(x) - \varrho_0(x)\,g_0(x) = 0
$$ (M:55)

```

### Incremental momentum equation

```{div} full-width

By subtracting eq. {eq}`M:53`  from eq. {eq}`M:51` and neglecting second-order terms, we thus obtain the incremental momentum equation

$$
\begin{align}
& \bar{\grad}\cdot\loc\bar{\sb{\sigma}}(\bb{r},t) -\varrho_0(\bb{r},t)\,\bar{\grad}\loc\bar{\Psi}(\bb{r},t) - \loc\bar{\rho}^E(\bb{r},t)\,\bar{\grad}\phi_0^E(\bb{r})
= \varrho_0(\bb{r})\,\bar{\bb{a}}(\bb{r},t) 
\end{align}
$$ (M:56)

Furthermore, by considering the Lagrangian representation of the local incremental fields and that the gradient or Laplacian of fields with respect to the current coordinates can be approximated, at the first order, with their gradient or Laplacian with respect to the initial coordinates, eqs. {eq}`O:31` and {eq}`O:32`, we can also write

$$
\begin{align}
& \grad\cdot\loc\sb{\sigma}(\bb{x},t) -\varrho_0(\bb{x})\,\grad\loc\Psi(\bb{x},t) - \loc\rho^E(\bb{x},t)\,\grad\phi_0^E(\bb{x}) = \varrho_0(\bb{x})\,\bb{a}(\bb{x},t) 
\end{align}
$$ (M:57)

Then, by substituting eq. {eq}`M:12`  into the incremental momentum equation and making use of the spherical symmetry of the initial statse, we thus obtain 

$$
\grad\cdot\mat\sb{\sigma}(\bb{x},t) - \grad\big(\varrho_0(x)\,g_0(x)\,u(\bb{x},t)\big) -\varrho_0(x)\,\grad\loc\Psi(\bb{x},t) - \loc\rho^E(\bb{x},t)\,g_0(x)\,\bn{r}(\vartheta,\varphi)
= \rho_0(x)\,\bb{a}(\bb{x},t) 
$$ (M:58)

```

## Self-gravitating spherically-symmetric Earth models

```{div} full-width

Let us now summarize the differential equations describing the initial and incremental state of the Earth and consider both the momentum and Poisson equation. For the initial state, from eqs. {eq}`M:26` and {eq}`M:55`, we have

$$
\begin{align}
& -\dot{p}_0(x) - \varrho_0(x)\,g_0(x) = 0
\\
& \dot{g}_0 + \frac{2}{x}\,g_0(x) =  C\,\varrho_0(x)
\end{align}
$$ (M:59)

As it concern the incremental state, instead, from eqs. {eq}`M:24` and {eq}`M:58`

$$
\begin{align}
& \grad\cdot\mat\sb{\sigma}(\bb{x},t) - \grad\big(\varrho_0(x)\,g_0(x)\,u(\bb{x},t)\big) -\varrho_0(x)\,\grad\loc\Psi(\bb{x},t) - \loc\rho^E(\bb{x},t)\,g_0(x)\,\bn{r}(\vartheta,\varphi)
= \rho_0(x)\,\bb{a}(\bb{x},t) 
\\
& \nabla^2\loc\Phi(\bb{x},t) = C\,\big(\loc\rho^E(\bb{x},t) + \rho^A(\bb{x},t)\big) - 2\,\omega(t)^2
\end{align}
$$ (M:60)

Note that we have included the density distribution of the athmosphere in the incremental Poisson equation due to the fact that the (generalized) total potential accounts also for the gravitational potential of the athmosphere. In order to fix the ideas, the incremental momentum equation will be solved within the solid Earth, i.e. for all $\bb{x}$ such that $x=\big\|\bb{x}\big\|<a$, while the incremental Poisson equation will be solved up to just above the Earth surface, i.e., for all $\bb{x}$ such that $x=\big\|\bb{x}\big\|\leq a^+$.

```

### Internal and surface loads

```{div} full-width

So far, we have not considered internal and surface loads describing mantle density heterogeneities due to cold or hot materials like sinking slabs and plumes or ice and water redistributions over the Earth surface. In principle, these loads could be described as density perturbations of the initial density distribution of the Earth. From a practical point of view, however, this possibility would require to take into account for the energy equation for describing the theremal state of the Earth, as well as specific equations governing the physics of criosphere and hydrosphere, ths describing in detail glaciers, acquifers, oceans and other form of water storage.

Due to these difficulties, as first approximation, we will describe internal and surface loads as prescribed time-dependent density distributions. In particular, we will denote with $\rho^I$ the internal load density distribution and with $\rho^S$ the surface load one, for which we can also write

$$
\rho^S(\bb{x},t) = \sigma^S(\vartheta,\varphi)\,\delta(x-a)
$$ (M:61)

with $a$ being the radius of the Earth.

According to this approach, the local incremental density, $\loc\rho$, and potential, $\loc\phi$ of the Earth will be incremnted by the loading contributions

$$
\loc\rho^E(\bb{x},t) \rightarrow \loc\rho^E(\bb{x},t) + \rho^I(\bb{x},t) + \rho^S(\bb{x},t)
\qquad
\loc\phi^E(\bb{x},t) \rightarrow \loc\phi^E(\bb{x},t) + \phi^I(\bb{x},t) + \phi^S(\bb{x},t)
$$ (M:62)

where $\phi^I$ and $\phi^S$ solve the following Poisson equations

$$
\nabla^2\phi^I(\bb{x},t) = C\,\rho^I(\bb{x},t)
\qquad
\nabla^2\phi^S(\bb{x},t) = C\,\rho^S(\bb{x},t)
$$ (M:63)

This means that the total (generilized potential) becomes

$$
\Psi(\bb{x},t) = \phi^E(\bb{x},t) +\phi^I(\bb{x},t)+\phi^S(\bb{x},t)+\phi^A(\bb{x},t) +\psi^T(\bb{x},t)  +\psi^C(\bb{x},t) 
$$ (M:64)

and that its local increment reads

$$
\loc\Psi(\bb{x},t) = \loc\phi^E(\bb{x},t) +\phi^I(\bb{x},t)+\phi^S(\bb{x},t)+\phi^A(\bb{x},t) +\psi^T(\bb{x},t)  +\psi^C(\bb{x},t) 
$$ (M:65)

In this respect, the incremental momentum and Poisson equations become

$$
\begin{align}
& \grad\cdot\mat\sb{\sigma}(\bb{x},t) - \grad\big(\varrho_0(x)\,g_0(x)\,u(\bb{x},t)\big) -\rho_0(x)\,\grad\loc\Psi(\bb{x},t) - \big(\loc\rho^E(\bb{x},t)+\rho^I(\bb{x},t)\big)\,\grad\phi_0(x)
= \rho_0(x)\,\bb{a}(\bb{x},t) 
\\
& \nabla^2\loc\Psi(\bb{x},t) = C\,\big(\loc\rho^E(\bb{x},t)+\rho^I(\bb{x},t)+\rho^S(\bb{x},t)+\rho^A(\bb{x},t)\big)
\end{align}
$$ (M:66)
 
where we have included only the internal load density in the momentum equation because we will solve this equation only within the solid Earth.

Another aspect that we need to remark is that the total mass of the solid Earth conserve and that the choice that the translating origin coincide with the centre of mass of the Earth (including the athmoshere) system now have to include also the internal and surface loads. In other words, eq. {eq}`M:36` means also that
 
$$
\int \big(\loc\rho_{0,0}^I(x)+\loc\rho_{0,0}^S(x)\big)\,x^2\,\md x = 0 \qquad \loc\phi_{0,0}^I(x\geq a,t)+\loc\phi_{0,0}^S(x\geq a,t)  = 0
$$ (M:67)

while eq. {eq}`M:38` becomes

$$
\begin{align}
\bb{C}_E(t) &= \sqrt{3}\,M_E\,\sum_m \big(\loc G^E_{1,m}(a,t)+\loc G^I_{1,m}(a,t)+\loc G^S_{1,m}(a,t)\big)\,\bb{p}_m(t)
\end{align}
$$ (M:68)

```
