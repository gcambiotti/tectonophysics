

$\newcommand{\md}{\mathrm{d}\hspace{0cm}}$
$\newcommand{\mdd}{\mathrm{d}\hspace{0cm}^2}$
$\newcommand{\grad}{\boldsymbol{\nabla}\hspace{-0.01cm}}$
$\newcommand{\stress}{\boldsymbol{\sigma}}$
$\newcommand{\bn}[1]{\hat{\mathbf{#1}}}$
$\newcommand{\bb}[1]{\mathbf{#1}}$
$\newcommand{\sn}[1]{\hat{\boldsymbol{#1}}}$
$\newcommand{\sb}[1]{\boldsymbol{#1}}$
$\newcommand{\zero}{\mathbf{0}}$
$\newcommand{\one}{\mathbf{1}}$
$\newcommand{\loc}{\Delta\hspace{0cm}}$
$\newcommand{\mat}{\delta\hspace{0cm}}$
$\newcommand{\det}{\mathrm{\det}\hspace{0cm}}$
$\newcommand{\cm}[1]{\bar{\bb{#1}}}$
$\newcommand{\lmsum}{\sum_{n=0}^\infty\sum_{m=-n}^n}$
$\newcommand{\kqsum}{\sum_{k=0}^\infty\sum_{q=-k}^k}$
$\newcommand{\dif}[1]{#1\hspace{0cm}\prime}$
$\newcommand{\lap}{\nabla^2\hspace{0cm}}$
$\newcommand{\dermat}[1]{\frac{\mathrm{D}\hspace{0.02cm}#1}{\mathrm{D}\hspace{0.02cm} t}}$
$\newcommand{\dercla}[1]{\frac{\mathrm{d}\hspace{0.02cm}#1}{\mathrm{d}\hspace{0.02cm} t}}$
$\newcommand{\derpar}[1]{\frac{\partial\hspace{0.02cm}#1}{\partial\hspace{0.02cm} t}}$
$\newcommand{\rpar}[2]{\frac{\partial\hspace{0.02cm}#1}{\partial\hspace{0.02cm} r_{#2}}}$
$\newcommand{\xpar}[2]{\frac{\partial\hspace{0.02cm}#1}{\partial\hspace{0.02cm} x_{#2}}}$
$\newcommand{\ddermat}[1]{\frac{\mathrm{D}^2\hspace{0.02cm}#1}{\mathrm{D}\hspace{0.02cm} t^2}}$
$\newcommand{\ddercla}[1]{\frac{\mathrm{d}^2\hspace{0.02cm}#1}{\mathrm{d}\hspace{0.02cm} t^2}}$
$\newcommand{\dderpar}[1]{\frac{\partial^2\hspace{0.02cm}#1}{\partial\hspace{0.02cm} t^2}}$
$\newcommand{\drpar}[2]{\frac{\partial^2\hspace{0.02cm}#1}{\partial\hspace{0.02cm} r_{#2}^2}}$
$\newcommand{\dxpar}[2]{\frac{\partial^2\hspace{0.02cm}#1}{\partial\hspace{0.02cm} x_{#2}^2}}$

## Mass conservation

```{div} full-width

The mass contained with a time-dependent region $\mathcal{V}_t$ of a deformable continuum body evolving according to the displacement field  $\bb{u}$ can be obtained as

$$
M = \int_{\mathcal{V}_t} \bar{\rho}(\bb{r},t)\,\md \bar{V}
$$ (E:3)

where $\md \bar{V}=\md r_1\,\md r_2\,\md r_3$ being the infinitesimal volume element. By noting that

$$
\md \bar{V} = J(\bb{x},t)\,\md V
$$ (E:4)

with $\md V=\md x_1\,\md x_2\,\md x_3$ and $J$ being the Jacobian of the transformation $\bb{r}\rightarrow \bb{x}$ as in eq. {eq}`OC:4`

$$
J(\bb{x},t) = \mathrm{det}\big(\bb{J}(\bb{x},t)\big) \approx 1+\grad\bb{u}(\bb{x},t) \qquad \mathrm{with}\qquad \bb{J}(\bb{x},t) = \grad\bb{r} = \one + \grad\bb{u}(\bb{x},t)
$$ (E:5)

where the approximation holds at the first-order approximation. In light of this, after the change of variable, eq. {eq}`E:3` becomes 

$$
M = \int_{\mathcal{V}_0} \rho(\bb{x},t)\,J(\bb{x},t)\,\md V
$$ (E:6)

In this respect and requiring the mass conservation of any deformed region system, we thus understand that the product between $\rho$ and $J$ corresponds to the initial density distribution $\rho_0$ that, by definition, does not depend on time

$$
\rho_0(\bb{x}) =  \rho(\bb{x},t)\,J(\bb{x},t)
$$ (E:8)

This interpretation is correct also in view of the fact that $J$ represent the relative volume variations of the infinitesimal volume element of the deformable continuum body.

```

## Centre of mass and orbital motion

```{div} full-width

Let us now consider the mass, $M$ and the centre of mass, $\bb{Z}$, of the Earth system

$$
\begin{align}
& M = \int \bar{\rho}(\bb{r},t)\,\md \bar{V} = \int \rho_0(\bb{x})\,\md V
\\
& M\,\bb{Z}(t) = \int \bar{\rho}(\bb{r},t)\,\bb{R}(\bb{r},t)\,\md \bar{V} = M\,\bb{X}_O(t) + \bb{C}(t)
\end{align}
$$ (E:9)

with $\bb{C}$ being the first moment of the density distribution with respect to the translating origin

$$
\bb{C}(t) = \int \bar{\rho}(\bb{r},t)\,\bb{r}\,\md \bar{V} = \int \rho_0(\bb{x})\,\bb{r}\,\md V
$$ (E:10)

```


```{admonition} Time derivatives of the density first moment 
:class: full-width, dropdown

We note that

$$
\begin{align}
&\dercla{\bb{C}(t)} = \int \rho_0(\bb{x})\,\big(\sb{\omega}\times\bb{r}+\bb{v}(\bb{r},t)\big)\,\md V = \sb{\omega}\times\bb{C}(t) + \derpar{\bb{C}(t)}
\\
& \ddercla{\bb{C}(t)} = \sb{\omega}(t)\times\sb{\omega}(t)\times\bb{C}(t) + \derpar{\sb{\omega}(t)}\times\bb{C} + 2\,\sb{\omega}(t)\times\derpar{\bb{C}(t)} + \dderpar{\bb{C}(t)}
\end{align}
$$ (E:11)

with

$$
\begin{align}
&\derpar{\bb{C}(t)} = \int \bar{\rho}(\bb{r},t)\,\bb{v}(\bb{r},t)\,\md \bar{V} = \int \rho_0(\bb{x})\,\bb{v}(\bb{r},t)\,\md V 
\\
& 
\dderpar{\bb{C}(t)} = \int \bar{\rho}(\bb{r},t)\,\bb{a}(\bb{r},t)\,\md \bar{V} = \int \rho_0(\bb{x})\,\bb{a}(\bb{r},t)\,\md V
\end{align} 
$$ (E:12)

```

### Linear momentum

```{div} full-width

The total linear momentum and its time derivative read

$$
\begin{align}
&\bb{P}(t) = \int \bar{\rho}(\bb{r},t)\,\bb{V}(\bb{r},t)\,\md \bar{V} = M\,\dercla{\bb{X}_O(t)} +\dercla{\bb{C}(t)}
\\
& \dercla{\bb{P}(t)} = \int \bar{\rho}(\bb{r},t)\,\bb{A}(\bb{r},t)\,\md \bar{V} = M\,\ddercla{\bb{X}_O(t)} +\ddercla{\bb{C}(t)}
\end{align}
$$ (E:13)

As first choice, we fix the translating origin $\bb{X}_O$ so that it coincides with the centre of mass

$$
\bb{Z}(t) = \bb{X}_O(t) \qquad \bb{C}(t) = \zero
$$ (E:14)

and write the total linear momentum and its derivative simply as

$$
\bb{P}(t) =  M\,\dercla{\bb{Z}(t)}\qquad
\dercla{\bb{P}(t)} = M\,\ddercla{\bb{Z}(t)} 
$$ (E:15)

Then by considering that the total forces from the celestial body acting on nearly spherical density distribution of total mass $M$ reads

$$
\bb{F}_T(t) = M\,\bb{g}_1(t)
$$ (E:16)

from the second Newton law, we obatin the acceleration of the centre of mass in the inertial frame


$$
\ddercla{\bb{P}(t)} = \bb{F}_T(t)
\qquad \Rightarrow \qquad 
\ddercla{\bb{Z}(t)}  = \bb{g}_1^T(t)
$$ (E:17)

Note that $\bb{g}_1^T$ depends on time because the spherical coordinate of the celestial body with respect to Earth system can vary throug time and due to fact that we are considering the rotating unit vectors

$$
\bb{g}_1^T = \frac{G\,M_T}{x_T^2(t)}\,\bn{r}(\vartheta_T(t),\varphi_T(t),t) = \frac{G\,M_T}{x_T^2(t)}\,\Big[\sin\vartheta_T(t)\,\big(\sin\varphi_T(t)\,\bb{e}_1(t)+\cos\varphi_T(t)\,\bn{e}_2(t)\big)+\cos\vartheta_T(t)\,\bn{e}_3(t)\Big]
$$ (E:16a)

```

### Solid earth and athmosphere

```{div} full-width

Let us now distinguish the density distribution $\rho$ into the solid Earth, $\rho^E$, and athmosphere, $\rho^A$, contributions

$$
\rho(\bb{x},t) = \rho^E(\bb{x},t) + \rho^A(\bb{x},t)
$$ (E:18)

and write the centres of mass and the linear momenta of the solid Earth or the athmosphere

$$
\begin{align}
&M_X\,\bb{Z}_X(t) = M_X\,\bb{Z}(t)+\bb{C}_X(t)
\qquad
\bb{P}_X(t) = M_X\,\dercla{Z}(t) + \dercla{\bb{C}_X(t)}
\end{align}
$$ (E:19)

with $X=E$ or $A$, and

$$
\bb{C}_X(t) = \int \bar{\rho}^X(\bb{r},t)\,\bb{r}\,\md \bar{V}
$$ (E:20)

By definition, the following relation between the masses, centres of mass and linear momenta of the Earth system, of the solid Earth and of the athmosphere hold

$$
M_E + M_A = M
\qquad
M_E\,\bb{Z}_E(t) + M_A\,\bb{Z}_A(t) = M\,\bb{Z}(t)
\qquad
\bb{P}_E(t) + \bb{P}_A(t) = \bb{P}(t)
$$ (E:21)

We note that, from eq. {eq}`E:17`, we can obtain

$$
\bb{C}_E(t) = - \bb{C}_A(t)
$$ (E:22)

Let us now consider the total force from the athmosphere on the Earth

$$
\bb{F}_{AE}(t) =  \int_{\partial V_t}\bb{T}^A(\bb{r},t)\,\md S - \int  \bar{\rho}^E(\bb{r},t)\,\bar{\grad}\bar{\phi}^A(\bb{r},t)\,\md \bar{V}
$$ (E:23)

where $\phi^A$ and $\bb{T}^A$ being the athmosphere gravitational potential and the traction of the athmosphere acting on the solid Earth. Also, let us consider the total force from the celestial body on the solid Earth, $\bb{F}_{TE}$, and on the athmosphere, $\bb{F}_{TA}$, 

$$
\bb{F}_{TX}(t) = M_X\,\bb{g}_1^T(t)
$$ (E:24)

with $X=E$ or $A$. In view of the second Newton law applied to the solid Earth or athmosphere

$$
\dercla{\bb{P}_E(t)} = \bb{F}_{TE}(t) + \bb{F}_{AE}(t)  \qquad \dercla{\bb{P}_A(t)} = \bb{F}_{TA}(t) - \bb{F}_{AE}(t)
$$ (E:25)

we obatin the following relation between the acceleration of the density first momenta of the solid Earth and the athmosphere and the total force between them

$$
\ddercla{\bb{C}_E(t)} = -\ddercla{\bb{C}_A(t)} = \bb{F}_{AE}(t) 
$$ (E:26)

Note that the total force from the solid Earth to the athmosphere used in eq. {eq}`E:25` is just $-\bb{F}_{AE}$.

```

## The choice of the angular velocity

```{div} full-width

Let us now consider the total linear momenta of the solid Earth or the athmosphere under the assumption that $\bb{Z}=\bb{X}_O$

$$
\begin{align}
\bb{H}_X(t) &= \int\bar{\rho}^X(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\bb{V}}(\bb{r},t)\,\md \bar{V} = \bb{X}_O(t)\times\bb{P} + \int\bar{\rho}^X(\bb{r},t)\,\bb{r}\times\bar{\bb{V}}(\bb{r},t)\,\md \bar{V}
\\
&= \bb{Z}(t)\times\bb{P}_X(t) + \bb{C}_X(t)\times\dercla{\bb{Z}(t)} + \int\bar{\rho}^X(\bb{r},t)\,\bb{r}\times\big(\sb{\omega}(t)\times\bb{r}+\bar{\bb{v}}(\bb{r},t)\big)\,\md \bar{V}
\\
&= \bb{Z}(t)\times\bb{P}_X(t) + \bb{C}_X(t)\times\dercla{\bb{Z}(t)} + \bb{I}_X(t)\cdot\sb{\omega}(t) + \bb{h}_X(t)
\end{align}
$$ (E:27)

with $\bb{I}_X$ and $\bb{h}_X$ being the inertia tensor and the relative angular momentum

$$
\begin{align}
& \bb{I}_X(t) = \int\bar{\rho}^X(\bb{r},t)\,\big(r^2\,\one - \bb{r}\,\bb{r}\big)\,\md \bar{V}
\\
& \bb{h}_X(t) = \int\bar{\rho}^X(\bb{r},t)\,\bb{r}\times\bar{\bb{v}}(\bb{r},t)\,\md \bar{V} 
\end{align}
$$ (E:28)

In order to define the angular velocity, we adopt the Tisserand reference frame in which the angular velocity $\sb{\omega}$ is choosen in such a way that relative angular momentum of the solid Earth is zero

$$
\bb{h}_E(t) = \zero
$$ (E:29)

```

```{note}
:class: full-width, dropdown

This choice is always possible. Let us indeed consider the following angular velocity

$$
\sb{\omega}^*(t) = \sb{\omega}(t)+\mat\sb{\omega}(t)
$$ (E:30)

and denote with $\bb{e}^*_i$ the respective rotating unit vectors. Then, let us expresse the velocity in the two distinct rotating reference frame

$$
\bb{V}(\bb{x},t) = \dercla{\bb{Z}(t)} + \sb{\omega}^*(t)\times\bb{r} + \bb{v}^*(\bb{x},t) = \dercla{\bb{Z}(t)} + \sb{\omega}(t)\times\bb{r} + \bb{v}(\bb{x},t)
$$ (E:31)

We thus understand that

$$
\bb{v}^*(\bb{x},t) = \bb{v}(\bb{x},t) - \mat\sb{\omega}(t)\times\bb{r}
$$ (E:32)

and the relative angular momentum can be arrange as follows

$$
\bb{h}_E^*(t) = \int\bar{\rho}^E(\bb{r},t)\,\bb{r}\times\bar{\bb{v}}^*(\bb{r},t)\,\md \bar{V} =  \bb{h}_E(t) - \int \bar{\rho}^E(\bb{r},t)\,\bb{r}\times \mat\sb{\omega}(t)\times\bb{r}\,\md \bar{V} = \bb{h}_E(t) - \bb{I}_E(t)\cdot\mat\sb{\omega}(t)
$$ (E:33)

In this respect, we thus understand that it can be set to zero by properly choosing $\mat\sb{\omega}$

$$
\mat\sb{\omega}(t) = \bb{I}_E(t)^{-1}(t)\cdot\bb{h}_E(t) = \zero  \qquad\Rightarrow \qquad \bb{h}_E^*(t)
$$ (E:34)

and, so, there exist an angular velocity vector $\sb{\omega}^*$ such that the relative angular momentum is zero.

```




```{div} full-width

This choice of the angulra velocity $\sb{\omega}$ based on the solid Earth system, along with the choice of the translating origin $\bb{X}_O$ taken as the centre of mass of the Earth system (including the athmosphere), fully defines the translating and rotating reference system discussed in  {numref}`Chapter %s <chap-kinematics>` - {ref}`chap-kinematics`. 

```

### Rotational Euler equation

```{div} full-width

Within this framework, the angular momentum of the solid Earth becomes

$$
\bb{H}_E(t) = \bb{Z}(t)\times\bb{P}_E(t) + \bb{C}_E(t)\times\dercla{\bb{Z}(t)} + \bb{I}_E(t)\cdot\sb{\omega}(t) 
$$ (E:35)

and its time derivative yields

$$
\begin{align}
\dercla{\bb{H}_E} &= \dercla{\bb{Z}(t)}\times\bb{P}_E(t) + \bb{Z}\times\dercla{\bb{P}_E} + \dercla{\bb{C}_E(t)}\times\dercla{\bb{Z}(t)} + \bb{C}_E\times\ddercla{\bb{Z}} + \dercla{\big(\bb{I}_E(t)\cdot\sb{\omega}(t) \big)} 
\\
&= \dercla{\bb{Z}(t)}\times\left(M_E\,\dercla{\bb{Z}} + \dercla{\bb{C}_E}\right) + \bb{Z}\times\left(M_E\,\bb{g}_1^T + \bb{F}_{AE}\right) + \dercla{\bb{C}_E(t)}\times\dercla{\bb{Z}(t)} + \bb{C}_E\times\ddercla{\bb{Z}} + \dercla{\big(\bb{I}_E(t)\cdot\sb{\omega}(t) \big)} 
\\
&= \big(M_E\,\bb{Z}+\bb{C}_E\big)\times\bb{g}_1^T + \bb{Z}\times\ddercla{\bb{C}_E} + \dercla{\big(\bb{I}_E(t)\cdot\sb{\omega}(t) \big)} 
\end{align}
$$ (E:36)

Let us now consider the torque from the athmosphere and from the celestial bodies acting on the solid Earth

$$
\begin{align}
\bb{K}_{AE}(t) &= \int_{\partial V_t} \bar{\bb{R}}(\bb{r},t)\times\bb{T}^A(\bb{r},t)\,\md S - \int  \bar{\rho}^E(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\grad}\bar{\phi}^A(\bb{r},t)\,\md \bar{V} = \bb{Z}\times\bb{F}_{AE} + \bb{k}_{AE}
\\
&= \bb{Z}(t)\times\ddercla{\bb{C}_E(t)} + \bb{k}_A(t)
\\
\bb{K}_{TX}(t) &= -\int \bar{\rho}^X(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\grad}\bar{\phi}^T(\bb{r},t)\,\md \bar{V} = \big(M_X\,\bb{Z} +\bb{C}_X\big)\times\bb{g}_1 -\int \bar{\rho}^X(\bb{r},t)\,\bar{\bb{R}}(\bb{r},t)\times\bar{\grad}\bar{\psi}^T(\bb{r},t)\,\md \bar{V}
\\
&= \big(M_X\,\bb{Z}(t) +\bb{C}_X(t)\big)\times\bb{g}_1^T(t) + \bb{Z}(t)\times\int \bar{\rho}^X(\bb{r},t)\,\bar{\grad}\bar{\psi}^T(\bb{r},t)\,\md\bar{V} + \int \bar{\rho}^X(\bb{r},t)\,\bb{r}\times\bar{\grad}\bar{\psi}^T(\bb{r},t)\,\md \bar{V}
\\
&= \big(M_X\,\bb{Z}(t) +\bb{C}_X(t)\big)\times\bb{g}_1^T(t) 
\end{align}
$$ (E:37)

where, in the last identities for $\bb{K}_{TX}$, we have neglected the remaining volume integrals in view of the results of {numref}`Section %s <sec-total-force>` - {ref}`sec-total-force` and {numref}`Section %s <sec-total-torque>` - {ref}`sec-total-torque`, and defined the relative torque $\bb{k}_{AE}$ from the athmosphere with respect to the centre of mass on the solid Earth

$$
\bb{k}_{AE}(t) = \int_{\partial V_t} \bb{r}\times\bb{T}^A(\bb{r},t)\,\md S - \int  \bar{\rho}^E(\bb{r},t)\,\bb{r}\times\bar{\grad}\bar{\phi}^A(\bb{r},t)\,\md \bar{V} 
$$ (E:38)

In light of this, the conservation of the angular momentum of the solid Earth system

$$
\dercla{\bb{H}_E(t)} = \bb{K}_{TE}(t) + \bb{K}_{AE}(t)
$$ (E:39)

yields the so called Euler rotational equation

$$
\dercla{\big(\bb{I}_E(t)\cdot\sb{\omega}(t) \big)}  = \derpar{\big(\bb{I}_E(t)\cdot\sb{\omega}(t) \big)} + \sb{\omega}\times\big(\bb{I}_E(t)\cdot\sb{\omega}(t) \big) = \bb{k}_{AE}(t)
$$ (E:40)

We note that the relative torques of the athmosphere can thus change the angular momentum of the solid Earth. Different, the relative torque from celestial bodies does not enter the above equation because, within the assumption of infinitesimal departure from the spherical symmetry and as already pointed out by eq. {eq}`G:73`, it is zero at the first order.

```
