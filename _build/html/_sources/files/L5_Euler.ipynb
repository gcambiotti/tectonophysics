## Inertia tensor perturbations

```{div} full-width

In order to obtain the inertia tensor of the solid Earth and loads, we rewrite  eq. {eq}`M:39` as follows

$$
\bb{I}(t) = \big(I_0+\loc  I(t))\big)\,\bb{1} + \bb{D}(t)
$$ (R:1)

that we have decomposed into the initial scalar inertia $I_0$

$$
\begin{align}
& I_0 = \frac{8\,\pi}{3}\,\int_0^a \varrho_0(x)\,x^4\,\md x 
\end{align}
$$ (R:2)

its perturbation $\loc I$ and the deviatoric inertia tensor $\bb{D}$. The latter two contributions, in view of eqs. {eq}`M:40`, {eq}`M:41` and {eq}`M:62` and  {eq}`M:62`, give

$$
\begin{align}
& \loc I(t) = \frac{8\,\pi}{3}\, \left(\sigma^S_{0,0}(t)\,a^4  + \int_0^a \loc\rho^E_{0,0}(x,t)\,x^4\,\md x \right)
\\
& \bb{D}(t) = 5\,a\,M_E\,\sum_m \big(\loc G^S_{2,m}(a,t)+\loc G^E_{2,m}(a,t)\big)\,\bb{Q}_m(t)
\end{align}
$$ (R:3)

Furthermore, it will be convinient to introduce the deviatoric inertia of the only surface load, without considering the induced deformation on the Earth, that, similar to eq. {eq}`R:3` reads

$$
\bb{D}^S(t) =  5\,a\,M_E\,\sum_m \loc G^S_{2,m}(a,t)\,\bb{Q}_m
$$ (R:4)

By considering that the geoid perturbation $\loc G^E$ can be expressed in terms of the surface load, $\loc G^S$, and centrifugal, $\loc G^C$, geoid perturbations multiplied by the gravitational loading, $k_n^L$, and tidal, $k_n^T$, Love numbers

$$
\loc G^E_{n,m}(a,t) = k_n^L(a)\,\loc G^S_{n,m}(a,t) + k_n^
T(a)\, \loc G^C_{n,m}(a,t) 
$$ (R:5)

and that

$$
5\,\sum_{m=-2}^2\,\bb{Q}_m\,\loc G^C_{2,m}(a,t) = \frac{\omega^2(t)}{g_0(a)}\,\bb{Q}(\vartheta_C(t),\varphi_C(t)) = \frac{\sb{\Omega}(t)}{g_0(a)}
$$ (R:6)

with

$$
\sb{\Omega}(t) = \boldsymbol{\omega}(t)\,\boldsymbol{\omega}(t)-\frac{\omega^2(t)}{3}\,\bb{1}
$$ (R:7)

the deviatoric inertia tensor can be further recast in the following compact expression

$$
\bb{D}(t) = \big(1+k_2^L(a)\big)\,\bb{D}_L(t) + k_2^T(a)\,\frac{a^5}{3\,G}\,\sb{\Omega}(t)
$$ (R:8)

It provides a straightforward physical interpretation of the deviatoric inertia tensor in terms of the surface loading deviatoric inertia and of the just defined tensor $\sb{\Omega}$ describing the Earth's rotation and of the role 

In the following we will neglect $\loc  I$ (this is correct within the present framework and under the assumption of incompressibility). Indeed, $\sigma_{0,0}^S=0$ because $\sigma^S$ describe a rearrangement of surface water and ice and, so, the total mass of the exange between water and ice must be zero at the Earth surface. Furthermore, $\loc  \rho^E_{0,0}=0$ is zero and there are no degree $n=0$ contributions from topography perturbations just because they would implies an expansion or contration of the radii of the internal interfaces and of the Earth's surface. In this respect, eq. {eq}`R:1` simplifies as follows

$$
\bb{I}(t) = I_0\,\one + \bb{D}(t)
$$ (R:9)

```

### Viscoelastic perturbations

```{div} full-width

Let us now generalize the deviatoric inertia tensor $\bb{D}$ given by eq. {eq}`R:8` to the viscoelastic case and write

$$
 \bb{D}(t) =  \int_{-\infty}^\infty (\delta(\tau)+k_2^L(a,\tau))\,\bb{D}_S(t-\tau)\,\md \tau  + \frac{a^5}{3\,G}\,\int_{-\infty}^\infty k_2^T(a,\tau)\,\sb{\Omega}(t-\tau)\,\md \tau
$$ (R:10)


where we have used the fact that

$$
\loc G^E_{n,m}(a,t) = \int \big(k_n^L(a,\tau)\,\loc G^S_{n,m}(a,t-\tau) + k_n^
T(a,\tau)\, \loc G^C_{n,m}(a,t-\tau)\big)\,\md \tau 
$$ (R:11)

In the following, we will omit the dependence on $a$ of the gravitational Love number.

```



```{warning} 
:class: full-width

Similar to the warning of {numref}`Section %s <sec-linear-viscoelastic>` - {ref}`sec-linear-viscoelastic`, it can be shown that the time convolution in eq. {eq}`R:10` must be understood in the following sense


$$
 \bb{D}(t) =  \left[\int_{-\infty}^\infty (\delta(\tau)+k_2^L(\tau))\,D_{ij}^S(t-\tau)\,\md \tau\,  + \frac{a^5}{3\,G}\,\int_{-\infty}^\infty k_2^T(\tau)\,\Omega_{ij}(t-\tau)\,\md \tau\right]\,\bb{e}_i(t)\,\bb{e}_j(t)
$$ (R:12)

where

$$
\sb{\Omega}(t) = \Omega_{ij}(t)\,\bb{e}_i(t)\,\bb{e}_j(t)
$$ (R:13)

with

$$
\Omega_{ij}(t) = \omega_i(t)\,\omega_j(t) - \frac{\omega^2(t)}{3}\,\delta_{ij}
$$ (R:14)


```


```{div} full-width

Then, we decompose the angular velocity into the initial angular velocity and its perturbation

$$
\boldsymbol{\omega}(t) =  \boldsymbol{\omega}_0(t) + \mat\boldsymbol{\omega}(t)
$$ (R:15)

with

$$
\boldsymbol{\omega}_0(t) = \omega_0\,\bb{e}_3(t) \qquad \mat\boldsymbol{\omega}(t) = \omega_0\,m_k(t)\,\bb{e}_k(t)
$$ (R:16)

with $m_k$ being the so-called direction cosines of the angular velocity perturbation $\mat\sb{\omega}$.

```

```{admonition} The initial angular velocity
:class: full-width, dropdown
    
Note that we, as far as there is no perturbation, the third rotating unit vector $\bb{e}_3$ does not vary through time


$$
\dercla{\bb{e}_3(t)} = \sb{\omega}_0(t)\times\bb{e}_3(t) = \omega_0\,\bb{e}_3(t)\times\bb{e}_3(t) = \zero
$$ (R:17)

and, so, the initial angular velocity, $\boldsymbol{\omega}_0$, can be seen as a vector constant in space in the inertial frame. When the angular velocity is perturbed, instead, the third rotating unit vector $\bb{e}_3$  will vary through time and, aware of this circumstance, we decide to define the perturbation of the angular velocity with respect to the time dependent vector $\sb{\omega}_0$. This choice, indeed, simplifies the formalism and make clearer the physical meaning of the equations that we are going to obtain.

```



```{admonition} Direction cosines
:class: full-width, dropdown

In order to provide a physical interpretation of the direction cosines, let us consider, at the first order, the angular rate

$$
\omega(t) = \sqrt{\sb{\omega}(t)\cdot\sb{\omega}(t)} \approx \omega_0\,\sqrt{1+2\,m_3(t)} \approx \omega_0\,(1+m_3)
$$

In this respect, $m_3$ represents the relative change of the angular rate

$$
m_3 = \frac{\omega(t)-\omega_0}{\omega_0}
$$

Let us now consider the rotation axis

$$
\bb{n}_C(t) = \frac{\sb{\omega}(t)}{\omega(t)} \approx \bb{e}_3(t) + m_1(t)\,\bn{e}_1(t)  + m_2(t)\,\bn{e}_2(t)
$$


```

```{div} full-width

Keep only first-order terms

$$
\sb{\Omega}(t) = \sb{\Omega}_0(t) + \mat\sb{\Omega}(t)
$$ (R:18)

with

$$
\begin{align}
& \sb{\Omega}_0(t) = \sb{\omega}_0(t)\,\sb{\omega}_0(t)-\frac{\omega_0^2}{3}\,\one
\\
& \mat\sb{\Omega}(t) = \boldsymbol{\omega}_0(t)\,\mat\boldsymbol{\omega}(t)+\mat\boldsymbol{\omega}(t)\,\boldsymbol{\omega}_0(t)-\frac{2\,\boldsymbol{\omega}_0(t)\cdot\delta\boldsymbol{\omega}(t)}{3}\,\bb{1}
\end{align}
$$ (R:19)

eq. {eq}`R:10` becomes

$$
\begin{align}
\bb{D}(t) &= \int_{-\infty}^\infty (\delta(\tau)+k_2^L(\tau))\,\bb{D}^S(t-\tau)\,\md \tau
+\frac{a^5}{3\,G}\,\left[\left(\int_{-\infty}^\infty k_2^T(\tau)\,\md \tau\right)\,\sb{\Omega}_0(t) 
+ \int_{-\infty}^\infty k_2^T(\tau)\,\mat\sb{\Omega}(t-\tau)\,\md \tau\right]
\end{align}
$$ (R:20)

Then, we define the fluid tidal Love number

$$
k_2^{T,F} = \int_{-\infty}^\infty k_2^T(\tau)\,\md \tau  = \lim_{t\rightarrow \infty} \int_{-\infty}^t k_2^T(\tau)\,\md \tau  = \lim_{t\rightarrow \infty} \int_{-\infty}^\infty k_2^T(\tau)\,H(t-\tau)\,\md \tau 
$$ (R:21)

and recast eq. {eq}`R:20` as follows

$$
\begin{align}
\bb{D}(t) &= \int_{-\infty}^\infty (\delta(\tau)+k_2^L(\tau))\,\bb{D}^S(t-\tau)\,\md \tau
\\
&+ \frac{a^5\,\omega_0^2}{3\,G}\,\left[\big(k_2^{T,F}+2\,M_3(t)\big)\,\left(\bb{e}_3(t)\,\bb{e}_3(t)-\tfrac{1}{3}\,\bb{1}\right) 
+\sum_{k=1}^2\,M_k(t)\,\big(\bb{e}_3(t)\,\bb{e}_k(t)+\bb{e}_k(t)\,\bb{e}_3(t)\big)\right]
\end{align}
$$ (R:22)

with

$$
M_k(t) = (k_2^T\star m_k)(t) = \int_{-\infty}^\infty k_2^T(\tau)\,m_k(t-\tau)\,\md \tau
$$ (R:22a)

Defining the equatorial, $A$, and polar, $C$, inertia moments and their difference as folos

$$
A = I_0 - \frac{1}{3}\,\frac{a^5\,\omega_0^2}{3\,G}\,k_2^{T,F}
\qquad
C = I_0 + \frac{2}{3}\,\frac{a^5\,\omega_0^2}{3\,G}\,k_2^{T,F}
\qquad
C-A = \frac{a^5\,\omega_0^2}{3\,G}\,k_2^{T,F}
$$ (R:23)

eq. {eq}`R:22` becomes


$$
\begin{align}
\bb{D}(t) &=  \int_{-\infty}^\infty (\delta(\tau)+k_2^L(\tau))\,\bb{D}^S(t-\tau)\,\md \tau
\\
&+(C-A)\,\left[\left(1+\frac{2\,M_3(t)}{k_2^{T,F}}\right)\,\left(\bb{e}_3(t)\,\bb{e}_3(t)-\tfrac{1}{3}\,\bb{1}\right) 
+\sum_{k=1}^2\frac{M_k(t)}{k_2^{T,F}}\,\big(\bb{e}_3(t)\,\bb{e}_k(t)+\bb{e}_k(t)\,\bb{e}_3(t)\big)\right]
\end{align}
$$ (R:24)

```

```{admonition} Axially symmetric surface loads
:class: full-width, dropdown

Let us consider a surface load density distribution as in eq. {eq}`M:61` and the associated geoid perturbation in terms of its spherical harmonic coefficients

$$
\loc G_{n,m}^S = \frac{C\,\sigma^S_{n,m}\,a}{(2\,n+1)\,g_0(a)}
$$ (R:25)

In view of eq. {eq}`R:4`, the deviatoric inertia tensor due to the surface load becomes

$$
\bb{D}^S(t) =  \frac{C\,a\,M_E}{g_0(a)}\,\sum_m \sigma_{2,m}^S\,\bb{Q}_m(t) =  4\,\pi\,a^3\,\sum_m \sigma_{2,m}^S\,\bb{Q}_m(t)
$$ (R:26)

Let us now further assume that $\sigma^S$ is axially symmetric around $(\vartheta_0,\varphi_0)$

$$
\sigma^S(\vartheta,\varphi) = \sigma^S(\cos\gamma) = \sum \sigma_n\,P_n(\cos\gamma)
$$ (R:27)

with

$$
\cos\gamma = \cos\vartheta_0\,\cos\vartheta+\sin\vartheta_0\,\sin\vartheta\,\cos(\varphi-\varphi_0)
$$ (R:28)

Making use of the spherical harmonic addition theorem, we thus obtain that

$$
\sigma_{n,m}^S = \frac{\sigma_n}{2\,n+1}\,Y_n^m(\vartheta_0,\varphi_0)
$$ (R:29)

and eq. {eq}`R:28` becomes

$$
\bb{D}^S(t) =  \frac{4\,\pi\,a^3}{5}\,\sigma_2^S\,\sum_m \bb{Q}_m(t)\,,Y_2^m(\vartheta_0,\varphi_0) = \frac{4\,\pi\,a^3}{5}\,\sigma_2^S\,\bb{Q}(\vartheta_0,\varphi_0)
$$ (R:30)

where we have made use of eq. {eq}`G:32`. We note that, in matrix notation, we have

$$
\bb{Q}(\vartheta_0,\varphi_0) = \frac{1}{3}\,\one - \bn{r}(\vartheta_0,\varphi_0)\,\bn{r}(\vartheta_0,\varphi_0) = \left(\begin{matrix}
\tfrac{1}{3}-\sin^2\vartheta_0\,\cos^2\varphi_0 & - \tfrac{1}{2}\,\sin^2\vartheta_0\,\sin(2\,\varphi_0) & -\tfrac{1}{2}\,\sin(2\,\vartheta_0)\,\cos\varphi_0 \\
- \tfrac{1}{2}\,\sin^2\vartheta_0\,\sin(2\,\varphi_0)  & \tfrac{1}{3}-\sin^2\vartheta_0\,\cos^2\varphi_0 & -\tfrac{1}{2}\,\sin(2\,\vartheta_0)\,\sin\varphi_0 \\ -\tfrac{1}{2}\,\sin(2\,\vartheta_0)\,\cos\varphi_0 & -\tfrac{1}{2}\,\sin(2\,\vartheta_0)\,\sin\varphi_0 & \cos^2\vartheta_0
\end{matrix}\right)
$$ (R:31)

and that the direction $\bn{x}_\min$ of minimum inertia $I_\min$ is

$$
\bn{x}_\min = \bn{r}(\vartheta_0,\varphi_0) \qquad I_\min = -\frac{2}{3}\,\frac{4\,\pi\,a^3}{5}\,\sigma_2^S
$$ (R:32)

while the other two perpendicular directions are of maximum inertia $I_\max$

$$
I_\max = \frac{1}{3}\,\frac{4\,\pi\,a^3}{5}\,\sigma_2^S
$$ (R:33)

In order to further fix the ideas, we can also provide an analytical expression for $\sigma_2^S$ assuming that the surface load is an axially symetric ice shield of maximum height $H_0$ and angular amplitude $\alpha$ (see {numref}`Section %s <sec-shield>` - {ref}`sec-shield`). We thus write

$$
\sigma_2^S = \rho_\mathrm{ice}\,H_0\,h_2 
$$ (R:34)

with

$$
h_2 = \frac{5}{16}\,\frac{(4\,\cos^2\alpha-1)\,P_2(\cos\alpha)-2\,\cos\alpha\,P_3(\cos\alpha)}{1-\cos\alpha} > 0
$$ (R:35)
```

## Rotational Euler equation

```{div} full-width

Let us recall the rotational Euler equation {eq}`E:40` that we report here for the convinience of the reader

$$
\derpar{\bb{H}_E(t)} + \sb{\omega}(t)\times\bb{H}_E(t) = \bb{k}^A(t)
$$ (R:36)

with $\bb{k}^A$ being the relative torque from the athmosphere on the solid Earth and $\bb{H}_E$ being the solid Earth (and loads) angular momentum

$$
\bb{H}_E(t) = \bb{I}(t)\cdot\sb{\omega}(t)
$$ (R:37)

```


```{note}
Hereinafter, for the sake of brevity, we will note explicitate the time dependence of the rotating unit vectors $\bb{e}_i$ and those coming from them.
```


```{div} full-width

According to the results of the previous section, the inertia tensor of the solid Earth (and load) can be decomposed as follows

$$
\bb{I}(t) = \bb{I}_0+\loc\bb{I}(t)
$$ (R:38)

with

$$
\begin{align}
\bb{I}_0 &= I_0\,\bb{1}  + (C-A)\,\left(\bb{e}_3\,\bb{e}_3-\tfrac{1}{3}\,\bb{1}\right) 
= A\,\big(\bb{e}_1\,\bb{e}_1+\bb{e}_2\,\bb{e}_2\big)+C\,\bb{e}_3\,\bb{e}_3 = \left(\begin{matrix} A & 0 & 0 \\ 0 & A & 0 \\ 0 & 0 & C \end{matrix}\right)
\end{align}
$$ (R:39)

and

$$
\begin{align}
\loc\bb{I}(t) &=  \int_{-\infty}^\infty (\delta(\tau)+k_2^L(\tau))\,\bb{D}^{S}(t-\tau)\,\md \tau
\\
& + \frac{C-A}{k_2^{T,F}}\,\left[\left(2\,\bb{e}_3\,\bb{e}_3-\tfrac{2}{3}\,\bb{1}\right) \,M_3(t)
+ \sum_{k=1}^2\big(\bb{e}_3\,\bb{e}_k+\bb{e}_k\,\bb{e}_3\big)\,M_k(t)\right]
\end{align}
$$ (R:40)

Then, keeping only first-order terms,  the angular momentum of solid Earth (and loads) gives

$$
\begin{align}
\bb{H}_E(t) &= \bb{I}(t)\cdot\boldsymbol{\omega}(t) = \bb{I}_0\cdot\boldsymbol{\omega}_0 + \bb{I}_0\cdot\mat\boldsymbol{\omega}(t) + \loc\bb{I}(t)\cdot\boldsymbol{\omega}_0
\\
&=\omega_0\,\left(m_1(t)\,A\,\bb{e}_1+m_2(t)\,A\,\bb{e}_2+(1+m_3(t))\,C\,\bb{e}_3+\sum_{j=1}^3\loc I_{j3}(t)\,\bb{e}_j\right)
\end{align} 
$$ (R:39)

and its cross product wth the angular velocity vector

$$
\begin{align}
\boldsymbol{\omega}\times\bb{H} &= \omega_0^2\,\left[\bb{e}_3\times\left(m_1(t)\,A\,\bb{e}_1+m_2(t)\,A\,\bb{e}_2+\sum_{j=1}^2 \loc I_{j3}(t)\,\bb{e}_j\right)+\sum_{j=1}^2 m_j(t)\,C\,\bb{e}_j\times\bb{e}_3\right]
\\
&= \omega_0^2\,\Big[\big(\loc I_{13}(t)-(C-A)\,m_1(t)\big)\,\bb{e}_2-\big(\loc I_{23}(t)-(C-A)\,m_2(t)\big)\,\bb{e}_1\Big]
\end{align}
$$ (R:40)

In light of this, the Euler's rotational equation {eq}`R:36` becomes

$$
\omega_0^2\,\Big[\big(\loc I_{13}(t)-(C-A)\,m_1(t)\big)\,\bb{e}_2-\big(\loc I_{23}(t)-(C-A)\,m_2(t)\big)\,\bb{e}_1\Big] +  \omega_0\,\left(\dot{m}_1(t)\,A\,\bb{e}_1+\dot{m}_2(t)\,A\,\bb{e}_2+\dot{m}_3(t)\,C\,\bb{e}_3+\sum_{j=1}^3\loc \dot{I}_{j3}(t)\,\bb{e}_j\right) = k^A_j(t)\,\bb{e}_j
$$ (R:42)

or equivalently,

$$
\begin{align}
& -\omega_0^2\,\big(\loc  I_{23}(t)-(C-A)\,m_2(t)\big) +  \omega_0\,\big(\dot{m}_1(t)\,A+ \loc  \dot{I}_{13}(t)\big) = k^A_1 (t)
\\
& \omega_0^2\,\big(\loc  I_{13}(t)-(C-A)\,m_1(t)\big) +  \omega_0\,\big(\dot{m}_2(t)\,A+ \loc  \dot{I}_{23}(t)\big) = k_2^A(t)
\\
& \omega_0\,\big(\dot{m}_3(t)\,C + \loc  \dot{I}_{33}(t)\big) = k_3^A(t)
\end{align}
$$ (R:43)

with the dot $\cdot$ standing for the time derivative.

```

### Length of the day (LOD)

```{div} full-width

Within the first-order perturbation theory, the third component of eq. {eq}`R:43` is decoupled from the first two components. It thus can be solved separately and, after time integration (from $0^-$ to $t$), we obtain

$$
\omega_0\,\big(C\,m_3(t) + \loc I_{33}(t)\big) = \int_{0^-}^t\,k_3^A(\tau)\,\md \tau
$$

where we have assume that $m_(t<0)=0$ and $\loc I_{33}(t<0)=0$. 

By considering that

$$
\begin{align}
\loc I_{33}(t) &=  \int_{-\infty}^\infty (\delta(\tau)+k_2^L(\tau))\,D_{33}^{S}(t-\tau)\,\md \tau + \frac{4\,(C-A)}{3\,k_2^{T,F}}\,M_3(t)
\end{align}
$$ (R:40)

after Laplace transform, the third component of eq. {eq}`R:43` gives

$$
s\,\left(C+\frac{4\,(C-A)}{3\,k_2^{T,F}}\,\tilde{k}_2^T(s)\right)\,\tilde{m}(s) + s\,\big(1+\tilde{k}_2^L(s)\big)\,\tilde{D}_{33}^S(s) = \tilde{k}^A(s)
$$

and, so, $\tilde{m}_3$ can be obtained as follows

$$
\tilde{m}(s) = \frac{\tilde{k}^A(s)-s\,\big(1+\tilde{k}_2^L(s)\big)\,\tilde{D}_{33}^S(s) }{s\,\left(C+\frac{4\,(C-A)}{3\,k_2^{T,F}}\,\tilde{k}_2^T(s)\right)}
$$

```

### Polar motion

```{div} full-width

In order to solve the first two components of eq. {eq}`R:36`, let us define

$$
m(t) = m_1(t) + i\,m_2(t) \qquad J = \frac{\loc  I_{13}(t) + i\,\loc  I_{23}(t)}{C-A} \qquad \psi(t) =\frac{k^A_1(t)+i\,k^A_2(t)}{\omega_0^2\,(C-A)}
$$ (R:44)

and combine the two components in order to obtain the following complex differential equation

$$
\begin{align}
&  \left(\frac{\dot{m}(t)}{\sigma_e}+ \frac{\dot{J}(t)}{\omega_0}\right) + i\,\big(J(t)-m(t)\big)  = \psi(t)
\end{align}
$$ (R:45)

with $\sigma_e$ being the Euler frequency

$$
\sigma_e = \omega_0\,\frac{C-A}{A}
$$ (R:46)

```



```{admonition} Euler frequency
:class: full-width, dropdown

By considering that

$$
\frac{C-A}{A} = 0.0035
$$ (R:46a)

the Euler frequency and period yields

$$
\sigma_e = 2.5767\times 10^{-7}\,\mathrm{s}^{-1} \qquad T_e = \frac{2\,\pi}{\sigma_e} = 0.7727\,\mathrm{yr}
$$ (R:46b)


```



```{div} full-width

In the perspective of further specifying eq. {R:45}, we consider that

$$
\begin{align}
\loc I_{k3}(t) &=  \int_{-\infty}^\infty (\delta(\tau)+k_2^L(\tau))\,D_{k3}^{S}(t-\tau)\,\md \tau + \frac{C-A}{k_2^{T,F}}\,M_k(t)
\end{align}
$$ (R:47)

and define

$$
J^S(t) = \frac{D_{13}^S(t) + i\,D_{23}^S(t)}{C-A} \qquad M(t) = M_1(t) + i\,M_2(t) = \int_\infty^\infty k_2^T(\tau)\,m(t-\tau)\,\md \tau
$$ (R:48)

In this way, $J$ gives

$$
\begin{align}
J(t) &=  \int_{-\infty}^\infty (\delta(\tau)+k_2^L(\tau))\,J^S(t-\tau)\,\md \tau + \frac{M(t)}{k_2^{T,F}}
\end{align}
$$ (R:49)

and, making use of it and after Laplace transform, eq. {eq}`R:45` becomes

$$
\begin{align}
&  s\,\left(\frac{\tilde{m}(s)}{\sigma_e}+ \frac{\tilde{J}(s)}{\omega_0}\right) + i\,\big(\tilde{J}(s)-\tilde{m}(s)\big)  = \tilde{\psi}(s)
\end{align}
$$ (R:50)

with

$$
\tilde{J}(s)  = \big(1+\tilde{k}_2^L(s)\big)\,\tilde{J}^S(s) + \frac{\tilde{k}_2^T(s)}{k_2^{T,F}}\,\tilde{m}(s)
$$ (R:51)

Before considering the solution for $\tilde{m}$ and discuss effective strategy for evaluating its inverse Laplace transform to get the solution in the Laplace domain, we shall get some physical insight into eq. {eq}`R:45` under a few simplyfing assumptions.

```

## Euler and Chandler wobbles

### Rigid Earth and Euler wobble

```{div} full-width

Let us consider the case of a rigid solid Earth for which the gravitational viscoelastic Love numbers are zero

$$
k_2^{X}(t) = 0
$$ (R:52a)

Within this assumption, making use of eq. {eq}`R:52a` into eq. {eq}`R:49`, we obtain

$$
J(t) = J^S(t)
$$ (R:52b)


Let us now set to zero the athmospheric torque, $\psi=0$, and prescribe the surface load inertia tensor perturbation with a step-like time history

$$
J^S(t) = \bar{J}^S\,H(t)
$$ (R:52)

Within these assumptions, eq. {eq}`R:45` becomes

$$
\begin{align}
&  \frac{\dot{m}(t)}{\sigma_e}+ \frac{\bar{J}^S}{\omega_0}\,\delta(t) + i\,\big(\bar{J}^S\,H(t) -m(t)\big)  = 0
\end{align}
$$ (R:53)

In order to solve it, we assume that

$$
m(t) = \big(f_0 + f(t)\big)\,H(t) \qquad \dot{m}(t) = f_0\,\delta(t) + \dot{f}(t)\,H(t)
$$ (R:54)

with $f(0)=0$.  After substitution of eq. {eq}`R:54` into eq. {eq}`R:53`, we obtain

$$
\begin{align}
&  \left(\frac{f_0}{\sigma_e}+\frac{\bar{J}^S}{\omega_0}\right)\,\delta(t) + \left[\frac{\dot{f}(t)}{\sigma_e} + i\,\big(\bar{J}^S -f_0-f(t)\big) \right]\,H(t) = 0
\end{align}
$$ (R:55)

In light of this, we first obtain

$$
f_0 = \frac{\sigma_e}{\omega_0}\,\bar{J}^S  
$$ (R:56)

and the following first-order differential equation

$$
f(t) + i\,\frac{\dot{f}(t)}{\sigma_e} = \bar{J}^S - f_0 = \bar{J}^S\,\left(1-\frac{\sigma_e}{\omega_0}\right)\qquad\mathrm{with}\qquad f(0) = 0
$$ (R:57)

that is solve by

$$
f(t) = \bar{J}^S\,\left(1-\frac{\sigma_e}{\omega_0}\right)\,\left(1-e^{i\,\sigma_e\,t}\right)
$$ (R:58)

From eqs. {eq}`R:54`, we thus obtain

$$
\begin{align}
m(t) &= \bar{J}^S\,\left[ \frac{\sigma_e}{\omega_0}  +  \left(1-\frac{\sigma_e}{\omega_0}\right)\,\left(1-e^{i\,\sigma_e\,t}\right)\right]\,H(t) =
\\
&=\bar{J}^S\,\left[ \frac{\sigma_e}{\omega_0}\,e^{i\,\sigma_e\,t}  +  \left(1-e^{i\,\sigma_e\,t}\right)\right]\,H(t)
\end{align}
$$ (R:59)

Let us now implement and plot eq. {eq}`R:59` (blue line, `ms`) and distinguish the contributions from the first (orange line, `ms_jump`) and second (green line, `ms_main`) terms in the square brackets over an Euler period

```

````python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

JS = 0.01+0.005j     
x = 0.0035          # (C-A)/A

day = 24*3600
w = 2*np.pi/day     # angular rate
we = x*w            # Euler angular frequency
Te = 2*np.pi/we     # Euler period


ts = np.linspace(0,Te,10001)

ms_main = JS*(1-np.exp(1j*we*ts))
ms_jump = JS*(-x*np.exp(1j*we*ts))
ms  = ms_main + ms_jump


import matplotlib.pyplot as plt
fig,axes = plt.subplots(1,2,tight_layout=True,figsize=(8,4))

for ax in axes:
    ax.set_aspect("equal")
    ax.scatter(JS.real,JS.imag,marker=".")
    ax.plot(ms.real,ms.imag)
    ax.plot(ms_jump.real,ms_jump.imag)
    ax.plot(ms_main.real,ms_main.imag)
    ax.axhline(0,color="b",linewidth=0.5)
    ax.axvline(0,color="b",linewidth=0.5)
    ax.set_xlabel("$m_1$")
    ax.set_ylabel("$m_2$")
axes[-1].set_xlim(-0.0007,0.0007)
axes[-1].set_ylim(-0.0007,0.0007)
axes[-1].set_xticks(np.linspace(-0.0006,0.0006,5))
axes[-1].set_yticks(np.linspace(-0.0006,0.0006,5));
````

```{div} full-width

We note that the first term in the square brackets is very small compared to the second one and, so, it can be neglected. This negligible term comes from the time derivative of $J$ in eq. {eq}`R:45` and, so, we can neglect it. From now, we thus decide to simplify  eq. {eq}`R:` as follows

$$
\begin{align}
&  \frac{\dot{m}(t)}{\sigma_e} + i\,\big(J(t)-m(t)\big)  = \psi(t)
\end{align}
$$ (R:60)

of which the solution reads

$$
\begin{align}
m(t) &=\bar{J}\, \left(1-e^{i\,\sigma_e\,t}\right)\,H(t)
\end{align}
$$ (R:61)
```

### Elastic Earth and Chandler wobble

```{div} full-width

Let us now assume that the perturbation of the solid Earth are elastic. This means that the we are assuming that the viscoelastic Love numbers are simpy given by

$$
k_2^X(t) = k_2^{X,E}\,\delta(t)
$$ (R:62)

Within this assumption, eq. {eq}`R:22a` gives

$$
M_k = k_2^{X,E}\,m_k(t)
$$ (R:63)

and eq. {eq}`R:49` becomes

$$
J(t) = q_L^E\,J^S(t) + (1-q_T^E)\,m(t)
$$ (R:64)

Here, for the sake of brevity, we have defined

$$
q_L^E = 1+k_2^{L,E} = 0.5903 \qquad q_T^E = 1-\frac{k_2^{T,E}}{k_2^{T,F}} = 0.5265
$$ (R:65)

where the numerical values comes from the three-layered incompressible Earth model.

Then, the simplified rotational Euler equation {eq}`R:60` becomes

$$
\begin{align}
&  \frac{\dot{m}(t)}{\sigma_e} + i\,\big(q^E_L\,J^S(t)-q_T^E\,m(t)\big)= \psi(t) 
\end{align}
$$ (R:66)

which can be recast as follows

$$
\begin{align}
&  \frac{\dot{m}(t)}{\sigma_c} + i\,\left(\frac{q^E_L}{q_T^E}\,J^S(t)-m(t)\right)= \frac{\psi(t)}{q_T^E}
\end{align}
$$ (R:67)

in terms of the Chandler frequency

$$
\sigma_c = \sigma_e\,q_T^E = 1.3566\times 10^{-7}\,\mathrm{s}^{-1} \qquad T_c = \frac{2\,\pi}{\sigma_c} = \frac{T_e}{q_T^E} = 1.8994\,T_e = 1.4676\,\mathrm{yr}
$$ (R:68)

Within the same assumption made in the previous section

$$
J^S(t) = \bar{J}^S\,H(t) \qquad \psi(t) = 0
$$ (R:69)

we thus obtain

$$
\begin{align}
&  \frac{\dot{m}(t)}{\sigma_c} + i\,\left(\frac{q^E_L}{q_T^E}\,\bar{J}^S\,H(t)-m(t)\right)= 0\end{align}
$$ (R:70)

of which the solution reads

$$
\begin{align}
m(t) &=\frac{q_L^E}{q_T^E}\,\bar{J}^S\, \left(1-e^{i\,\sigma_c\,t}\right)\,H(t)
\end{align}
$$ (R:71)

In this respect, we obtain a wobble of frequency $\sigma_c$, the so-called Chandler wobble, and of amplitude $q_L^E/q_T^E\,\bar{J}^S$, with $q_L^E/q_T^E=1.1211$. 

```

## Rotational dynamics of a viscoelastic Earth

By defining

$$
q_L(t) = \delta(t)+k_2^{L}(t) \qquad q_T(t) = 1-\frac{k_2^{T}(t)}{k_2^{T,F}}
$$ (R:72)

we can write $J$ given by eq. {eq}`R:49` as follows

$$
J(t) = \int_{-\infty}^\infty \Big[q_L(\tau)\, J^S\big)(t-\tau) + q_T(\tau)\,m(t-\tau)\Big]\,\md \tau \qquad \tilde{J}(s) = \tilde{q}_L(s)\,\tilde{J}(s) + \tilde{q}_T(s)\,\tilde{m}(s)
$$ (R:73)


After Laplace transform, the simplified rotational Euler equation {eq}`R:60` becomes

$$
\begin{align}
&  \tilde{m}(s)\,\left(i\,\frac{s}{\sigma_e} + \tilde{q}_T(s)\right)= \tilde{q}^L(s)\,\tilde{J}^S(s)
\end{align}
$$  (R:74)


By recasting $\tilde{q}_T$ and $\tilde{q}_L$ as follows

$$
\begin{align}
\tilde{q}_T(s) &= \frac{k_2^{T,F} - \tilde{k}_2^T(s)}{k_2^{T,F}} = -\frac{1}{k_2^{T,F}}\,\sum \left(\frac{k_{2,j}^T}{s_j}+\frac{k_{2,j}^T}{s-s_{2,j}}\right) = -\frac{s}{k_2^{T,F}}\,\sum_j\frac{k_{2,j}^T}{s_{2,j}\,(s-s_{2,j})} = i\,\frac{s}{\sigma_e}\,\sum_j\frac{x_{j}^T}{s-s_{2,j}}  
\\
\tilde{q}_L(s) &= 1 + \tilde{k}_2^L(s) = \tilde{k}_2^L(s) - k_2^{L,F} = \sum \left(\frac{k_{2,j}^L}{s_j}+\frac{k_{2,j}^L}{s-s_{2,j}}\right) = -s\,\sum_j\frac{k_{2,j}^L}{s_{2,j}\,(s-s_{2,j})} = i\,\frac{s}{\sigma_e}\,\sum_j\frac{x_{j}^L}{s-s_{2,j}}  
\end{align}
$$ (R:75)

where we used the fact that $k_2^{L,F}= -1$ and defined

$$
x_j^T =i\,\frac{\sigma_e}{k_2^{T,F}}\,\frac{k_{2,j}^T}{s_{2,j}}
\qquad
x_j^L = i\,\sigma_e\,\frac{k_{2,j}^L}{s_{2,j}}
$$ (R:76)

eq. {eq}`R:74` becomes


$$
\begin{align}
&  \frac{i\,s}{\sigma}_e\,\tilde{m}(s)\,\left(1 + \sum_j\frac{x_{j}^T}{s-s_{2,j}}  \right)= i\,\frac{s}{\sigma_e}\,\sum_j\frac{x_{j}^L}{s-s_{2,j}}  \,\tilde{J}^S(s)
\end{align}
$$  (R:77)

In light of this, $m$ can be obtained as follows

$$
\begin{align}
&  \tilde{m}(s) = \tilde{X}(s)\,\tilde{J}^S(s) \qquad m(t) = \int_{-\infty}^\infty X(\tau)\,J^S(t-\tau)\,\md \tau
\end{align}
$$ (R:78)

Here, $\tilde{X}(s)$ is given by

$$
\begin{align}
\tilde{X}(s)  &= \frac{ \sum_j\frac{x_{j}^L}{s-s_{2,j}} }{1 + \sum_j\frac{x_{j}^T}{s-s_{2,j}} }
= \frac{ \sum_j x_{j}^L\,\prod_{k\neq j} (s-s_{2,k}) }{\prod_k (s-s_{2,k}) + \sum_j x_{j}^T \prod_{k\neq j} (s-s_{2,k}) } = \frac{ \sum_j x_{j}^L\,\prod_{k\neq j} (s-s_{2,k}) }{\prod_k (s-a_{k}) }
\end{align}
$$  (R:79)

where $a_k$ are the $N$ roots of the polynomial at the denominator in the second identity

$$
\prod_k (s-s_{2,k}) + \sum_j x_{j}^T \prod_{k\neq j} (s-s_{2,k}) = \prod_k (s-a_{k}) 
$$ (R:80)

In view of eq. {eq}`R:79`, the function $\tilde{X}$ can be seen as the summation of $N$ first-order poles

$$
\tilde{X}(s) = \sum_q \frac{X_q}{s-a_q}  \qquad X(t) = H(t)\,\sum_q X_q\,e^{a_q\,t}
$$ (R:81)

with $X_q$ being the residue associated to the $q$-th pole

$$
X_q = \lim_{s\rightarrow a_q} \tilde{X}(s) = \frac{ \sum_j x_{j}^L\,\prod_{k\neq j} (a_q-s_{2,k}) }{\prod_{k\neq q} (a_q-a_{k}) }
$$ (R:82)

In light of this, once determined  the poles $s_{2,j}$ and the residues of the load, $k_{2,j}^L$, and tidal, $k_{2,j}^T$, gravitational Love numbers, we can first obtain the poles $a_q$ from eq. {eq}`R:80` and then the residues $X_q$ from eq. {eq}`R:82`.
