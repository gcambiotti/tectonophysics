## Spherical harmonic vectors

```{div} full-width

Starting from the spherical harmonics, we can define the three spherical harmonics vectors 

$$
\begin{align}
& \mathbf{P}_n^m(\vartheta,\varphi) = \hat{\mathbf{r}}\,Y_n^m(\vartheta,\varphi)
\\
& \mathbf{B}_n^m(\vartheta,\varphi) = \frac{x}{\gamma_n}\,\boldsymbol{\nabla} Y_n^m(\vartheta,\varphi) = \frac{1}{\gamma_n}\,\left(\hat{\boldsymbol{\vartheta}}\,\frac{\partial Y_n^m(\vartheta,\varphi)}{\partial \theta} + \hat{\boldsymbol{\varphi}}\,\csc\vartheta\,\frac{\partial Y_n^m(\vartheta,\varphi)}{\partial \varphi}\right)
\\
& \mathbf{C}_n^m(\vartheta,\varphi) = \mathbf{B}_n^m(\vartheta,\varphi)\times\hat{\mathbf{r}} = \frac{\boldsymbol{\nabla}\times(x\,\mathbf{P}_n^m(\vartheta,\varphi))}{\gamma_n}  = \frac{1}{\gamma_n}\,\left(\hat{\boldsymbol{\vartheta}}\,\csc\vartheta\,\frac{\partial Y_n^m(\vartheta,\varphi)}{\partial \varphi} - \hat{\boldsymbol{\varphi}}\,\frac{\partial Y_n^m(\vartheta,\varphi)}{\partial \vartheta}\right)
\end{align}
$$ (V:1)


with 

$$
\gamma_n=\sqrt{n\,(n+1)}
$$ (V:2)

Similar to the spherical harmonics, which can be used for representing scalar functions over the sphere, the spherical harmonic vectors can be used for representing vector fields 

$$
\begin{align}
& \mathbf{f}(\vartheta,\varphi) = \sum_{n=0}^\infty\sum_{m=-n}^n \big(f_{n,m}^P\,\mathbf{P}_{n}^m(\vartheta,\varphi)+ f_{n,m}^B\,\mathbf{B}_{n}^m(\vartheta,\varphi)+f_{n,m}^C\,\mathbf{C}_{n}^m(\vartheta,\varphi)\big)
\end{align}
$$ (V:3)

where $f_{n,m}^X$ (with $X=P,B$ or $C$) are the spherical harmonic coefficients. We note that $\mathbf{B}_{0,0}=\mathbf{C}_{0,0}=\mathbf{0}$ because $Y_0^0=1$ does not depend on the angular coordinates $(\vartheta,\varphi)$. In this respect, the spherical harmonic coefficients $f_{0,0}^B$ and $f_{0,0}^C$ have no meaning and, by convenction, we set them to zero

$$
\begin{align}
& f_{0,0}^B = f_{0,0}^C = 0
\end{align}
$$ (V:4)

Like the spherical harmonics, the spherical harmonic vectors are orthogonal in the following senses

$$
\int_{\mathcal{\Omega}} \mathbf{P}_n^m(\vartheta,\varphi)\cdot\mathbf{P}_p^q(\vartheta,\varphi)\,\mathrm{d} \Omega = 4\,\pi\,\delta_{n,p}\,\delta_{m,q}
\qquad
\int_{\mathcal{\Omega}} \mathbf{B}_n^m(\vartheta,\varphi)\cdot\mathbf{B}_p^q(\vartheta,\varphi)\,\mathrm{d} \Omega = 4\,\pi\,\delta_{n,p}\,\delta_{m,q}
\qquad
\int_{\mathcal{\Omega}} \mathbf{C}_n^m(\vartheta,\varphi)\cdot\mathbf{C}_p^q(\vartheta,\varphi)\,\mathrm{d} \Omega = 4\,\pi\,\delta_{n,p}\,\delta_{m,q}
$$ (V:5)

$$
\mathbf{P}_n^m(\vartheta,\varphi)\cdot\mathbf{B}_p^q(\vartheta,\varphi) = 0
\qquad
\mathbf{P}_n^m(\vartheta,\varphi)\cdot\mathbf{C}_p^q(\vartheta,\varphi) = 0
\qquad
\int_{\mathcal{\Omega}} \mathbf{B}_n^m(\vartheta,\varphi)\cdot\mathbf{C}_p^q(\vartheta,\varphi)\,\mathrm{d} \Omega = 0
$$ (V:6)

and, so, we can obtain the spherical harmonic coefficients of a vector field as follows

$$
f_{n,m}^X = \frac{1}{4\,\pi}\int_{\mathcal{\Omega}} \mathbf{f}(\vartheta,\varphi)\cdot\mathbf{X}_n^m(\vartheta,\varphi)\,\mathrm{d} \Omega
$$ (V:7)

```

```{admonition} A useful formalism
:class: full-width, dropdown

By noting that 

$$
\frac{\partial \mathrm{tri}_{m}(\varphi)}{\partial \varphi}  = m\,\mathrm{tri}_{-m}(\varphi)
= \begin{cases} m\,\sin(-m\,\varphi) = -m\,\sin(m\,\varphi)  & m < 0 \\
m\,\cos(-m\,\varphi) = m\,\cos(m\,\varphi) & m\geq 0
\end{cases}
$$ (V:8)

with $\mathrm{tri}_m$ given by eq. {eq}`P:29`, the derivative of the spherical harmonics with respect to the longitude become

$$
\frac{\partial Y_n^m(\vartheta,\varphi)}{\partial \varphi} = m\,P_n^{\mid m \mid}(\cos\theta)\,\mathrm{tri}_{-m}(\varphi) = m\,Y_n^{-m}(\vartheta,\varphi) 
$$ (V:9)

In this respect, the spherical harmonic vectors $\bb{B}_n^m$ and $\bb{C}_n^m$ yields

$$
\begin{align}
& \mathbf{B}_n^m(\vartheta,\varphi) =  \frac{1}{\gamma_n}\,\left(\hat{\boldsymbol{\vartheta}}\,\frac{\partial Y_n^m(\vartheta,\varphi)}{\partial \theta} + m\,\hat{\boldsymbol{\varphi}}\,\csc\vartheta\,Y_n^{-m}(\vartheta,\varphi)\right)
\\
& \mathbf{C}_n^m(\vartheta,\varphi) = \frac{1}{\gamma_n}\,\left(\hat{\boldsymbol{\vartheta}}\,\csc\vartheta\,m\,Y_n^{-m}(\vartheta,\varphi) - \hat{\boldsymbol{\varphi}}\,\frac{\partial Y_n^m(\vartheta,\varphi)}{\partial \vartheta}\right)
\end{align}
$$ (V:10)

```

### Basic properties

```{div} full-width

In the following we shall consider the following expressions for the divergence and the curl of the spherical harmonic vectors multiplied by a function $f$ that depends only on the radial distance $x$ yield

$$
\begin{align}
&\boldsymbol{\nabla}\cdot\big(f(x)\,\mathbf{P}_n^m(\vartheta,\varphi)\big)= \left(\dot{f}(x)+\frac{2}{x}\,f(x)\right)\,Y_n^m(\vartheta,\varphi) 
\\
& \boldsymbol{\nabla}\cdot\big(f(x)\,\mathbf{B}_n^m(\vartheta,\varphi)\big) = -\frac{f(x)}{x}\,\gamma_n\,Y_n^m(\vartheta,\varphi) 
\\
&\boldsymbol{\nabla}\cdot\big(f(x)\,\mathbf{C}_n^m(\vartheta,\varphi)\big) = 0
\end{align}
$$ (V:11)

and

$$
\begin{align}
& \boldsymbol{\nabla}\times\big(f(x)\,\mathbf{P}_n^m(\vartheta,\varphi)\big)= \gamma_n\,\frac{f(x)}{r}\,\mathbf{C}_n^m(\vartheta,\varphi) 
\\
&\boldsymbol{\nabla}\times\big(f(x)\,\mathbf{B}_n^m(\vartheta,\varphi)\big) = -\left(\dot{f}(x)+\frac{f(x)}{x}\right)\,\mathbf{C}_n^m(\vartheta,\varphi)
\\
& \boldsymbol{\nabla}\times\big(f(x)\,\mathbf{C}_n^m(\vartheta,\varphi)\big) = \gamma_n\,\frac{f(x)}{r}\,\mathbf{P}_n^m(\vartheta,\varphi) + \left(\dot{f}(x) + \frac{f(x)}{x}\right)\,\mathbf{B}_n^m(\vartheta,\varphi)
\end{align}
$$ (V:11a)

```

### Identities

```{div} full-width

In the following, we shall consider the following identities for the integrals of the spherical harmonic vector over the unit sphere

$$
\int_{\Omega}\mathbf{P}_n^m(\vartheta,\varphi)\,\mathrm{d}\Omega = \frac{4\,\pi}{\sqrt{3}}\,\mathbf{p}_m\,\delta_{n,1}
\qquad
\int_{\Omega}\mathbf{B}_n^m(\vartheta,\varphi)\,\mathrm{d}\Omega = \frac{4\,\pi\,\sqrt{2}}{\sqrt{3}}\,\mathbf{p}_m\,\delta_{n,1}
\qquad
\int_{\Omega}\mathbf{C}_n^m(\vartheta,\varphi)\,\mathrm{d}\Omega =\mathbf{0}
$$ (V:12)

with

$$
\bb{p}_0 = \bn{x}_3 \qquad \bb{p}_{-1} = \bn{x}_1 \qquad \bb{p}_1 = \bn{x}_2
$$ (V:13)

We note also that

$$
\bb{P}_1^m(\vartheta,\varphi)+ \gamma_1\,\bb{B}_1^m = \bb{p}_m \qquad Y_1^m(\vartheta,\varphi)\,\bb{p}_m = \bn{r}(\vartheta,\varphi)
$$ (V:14)

Additional usefull identities are

$$
\int_{\Omega}\mathbf{P}_n^m(\vartheta,\varphi)\,Y_{k,q}(\vartheta,\varphi)\,\mathrm{d}\Omega = \frac{4\,\pi}{\sqrt{3}}\,\mathbf{p}^n_{m,q}\,\delta_{n,k+1}
$$ (V:15)

with $\bb{p}^n_{m,q}$ being constant vectors with $m=-n,\cdots,n$ and $q=-(n-1),\cdots,n-1$ at fixed $n$.

```
